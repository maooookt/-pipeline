```python
def now_str():
    return datetime.now().strftime("%Y-%m-%d %H:%M:%S")

def run_step(step_name, func, *args, **kwargs):
    try:
        func(*args, **kwargs)
        return True, f"{step_name}: OK"
    except NameError as e:
        return False, f"{step_name}: NameError (未定義関数/変数?) -> {e}"
    except subprocess.CalledProcessError as e:
        return False, f"{step_name}: CalledProcessError -> {e}"
    except Exception as e:
        return False, f"{step_name}: Exception -> {e}"

def ensure_dir(p):
    os.makedirs(p, exist_ok=True)
    return p

def append_results(rows, out_csv_path):
    ensure_dir(os.path.dirname(out_csv_path))
    df = pd.DataFrame(rows)
    header = not os.path.exists(out_csv_path)
    df.to_csv(out_csv_path, index=False, mode="a", header=header)

# --- 重要: 既存関数のシグネチャ微調整（※本体はあなたの実装を使う） ---
# make_rename_tcl に C_number を追加してください:
# def make_rename_tcl(input_list, value, C_number): ...

# --- 追加: パイプライン実行本体 ---
def run_pipeline_for_row(row, rules, token_kinds_set, out_csv_path, file_path):
    seq_id = row.ID
    started = now_str()

    try:
        input_list = ast.literal_eval(row.Clean_Sequence)
    except Exception as e:
        append_results([{
            "ID": seq_id, "Status": "X", "FailedStep": "parse_sequence",
            "Message": f"Clean_Sequenceのパース失敗: {e}",
            "StartedAt": started, "EndedAt": now_str(),
            "WorkDir": "", "Logs": ""
        }], out_csv_path)
        return

    if not all(tok in token_kinds_set for tok in seq_list):
        append_results([{
            "ID": seq_id, "Status": "X", "FailedStep": "token_check",
            "Message": f"未知トークンを含む: {set(seq_list) - token_kinds_set}",
            "StartedAt": started, "EndedAt": now_str(),
            "WorkDir": "", "Logs": ""
        }], out_csv_path)
        return

    N_number = row.newN_number
    C_number = row.newC_number
    Modifier = row.Modifier
    Patch    = row.Cyc_Patch

    today = datetime.today().strftime('%Y.%m.%d')
    value = f"{seq_id}_{today}"
    work_dir = f"/Users/teraimao/experiment/confirm/{value}_work"
    logs_dir = ensure_dir(os.path.join(work_dir, "logs"))

    steps = []

    ok, msg = run_step("copy_seed_work", copy_seed_work, value)
    steps.append(msg)
    if not ok: return append_results([{
        "ID": seq_id, "Status": "X", "FailedStep": "copy_seed_work",
        "Message": msg, "StartedAt": started, "EndedAt": now_str(),
        "WorkDir": work_dir, "Logs": "\n".join(steps)
    }], out_csv_path)

    ok, msg = run_step("copy_seed_pdb", copy_seed_pdb, value, input_list, C_number)
    steps.append(msg)
    if not ok: return append_results([{
        "ID": seq_id, "Status": "X", "FailedStep": "copy_seed_pdb",
        "Message": msg, "StartedAt": started, "EndedAt": now_str(),
        "WorkDir": work_dir, "Logs": "\n".join(steps)
    }], out_csv_path)

    ok, msg = run_step("make_rename_tcl", make_rename_tcl, input_list, value, C_number)
    steps.append(msg)
    if not ok: return append_results([{
        "ID": seq_id, "Status": "X", "FailedStep": "make_rename_tcl",
        "Message": msg, "StartedAt": started, "EndedAt": now_str(),
        "WorkDir": work_dir, "Logs": "\n".join(steps)
    }], out_csv_path)

    ok, msg = run_step("in_charminp", in_charminp, value)
    steps.append(msg)
    if not ok: return append_results([{
        "ID": seq_id, "Status": "X", "FailedStep": "in_charminp",
        "Message": msg, "StartedAt": started, "EndedAt": now_str(),
        "WorkDir": work_dir, "Logs": "\n".join(steps)
    }], out_csv_path)

    ok, msg = run_step("send_dir", send_dir, value)
    steps.append(msg)
    if not ok: return append_results([{
        "ID": seq_id, "Status": "X", "FailedStep": "send_dir",
        "Message": msg, "StartedAt": started, "EndedAt": now_str(),
        "WorkDir": work_dir, "Logs": "\n".join(steps)
    }], out_csv_path)

    ok, msg = run_step("return_copy_work", return_copy_work, value)
    steps.append(msg)
    if not ok:
        return append_results([{
            "ID": seq_id, "Status": "X", "FailedStep": "return_copy_work",
            "Message": msg, "StartedAt": started, "EndedAt": now_str(),
            "WorkDir": work_dir, "Logs": "\n".join(steps)
        }], out_csv_path)

    ok, msg = run_step("save_results", save_results, value)
    steps.append(msg)
    if not ok:
        return append_results([{
            "ID": seq_id, "Status": "X", "FailedStep": "save_results",
            "Message": msg, "StartedAt": started, "EndedAt": now_str(),
            "WorkDir": work_dir, "Logs": "\n".join(steps)
        }], out_csv_path)

    ok, msg = run_step("in_csv", in_csv, file_path)  # ★ file_path を引数で受け取るようにした
    steps.append(msg)
    if not ok:
        return append_results([{
            "ID": seq_id, "Status": "X", "FailedStep": "in_csv",
            "Message": msg, "StartedAt": started, "EndedAt": now_str(),
            "WorkDir": work_dir, "Logs": "\n".join(steps)
        }], out_csv_path)

    append_results([{
        "ID": seq_id, "Status": "O", "FailedStep": "",
        "Message": "All steps completed",
        "StartedAt": started, "EndedAt": now_str(),
        "WorkDir": work_dir, "Logs": "\n".join(steps)
    }], out_csv_path)
```

```python
##必要！
def run_pipeline_for_row(row, rules, token_kinds_set, out_csv_path, file_path, run_remote, 
                         t_end_ps, dt_ps, skip_ps):
    
    seq_id = row.ID
    started = now_str()
    steps = []

    # --- 1. Clean_Sequence → list パース ---
    try:
        input_list = ast.literal_eval(row.Clean_Sequence)
    except Exception as e:
        append_results([{"ID": seq_id, "Status": "X", "FailedStep": "parse_sequence", 
                         "Message": f"パース失敗: {e}", "StartedAt": started, 
                         "EndedAt": now_str(), "WorkDir": "", "Logs": ""}], out_csv_path)
        return

    # --- 2. トークン検査 (ここがご質問の該当箇所です) ---
    # input_list の全要素が token_kinds_set (kind) に含まれているかチェック
    unknown_tokens = [tok for tok in input_list if tok not in token_kinds_set]
    if unknown_tokens:
        # 未知のトークンが見つかった場合、エラーを記録して「return」でこの行を終了
        append_results([{"ID": seq_id, "Status": "X", "FailedStep": "token_check", 
                         "Message": f"未知トークンあり: {unknown_tokens}", 
                         "StartedAt": started, "EndedAt": now_str(), "WorkDir": "", "Logs": ""}], out_csv_path)
        print(f"   [SKIP] ID {seq_id}: 未知のアミノ酸が含まれているためスキップします。")
        return # これにより、次のアミノ酸（次の行）の処理へ移行します

    # --- 3. パラメータ準備 ---
    C_number = int(row.newC_number) if pd.notna(row.newC_number) else None
    N_number = int(row.newN_number) if pd.notna(row.newN_number) else None
    Modifier = ("" if pd.isna(row.Modifier) else str(row.Modifier)).strip()
    Patch    = ("" if pd.isna(row.Cyc_Patch) else str(row.Cyc_Patch)).strip()
    
    today = datetime.today().strftime('%Y.%m.%d')
    #value = f"{seq_id}_{today}"
    value = f"{seq_id}_{today}_{int(t_end_ps)}ps"
    work_dir = f"/Users/teraimao/experiment/confirm/{value}_work"
    ensure_dir(os.path.join(work_dir, "logs"))

    # --- 4. ローカル処理ステップ ---
    local_steps = [
        ("copy_seed_work", copy_seed_work, [value]),
        ("copy_seed_pdb", copy_seed_pdb, [value, input_list, C_number]),
        ("make_rename_tcl", make_rename_tcl, [input_list, value, C_number]),
        ("in_charminp", in_charminp, [value, input_list, C_number, N_number, Modifier, Patch, rules]),
        # ★ この関数の中で「全inp書き出し」と「統合Run.csh作成」を完結させる
        ("write_all_inputs_integrated", write_all_inputs_integrated, 
         [value, input_list, C_number, N_number, Modifier, Patch, rules, t_end_ps, dt_ps, skip_ps, refz_map, seq_id])
    ]

    for name, func, args in local_steps:
        ok, msg = run_step(name, func, *args)
        steps.append(msg)
        if not ok:
            append_results([{"ID": seq_id, "Status": "X", "FailedStep": name, "Message": msg, 
                             "StartedAt": started, "EndedAt": now_str(), "WorkDir": work_dir, 
                             "Logs": "\n".join(steps)}], out_csv_path)
            return

    # --- 5. リモート処理ステップ ---
    if run_remote:
        client = connect_server()
        if client is None:
            append_results([{"ID": seq_id, "Status": "X", "FailedStep": "SSH_Connect", "Message": "接続失敗", 
                             "StartedAt": started, "EndedAt": now_str(), "WorkDir": work_dir, 
                             "Logs": "\n".join(steps)}], out_csv_path)
            return
        
        try:
            ok, msg = run_step("send_dir", send_dir, value, client)
            steps.append(msg)
            if not ok:
                append_results([{"ID": seq_id, "Status": "X", "FailedStep": "send_dir", 
                                 "Message": msg, "StartedAt": started, "EndedAt": now_str(), 
                                 "WorkDir": work_dir, "Logs": "\n".join(steps)}], out_csv_path)
                return
                
        finally:
            client.close()
    else:
        steps.append("remote: SKIP")

    # --- 6. 最終検証と結果保存 ---
    if run_remote:
        # リモート実行の場合は、投入成功をもって一旦 "O" とする
        # (本当の成功判定は、後で結果を回収したときに行う)
        status = "O"
        msg = "Jobs submitted to server. Waiting for execution."
    else:
        # ローカル実行の場合のみ検証を行う
        ok_verify, verify_msg = verify_charmm_success(value)
        status = "O" if ok_verify else "X"
        msg = verify_msg if not ok_verify else "All steps completed"

    append_results([{
        "ID": seq_id, "Status": status, "FailedStep": "" if status == "O" else "verify_charmm_success",
        "Message": msg,
        "StartedAt": started, "EndedAt": now_str(),
        "WorkDir": work_dir, "Logs": "\n".join(steps)
    }], out_csv_path)
```

```python
def copy_seed_work(value):
    src = "/Users/teraimao/experiment/confirm/seed_work"
    dst = f"/Users/teraimao/experiment/confirm/{value}_work"

    shutil.copytree(src, dst, dirs_exist_ok=True)

    # ここで「コピー先だけ」ロック解除＆書き込みOKにする
    try:
        # ロック(uchgフラグ)を外す
        subprocess.run(["chflags", "-R", "nouchg", dst], check=False)
        # 念のため自分に書き込み権限を付ける
        subprocess.run(["chmod", "-R", "u+rwX", dst], check=False)
    except Exception as e:
        return False, f"copy_seed_work: copied, but failed to clear flags: {e}"

    return True, f"Seed merged into '{dst}'."
```

```python
vmd_executable = "/Applications/VMD.app/Contents/vmd/vmd_MACOSXARM64"  # VMDのフルパス

def make_rename_tcl(input_list, value, C_number):
    # 1. 前の関数から「2つの値」を受け取る
    ok_rename, rename_build_tcl = rename_pdb_atom(input_list, C_number)
    
    # rename_pdb_atom が失敗していたらここで止める
    if not ok_rename:
        return False, f"rename_pdb_atom failed: {rename_build_tcl}"
    
    # 作業ディレクトリとスクリプトの出力先を定義
    working_dir = f"/Users/teraimao/experiment/confirm/{value}_work/rename"
    destination_path = os.path.join(working_dir, "build.tcl")
    
    try:
        # 作業ディレクトリが存在しなければ作成
        if not os.path.exists(working_dir):
            os.makedirs(working_dir)
        
        # build.tclファイルを作成
        with open(destination_path, 'w') as file:
            file.write(rename_build_tcl)
        
        # VMDの実行
        log_path = os.path.join(working_dir, "log")
        with open(log_path, "w") as log_file:
            subprocess.run(
                [vmd_executable, "-dispdev", "text", "-e", destination_path],
                stdout=log_file, 
                stderr=subprocess.PIPE, 
                check=True, 
                cwd=working_dir
            )
        
        msg = f"VMD executed successfully, output saved to '{log_path}'."
        print(msg)
        # ★成功を返す
        return True, msg

    except subprocess.CalledProcessError as e:
        err_msg = f"Error occurred during VMD execution: {e.stderr.decode() if e.stderr else e}"
        print(err_msg)
        return False, err_msg
    except Exception as e:
        err_msg = f"An unexpected error occurred in make_rename_tcl: {e}"
        print(err_msg)
        return False, err_msg
```

```python
def send_dir(value, client):
    sftp = client.open_sftp()
    local_path = f"/Users/teraimao/experiment/confirm/{value}_work"
    remote_path = f"/home/teraimao/experiment/confirm/{value}_work"

    try:
        # 1. ディレクトリ準備
        client.exec_command(f"rm -rf {remote_path}")
        client.exec_command(f"mkdir -p {remote_path}")
        
        # 2. アップロード
        upload_directory(sftp, local_path, remote_path)
        
        # 3. PSF作成 (直接実行)
        ok_csh, msg_csh = run_csh(value, client)
        if not ok_csh: return False, f"PSF failed: {msg_csh}"

        # 4. GBSW MDジョブ投入 (ここで sbatch された Run.csh が解析まで全自動で行う)
        print(f"  [SUBMIT] MD & Analysis Integrated Jobs...")
        ok_md, msg_md = run_sh_gbsw(value, client)
        if not ok_md: return False, f"Job submission failed: {msg_md}"

        # ★ run_sh_hbond と run_sh_surf は不要になったので削除！
        return True, "All jobs submitted. MD and Analysis will run sequentially on the server."

    except Exception as e:
        return False, f"Error: {str(e)}"
    finally:
        sftp.close()
```

```python
def upload_directory(sftp, local_path, remote_path):
    """ディレクトリを再帰的にアップロードする"""
    try:
        sftp.mkdir(remote_path)
    except IOError:
        pass  # 既に存在する場合は無視

    for item in os.listdir(local_path):
        lp = os.path.join(local_path, item)
        rp = os.path.join(remote_path, item)
        if os.path.isdir(lp):
            upload_directory(sftp, lp, rp)
        else:
            sftp.put(lp, rp)
```

```python
def run_csh(value, client):
    """
    Run.csh を直接実行する。
    swork のディレクトリへ移動し、パスを通した上で実行。
    """
    remote_dir = f"/home/teraimao/experiment/confirm/{value}_work/charmm"
    
    # 実行コマンド: 
    # 1. cd で移動
    # 2. chmod で実行権限付与
    # 3. ./Run.csh で実行
    # ※ ログを log ファイルに書き出すようにしておくと、後で確認しやすいです
    command = f"cd {remote_dir} && chmod +x Run.csh && ./Run.csh > log 2>&1"
    
    print(f"  [EXEC] Running Run.csh in {remote_dir}...")
    
    stdin, stdout, stderr = client.exec_command(command)
    
    # 直接実行の場合、終わるまで待機して結果を受け取ります
    exit_status = stdout.channel.recv_exit_status() 
    
    if exit_status == 0:
        return True, "Run.csh completed successfully."
    else:
        err_msg = stderr.read().decode().strip()
        return False, f"Run.csh failed with status {exit_status}. Error: {err_msg}"
```

### CSVファイルのヘッダー構成

```python
# --- 設定：保存先のディレクトリとファイル名 ---
# --- 設定：保存先のディレクトリとファイル名 ---
SAVE_DIR = "/Users/teraimao/experiment/data/解析結果/"
MASTER_CSV_PATH = os.path.join(SAVE_DIR, "master_database.csv")

if not os.path.exists(SAVE_DIR):
    os.makedirs(SAVE_DIR)

def parse_charmm_energies(log_lines):
    vdw, coul, gbenr = np.nan, np.nan, np.nan
    for line in log_lines:
        if "LAVE EXTERN>" in line:
            parts = line.split()
            vdw, coul = float(parts[2]), float(parts[3])
        elif "LAVE PBEQ>" in line:
            parts = line.split(); gbenr = float(parts[4])
    return vdw, coul, gbenr

def run_aggregation_progressive(client, start_id, end_id, target_date):
    results = []
    search_date = "*" if target_date.upper() in ["", "ALL", "すべて"] else target_date
    valid_pattern = re.compile(r"^(\d+)_(\d{4}\.\d{2}\.\d{2})_(\d+)ps_work$")

    total_ids = end_id - start_id + 1
    processed_count = 0

    for seq_id in range(start_id, end_id + 1):
        # ★進捗を表示（これが出るのでフリーズしていないことがわかります）
        print(f"🔎 [{seq_id}/{end_id}] フォルダを検索中...", end="\r")
        
        cmd = f"ls -d /home/teraimao/experiment/confirm/{seq_id}_{search_date}_*ps_work 2>/dev/null"
        _, stdout, _ = client.exec_command(cmd)
        folder_paths = stdout.read().decode().splitlines()
        
        if not folder_paths:
            continue
        
        for path in folder_paths:
            folder_name = os.path.basename(path)
            match = valid_pattern.match(folder_name)
            if not match: continue
            
            extracted_id = int(match.group(1))
            extracted_date = match.group(2)
            t_end_ps = float(match.group(3))
            expected = int(t_end_ps * 0.25)

            for pos in range(1, 6):
                # A. Z座標取得
                z_cmd = f"grep 'refz' {path}/gbsw_{pos}/charmm.inp | awk '{{for(i=1;i<=NF;i++) if($i==\"refz\") print $(i+1)}}'"
                _, out, _ = client.exec_command(z_cmd)
                z_coord = out.read().decode().strip()

                # B. エネルギー平均取得
                _, out, _ = client.exec_command(f"grep -A 10 'LAVE>' {path}/gbsw_{pos}/log")
                vdw, coul, solv = parse_charmm_energies(out.read().decode().splitlines())

                # C. HBOND取得
                hb_path = f"{path}/analysis/ana_gbsw_{pos}/hbond/output.dat"
                _, out, _ = client.exec_command(f"awk '{{sum+=$2; n++}} END {{if(n>0) print sum/n, n; else print \"0 0\"}}' {hb_path}")
                hb_res = out.read().decode().split()
                hb_avg, hb_n = (float(hb_res[0]), int(hb_res[1])) if len(hb_res) >= 2 else (np.nan, 0)

                # D. SASA取得
                sasa_path = f"{path}/analysis/ana_gbsw_{pos}/surf_area/sasa_{pos}.dat"
                _, out, _ = client.exec_command(f"awk '{{sum+=$2; n++}} END {{if(n>0) print sum/n, n; else print \"0 0\"}}' {sasa_path}")
                sasa_res = out.read().decode().split()
                sasa_avg, sasa_n = (float(sasa_res[0]), int(sasa_res[1])) if len(sasa_res) >= 2 else (np.nan, 0)

                status = "Success" if (hb_n >= expected and sasa_n >= expected) else "Failed"

                results.append({
                    "ID": extracted_id, "Date": extracted_date, "Pos_Num": pos, "Z_Coord": z_coord,
                    "t_end_ps": t_end_ps, "Status": status, "HB_Avg": hb_avg,
                    "SASA_Avg": sasa_avg, "Solv_Free_Avg": solv, "vdW_Avg": vdw, "Coul_Avg": coul
                })
        
        # IDが1つ終わるごとに改行して進捗を確定表示
        print(f"✅ ID {seq_id} 完了 ({len(folder_paths)}フォルダ分)")

    if not results: return 0
    
    new_df = pd.DataFrame(results)
    if os.path.exists(MASTER_CSV_PATH):
        old_df = pd.read_csv(MASTER_CSV_PATH)
        final_df = pd.concat([old_df, new_df]).drop_duplicates(subset=['ID', 'Date', 'Pos_Num'], keep='last')
    else:
        final_df = new_df
    
    final_df.sort_values(['ID', 'Date', 'Pos_Num']).to_csv(MASTER_CSV_PATH, index=False)
    return len(results)
```

```python
def parse_charmm_energies(log_lines):
    """LAVEセクションから特定のエネルギー値を抽出する"""
    vdw, coul, gbenr = np.nan, np.nan, np.nan
    for line in log_lines:
        if line.startswith("LAVE EXTERN>"):
            parts = line.split()
            vdw = float(parts[2])  # VDWaals
            coul = float(parts[3]) # ELEC (Coulomb)
        elif line.startswith("LAVE PBEQ>"):
            parts = line.split()
            gbenr = float(parts[4]) # GBEnr (Solv Free)
    return vdw, coul, gbenr

def aggregate_to_master(client, start_id, end_id, today, master_csv="master_database.csv"):
    results = []
    
    for seq_id in range(start_id, end_id + 1):
        # 1. サーバー上のディレクトリを特定 (100ps/500ps自動判別)
        cmd = f"ls -d /home/teraimao/experiment/confirm/{seq_id}_{today}_*ps_work"
        _, stdout, _ = client.exec_command(cmd)
        folder_paths = stdout.read().decode().splitlines()
        
        if not folder_paths:
            print(f"ID {seq_id}: フォルダが見つかりません。スキップします。")
            continue
            
        for path in folder_paths:
            folder_name = os.path.basename(path)
            t_end_ps = float(re.search(r"(\d+)ps", folder_name).group(1))
            expected_lines = int((t_end_ps * 0.25) / 1.0) # 生産相25%分

            for pos in range(1, 6):
                # --- A. Z座標取得 ---
                z_cmd = f"grep 'refz' {path}/gbsw_{pos}/charmm.inp | awk '{{print $8}}'"
                _, stdout, _ = client.exec_command(z_cmd)
                z_coord = stdout.read().decode().strip()

                # --- B. エネルギー平均取得 (LAVE) ---
                # LAVE> から始まる10行程度を抜き出して解析
                ene_cmd = f"grep -A 10 'LAVE>' {path}/gbsw_{pos}/log"
                _, stdout, _ = client.exec_command(ene_cmd)
                vdw, coul, solv = parse_charmm_energies(stdout.read().decode().splitlines())

                # --- C. 解析データ平均 (HBOND, SASA) と行数チェック ---
                # HBOND
                hb_path = f"{path}/analysis/ana_gbsw_{pos}/hbond/output.dat"
                _, stdout, _ = client.exec_command(f"awk '{{sum+=$2; n++}} END {{if(n>0) print sum/n, n; else print \"0 0\"}}' {hb_path}")
                hb_res = stdout.read().decode().split()
                hb_avg, hb_count = float(hb_res[0]), int(hb_res[1])

                # SASA
                sasa_path = f"{path}/analysis/ana_gbsw_{pos}/surf_area/sasa_1.dat"
                _, stdout, _ = client.exec_command(f"awk '{{sum+=$2; n++}} END {{if(n>0) print sum/n, n; else print \"0 0\"}}' {sasa_path}")
                sasa_res = stdout.read().decode().split()
                sasa_avg, sasa_count = float(sasa_res[0]), int(sasa_res[1])

                # --- D. Status 判定 ---
                status = "Success" if (hb_count >= expected_lines and sasa_count >= expected_lines) else "Failed"

                results.append({
                    "ID": seq_id, "Date": today, "Pos_Num": pos, "Z_Coord": z_coord,
                    "t_end_ps": t_end_ps, "Status": status, "HB_Avg": hb_avg,
                    "SASA_Avg": sasa_avg, "Solv_Free_Avg": solv, "vdW_Avg": vdw, "Coul_Avg": coul
                })

    # CSVへの書き出し (追記モード)
    new_df = pd.DataFrame(results)
    if os.path.exists(master_csv):
        old_df = pd.read_csv(master_csv)
        final_df = pd.concat([old_df, new_df]).drop_duplicates(subset=['ID', 'Date', 'Pos_Num'], keep='last')
    else:
        final_df = new_df
    
    final_df.to_csv(master_csv, index=False)
    print(f"集計完了: {len(new_df)} 行のデータを {master_csv} に反映しました。")
```

## 機械学習用のcsvファイル作成

```python
# --- パス設定 ---
PATH_SMILES = "/Users/teraimao/experiment/data/all_csv_data/this_is_CycPeptMPDB.csv"
PATH_MASTER = "/Users/teraimao/experiment/data/解析結果/master_database.csv"
PATH_OUTPUT = "/Users/teraimao/experiment/data/解析結果/training_dataset_final.csv"

def build_integrated_dataset(start_id, end_id):
    if not os.path.exists(PATH_MASTER) or not os.path.exists(PATH_SMILES):
        return False, "❌ 入力ファイルが見つかりません。パスを確認してください。"

    # 1. データの読み込み
    df_md = pd.read_csv(PATH_MASTER)
    df_sl = pd.read_csv(PATH_SMILES)

    # 2. 指定範囲のIDでフィルタリング
    df_md = df_md[(df_md['ID'] >= start_id) & (df_md['ID'] <= end_id)]

    # 3. 異常値・欠損値のチェックと除外
    # StatusがSuccessでないもの、または数値がNaNの行を特定
    initial_count = df_md['ID'].nunique()
    df_md = df_md[df_md['Status'] == 'Success'].dropna(subset=['HB_Avg', 'SASA_Avg', 'Solv_Free_Avg', 
                                                               'vdW_Avg', 'Coul_Avg'])
    
    # 5つのPosが揃っていないIDも除外（解析不完全なため）
    pos_counts = df_md.groupby('ID').size()
    valid_ids = pos_counts[pos_counts == 5].index
    df_md = df_md[df_md['ID'].isin(valid_ids)]
    
    final_count = df_md['ID'].nunique()
    removed_count = initial_count - final_count

    # 4. MDデータの横持ち変換（Pivot）
    # Z座標の情報を列名に反映させるため、まず各PosのZ座標を取得（代表値として平均を使用）
    z_map = df_md.groupby('Pos_Num')['Z_Coord'].mean().to_dict()

    target_values = ['HB_Avg', 'SASA_Avg', 'Solv_Free_Avg', 'vdW_Avg', 'Coul_Avg']
    df_pivot = df_md.pivot(index='ID', columns='Pos_Num', values=target_values)
    
    # 列名の整形: 例 vdW_Avg_z0, vdW_Avg_z7 ...
    new_columns = []
    for val, pos in df_pivot.columns:
        z_val = int(z_map.get(pos, pos))
        new_columns.append(f"{val}_z{z_val}")
    df_pivot.columns = new_columns
    df_pivot = df_pivot.reset_index()

    # 5. SMILES台帳（this_is_CycPeptMPDB.csv）と結合
    # すでにLogP, MW, PSA, HBD/HBAが含まれているとのことなのでそのままMerge
    final_df = pd.merge(df_sl, df_pivot, on='ID', how='inner')

    # 6. 保存
    final_df.to_csv(PATH_OUTPUT, index=False)
    
    msg = f"✅ 統合完了: {len(final_df)}件のペプチドを保存しました。\n"
    if removed_count > 0:
        msg += f"⚠️ 異常/欠損により {removed_count}件 のIDを除外しました。"
    return True, msg

# --- UI定義 ---
style = {'description_width': 'initial'}
txt_start = widgets.IntText(value=1, description='開始ID:', style=style)
txt_end = widgets.IntText(value=7500, description='終了ID:', style=style)
btn_exec = widgets.Button(description='🚀 データセットを統合・生成', button_style='primary', 
                          layout={'width': '300px'})
out_log = widgets.Output()

def on_click_exec(b):
    with out_log:
        clear_output()
        print("⏳ 処理中...")
        success, msg = build_integrated_dataset(txt_start.value, txt_end.value)
        print(msg)
        if success:
            df_check = pd.read_csv(PATH_OUTPUT)
            display(df_check.head(3))

btn_exec.on_click(on_click_exec)
display(widgets.VBox([widgets.HTML("<h3>📊 Dataset Integration Tool</h3>"), 
                    widgets.HBox([txt_start, txt_end]), btn_exec, out_log]))
```

```python
#差分特徴量を追加した改良版コード  主要な物理量（エネルギーや水素結合数など）について、水環境（z0）と
#膜内部（z7, z15, z17, z20）の差分を自動計算して追加します。

# --- 1. データの読み込み ---
PATH_INPUT = "/Users/teraimao/experiment/data/解析結果/training_dataset_final.csv"
df = pd.read_csv(PATH_INPUT, low_memory=False)

def generate_fingerprints(smiles, radius=2, nBits=2048):
    if pd.isna(smiles): return [0] * nBits
    mol = Chem.MolFromSmiles(smiles)
    if mol is None: return [0] * nBits
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits=nBits)
    return list(fp)

print(f"⏳ 物理的差分（Delta）の計算を開始...")

# --- 2. 特徴量エンジニアリング ---

# A: MDデータの差分計算 (Delta Features)
# 主要な物理量に対して、z0（水）を基準とした差分を生成
md_bases = ['Solv_Free_Avg', 'HB_Avg', 'vdW_Avg', 'Coul_Avg', 'SASA_Avg']
z_points = [7, 15, 17, 20]

delta_features = pd.DataFrame(index=df.index)
for base in md_bases:
    z0_col = f"{base}_z0"
    if z0_col in df.columns:
        for z in z_points:
            z_col = f"{base}_z{z}"
            if z_col in df.columns:
                # 差分（Δ）を計算: 膜中 - 水中
                delta_name = f"Delta_{base}_z{z}_vs_z0"
                delta_features[delta_name] = df[z_col] - df[z0_col]

# B: 構造情報 (FP)
fps = df['SMILES'].apply(generate_fingerprints)
df_fps = pd.DataFrame(fps.tolist(), index=df.index).add_prefix('FP_')

# C: 指定された既存記述子
target_keywords = [
    'Sequence_LogP', 'PEOE_VSA', 'SlogP_VSA', 'EState_VSA', 
    'HeavyAtomCount', 'FractionCSP3', 'MolLogP', 'MolMR', 
    'CHCl3_3DPSA', 'H2O_3DPSA', 'PC1', 'PC2'
]
existing_descs = [col for col in df.columns if any(k in col for k in target_keywords)]

# 特徴量の結合 (FP + 元のMD + 差分MD + 既存記述子)
X = pd.concat([df_fps, df[[c for c in df.columns if '_z' in c]], delta_features, df[existing_descs]], axis=1)

# 数値型のみを抽出（PC1などのリスト形式除外）
X = X.select_dtypes(include=[np.number])

# 目的変数 (Permeability)
y = pd.to_numeric(df['Permeability'], errors='coerce')

# --- 3. クリーニングと欠損値補完 ---
mask = y.notna()
X_masked = X[mask]
y_clean = y[mask]

imputer = SimpleImputer(strategy='mean')
X_clean = pd.DataFrame(imputer.fit_transform(X_masked), columns=X_masked.columns)

print(f"✅ 特徴量生成完了。総特徴量数: {X_clean.shape[1]}個")

# --- 4. 学習と評価 ---
X_train, X_test, y_train, y_test = train_test_split(X_clean, y_clean, test_size=0.2, random_state=42)

model = RandomForestRegressor(n_estimators=300, max_features='sqrt', random_state=42, n_jobs=-1)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

# --- 5. 結果の数値表示 ---
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"\n📊 --- 差分追加版 評価結果 ---")
print(f"決定係数 (R2 Score): {r2:.3f}")
print(f"平均誤差 (RMSE): {rmse:.3f}")

# --- 6. 重要度の可視化 ---
plt.figure(figsize=(14, 8))
feat_importances = pd.Series(model.feature_importances_, index=X_clean.columns)
# 上位20項目を表示
feat_importances.sort_values(ascending=False).head(20).plot(kind='barh', color='darkorange')

plt.title('Top 20 Important Features (Including Delta MD Features)')
plt.xlabel('Importance Score')
plt.gca().invert_yaxis() # 重要度が高い順に上から並べる
plt.tight_layout()
plt.show()
```

```python
#後半100psのhbondの平均を求めるコード
def calculate_hbond_average(value):
    column_index = 1  # 水素結合数がある列（2列目）
    time_index = 0    # 計算時間がある列（1列目）
    peptide_positions = ["in", "sur", "out"]  # 各位置で処理
    directory = "hbond"  # hbond専用ディレクトリ名

    try:
        for position in peptide_positions:
            dir_path = f"/Users/teraimao/experiment/work/{value}_work/analysis/ana_gbsw_{position}/{directory}"

            # ディレクトリ内のファイルをリスト
            if not os.path.isdir(dir_path):
                print(f"Directory not found: {dir_path}")
                continue

            data_files = [f for f in os.listdir(dir_path) if os.path.isfile(os.path.join(dir_path, f))]
            if not data_files:
                print(f"No files found in directory: {dir_path}")
                continue

            # 最初のファイルを選択（複数ファイルがある場合は要確認）
            file_path = os.path.join(dir_path, data_files[0])
            values = []

            print(f"Reading hbond file: {file_path}")
            try:
                with open(file_path, 'r') as file:
                    # ヘッダー行をスキップ
                    next(file)

                    # データ行を読み取る
                    for line in file:
                        if not line.strip():  # 空行をスキップ
                            continue
                        columns = line.split()
                        time = float(columns[time_index])  # 時間列を取得
                        if time >= 400.0:  # 時間が40以上の場合のみ値を取得
                            values.append(float(columns[column_index]))
                            print(f"Valid hbond data: {line.strip()}")
            except FileNotFoundError:
                print(f"File not found: {file_path}")
                continue
            except ValueError:
                print(f"Invalid data format in file: {file_path}")
                continue

            # 平均を計算
            if values:
                average = sum(values) / len(values)
                print(f"Calculated hbond average: {average}")
            else:
                print(f"No valid hbond values found in {file_path} for averaging.")
                average = 0.0

            # 保存先ディレクトリを作成
            average_dir_path = os.path.join(dir_path, "average")
            os.makedirs(average_dir_path, exist_ok=True)

            # 結果を保存
            average_file_path = os.path.join(average_dir_path, "result.txt")
            try:
                with open(average_file_path, 'w') as avg_file:
                    avg_file.write(f"{average:.6f}\n")  # 小数点以下6桁で保存
                    print(f"Average hbond value saved to {average_file_path}")
            except Exception as e:
                print(f"Failed to write hbond average to {average_file_path}: {e}")

    except Exception as e:
        print(f"An unexpected error occurred: {e}")

```

### 後半100psの溶媒露出表面積の平均を求めるコード

```python
#後半10psの溶媒露出表面積の平均を求めるコード
def calculate_surf_average(value):
    column_indices = [1, 2]  # 2列目、3列目、4列目、5列目のインデックス
    time_index = 0  # 計算時間のインデックス
    peptide_positions = ["in", "sur", "out"]  # 各位置で処理
    directory = "surf_area"  # hbond専用ディレクトリ名

    try:
        for position in peptide_positions:
            dir_path = f"/Users/teraimao/experiment/work/{value}_work/analysis/ana_gbsw_{position}/{directory}"

            # ディレクトリの存在確認
            if not os.path.isdir(dir_path):
                print(f"Directory not found: {dir_path}")
                continue

            # ディレクトリ内のファイルをリストアップ
            data_files = [f for f in os.listdir(dir_path) if os.path.isfile(os.path.join(dir_path, f))]
            if not data_files:
                print(f"No files found in directory: {dir_path}")
                continue

            # 最初のファイルを選択（複数ファイルがある場合は要確認）
            file_path = os.path.join(dir_path, data_files[0])
            column_values = {index: [] for index in column_indices}

            with open(file_path, 'r') as file:
                next(file)  # ヘッダーをスキップ
                for line in file:
                    if not line.strip():  # 空行をスキップ
                        continue
                    columns = line.split()
                    time = float(columns[time_index])  # 時間列を取得
                    if time >= 400.0:  # 時間が40以上の場合のみ値を取得
                        for index in column_indices:
                            column_values[index].append(float(columns[index]))

            # 各列の平均を計算
            averages = {
                index: (sum(column_values[index]) / len(column_values[index])) if column_values[index] else 0.0
                for index in column_indices
            }

            # データがなかった場合の処理
            if not any(column_values.values()):
                print(f"No valid hbond values found in {file_path} for averaging.")
                continue

            # "average" ディレクトリを作成
            average_dir_path = os.path.join(dir_path, "average")
            os.makedirs(average_dir_path, exist_ok=True)

            # 結果を保存
            average_file_path = os.path.join(average_dir_path, "result.txt")
            try:
                with open(average_file_path, 'w') as avg_file:
                    for index, avg in averages.items():
                        avg_file.write(f"Column {index+1} average: {avg:.6f}\n")
                    print(f"Averages saved to {average_file_path}")
            except Exception as e:
                print(f"Failed to write averages to {average_file_path}: {e}")

    except Exception as e:
        print(f"An unexpected error occurred: {e}")
```

```python
#平均値を計算する必要のない表面積を解析ディレクトリに入れる
import os
from collections import OrderedDict

def extract_total_from_log(log_file, target_commands):
    """ログファイルから指定されたターゲットコマンドの次の `total =` の値を抽出する"""
    target_areas = OrderedDict((cmd, None) for cmd in target_commands)  # 順番を保証する辞書

    try:
        with open(log_file, 'r') as file:
            lines = file.readlines()
            for i, line in enumerate(lines):
                for key in target_areas.keys():
                    if key in line:  # ターゲットコマンドを検出
                        for j in range(i + 1, len(lines)):
                            if "total  =" in lines[j]:
                                target_areas[key] = lines[j].split()[-1]  # 値を取得
                                break
        return target_areas
    except FileNotFoundError:
        print(f"Log file not found: {log_file}")
        return None
    except Exception as e:
        print(f"Error reading log file: {e}")
        return None

def save_results(value):
    peptide_positions = ["in", "sur", "out"]
    target_commands = [
        "COOR SURF RPROBE 1.4 SELE all END",  # surf_area 用
        "COOR SURF RPROBE 1.4 SELE type O* .or. type N* END",  # polar_surf_area 用
    ]

    for position in peptide_positions:
        # パスの定義
        log_file = f"/Users/teraimao/experiment/work/{value}_work/gbsw_{position}/log"
        dir_polar = f"/Users/teraimao/experiment/work/{value}_work/analysis/ana_gbsw_{position}/polar_surf_area/average"
        dir_surf = f"/Users/teraimao/experiment/work/{value}_work/analysis/ana_gbsw_{position}/surf_area/average"

        # ログファイルからデータを抽出
        target_areas = extract_total_from_log(log_file, target_commands)
        if not target_areas:
            print(f"Skipping {log_file} due to missing or invalid data.")
            continue

        # 値を保存
        if None not in target_areas.values():  # 全てのターゲットが見つかった場合
            os.makedirs(dir_polar, exist_ok=True)
            os.makedirs(dir_surf, exist_ok=True)

            # 正しい対応関係で保存する
            with open(os.path.join(dir_surf, "result.txt"), 'w') as file_surf:
                file_surf.write(target_areas["COOR SURF RPROBE 1.4 SELE all END"] + "\n")

            with open(os.path.join(dir_polar, "result.txt"), 'w') as file_polar:
                file_polar.write(target_areas["COOR SURF RPROBE 1.4 SELE type O* .or. type N* END"] + "\n")

            print(f"Values saved successfully for {position}:")
            print(f" - {target_areas['COOR SURF RPROBE 1.4 SELE all END']} saved to {os.path.join(dir_surf, 'result.txt')}")
            print(f" - {target_areas['COOR SURF RPROBE 1.4 SELE type O* .or. type N* END']} saved to {os.path.join(dir_polar, 'result.txt')}")
        else:
            print(f"Error: Missing data for some target commands in {log_file}.")
```
