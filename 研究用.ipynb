```python
def now_str():
    return datetime.now().strftime("%Y-%m-%d %H:%M:%S")

def run_step(step_name, func, *args, **kwargs):
    try:
        func(*args, **kwargs)
        return True, f"{step_name}: OK"
    except NameError as e:
        return False, f"{step_name}: NameError (未定義関数/変数?) -> {e}"
    except subprocess.CalledProcessError as e:
        return False, f"{step_name}: CalledProcessError -> {e}"
    except Exception as e:
        return False, f"{step_name}: Exception -> {e}"

def ensure_dir(p):
    os.makedirs(p, exist_ok=True)
    return p

def append_results(rows, out_csv_path):
    ensure_dir(os.path.dirname(out_csv_path))
    df = pd.DataFrame(rows)
    header = not os.path.exists(out_csv_path)
    df.to_csv(out_csv_path, index=False, mode="a", header=header)

# --- 重要: 既存関数のシグネチャ微調整（※本体はあなたの実装を使う） ---
# make_rename_tcl に C_number を追加してください:
# def make_rename_tcl(input_list, value, C_number): ...

# --- 追加: パイプライン実行本体 ---
def run_pipeline_for_row(row, rules, token_kinds_set, out_csv_path, file_path):
    seq_id = row.ID
    started = now_str()

    try:
        input_list = ast.literal_eval(row.Clean_Sequence)
    except Exception as e:
        append_results([{
            "ID": seq_id, "Status": "X", "FailedStep": "parse_sequence",
            "Message": f"Clean_Sequenceのパース失敗: {e}",
            "StartedAt": started, "EndedAt": now_str(),
            "WorkDir": "", "Logs": ""
        }], out_csv_path)
        return

    if not all(tok in token_kinds_set for tok in seq_list):
        append_results([{
            "ID": seq_id, "Status": "X", "FailedStep": "token_check",
            "Message": f"未知トークンを含む: {set(seq_list) - token_kinds_set}",
            "StartedAt": started, "EndedAt": now_str(),
            "WorkDir": "", "Logs": ""
        }], out_csv_path)
        return

    N_number = row.newN_number
    C_number = row.newC_number
    Modifier = row.Modifier
    Patch    = row.Cyc_Patch

    today = datetime.today().strftime('%Y.%m.%d')
    value = f"{seq_id}_{today}"
    work_dir = f"/Users/teraimao/experiment/confirm/{value}_work"
    logs_dir = ensure_dir(os.path.join(work_dir, "logs"))

    steps = []

    ok, msg = run_step("copy_seed_work", copy_seed_work, value)
    steps.append(msg)
    if not ok: return append_results([{
        "ID": seq_id, "Status": "X", "FailedStep": "copy_seed_work",
        "Message": msg, "StartedAt": started, "EndedAt": now_str(),
        "WorkDir": work_dir, "Logs": "\n".join(steps)
    }], out_csv_path)

    ok, msg = run_step("copy_seed_pdb", copy_seed_pdb, value, input_list, C_number)
    steps.append(msg)
    if not ok: return append_results([{
        "ID": seq_id, "Status": "X", "FailedStep": "copy_seed_pdb",
        "Message": msg, "StartedAt": started, "EndedAt": now_str(),
        "WorkDir": work_dir, "Logs": "\n".join(steps)
    }], out_csv_path)

    ok, msg = run_step("make_rename_tcl", make_rename_tcl, input_list, value, C_number)
    steps.append(msg)
    if not ok: return append_results([{
        "ID": seq_id, "Status": "X", "FailedStep": "make_rename_tcl",
        "Message": msg, "StartedAt": started, "EndedAt": now_str(),
        "WorkDir": work_dir, "Logs": "\n".join(steps)
    }], out_csv_path)

    ok, msg = run_step("in_charminp", in_charminp, value)
    steps.append(msg)
    if not ok: return append_results([{
        "ID": seq_id, "Status": "X", "FailedStep": "in_charminp",
        "Message": msg, "StartedAt": started, "EndedAt": now_str(),
        "WorkDir": work_dir, "Logs": "\n".join(steps)
    }], out_csv_path)

    ok, msg = run_step("send_dir", send_dir, value)
    steps.append(msg)
    if not ok: return append_results([{
        "ID": seq_id, "Status": "X", "FailedStep": "send_dir",
        "Message": msg, "StartedAt": started, "EndedAt": now_str(),
        "WorkDir": work_dir, "Logs": "\n".join(steps)
    }], out_csv_path)

    ok, msg = run_step("return_copy_work", return_copy_work, value)
    steps.append(msg)
    if not ok:
        return append_results([{
            "ID": seq_id, "Status": "X", "FailedStep": "return_copy_work",
            "Message": msg, "StartedAt": started, "EndedAt": now_str(),
            "WorkDir": work_dir, "Logs": "\n".join(steps)
        }], out_csv_path)

    ok, msg = run_step("save_results", save_results, value)
    steps.append(msg)
    if not ok:
        return append_results([{
            "ID": seq_id, "Status": "X", "FailedStep": "save_results",
            "Message": msg, "StartedAt": started, "EndedAt": now_str(),
            "WorkDir": work_dir, "Logs": "\n".join(steps)
        }], out_csv_path)

    ok, msg = run_step("in_csv", in_csv, file_path)  # ★ file_path を引数で受け取るようにした
    steps.append(msg)
    if not ok:
        return append_results([{
            "ID": seq_id, "Status": "X", "FailedStep": "in_csv",
            "Message": msg, "StartedAt": started, "EndedAt": now_str(),
            "WorkDir": work_dir, "Logs": "\n".join(steps)
        }], out_csv_path)

    append_results([{
        "ID": seq_id, "Status": "O", "FailedStep": "",
        "Message": "All steps completed",
        "StartedAt": started, "EndedAt": now_str(),
        "WorkDir": work_dir, "Logs": "\n".join(steps)
    }], out_csv_path)
```

```python
def run_step_if_exists(step_name, func_name: str, *args, **kwargs):
    """
    func_name が globals() にあれば実行して run_step と同じ戻り値。
    なければ (True, f"{step_name}: SKIP (未定義)") を返す。
    """
    func = globals().get(func_name)
    if func is None:
        return True, f"{step_name}: SKIP (関数未定義)"
    return run_step(step_name, func, *args, **kwargs)

# === ユーティリティ ===
def now_str():
    return datetime.now().strftime("%Y-%m-%d %H:%M:%S")

#def run_step(step_name, func, *args, **kwargs):
#    try:
#        func(*args, **kwargs)
#        return True, f"{step_name}: OK"
#    except NameError as e:
#        return False, f"{step_name}: NameError (未定義関数/変数?) -> {e}"
#    except subprocess.CalledProcessError as e:
#        return False, f"{step_name}: CalledProcessError -> {e}"
#    except Exception as e:
#        return False, f"{step_name}: Exception -> {e}"

def ensure_dir(p):
    os.makedirs(p, exist_ok=True)
    return p

def append_results(rows, out_csv_path):
    ensure_dir(os.path.dirname(out_csv_path))
    df = pd.DataFrame(rows)
    header = not os.path.exists(out_csv_path)
    df.to_csv(out_csv_path, index=False, mode="a", header=header)


def verify_charmm_success(value):
    """
    ローカル confirm/{value}_work/charmm を検査して
    CHARMM が正常終了したかを判定する。
    """
    d = f"/Users/teraimao/experiment/confirm/{value}_work/charmm"
    log_path = os.path.join(d, "log")
    pdb_path = os.path.join(d, "peptide.pdb")
    psf_path = os.path.join(d, "peptide.psf")

    # ファイル存在・サイズ
    for p in [log_path, pdb_path, psf_path]:
        if not os.path.isfile(p) or os.path.getsize(p) == 0:
            return False, f"missing_or_empty: {os.path.basename(p)}"

    # ログの NORMAL TERMINATION チェック
    try:
        with open(log_path, "r", errors="ignore") as f:
            text = f.read()
    except Exception as e:
        return False, f"log_read_error: {e}"

    if re.search(r"NORMAL TERMINATION", text, re.IGNORECASE):
        # 任意: BOMBS を軽くチェック（強すぎる閾値は好みで調整）
        bombs = re.findall(r"\bBOMBS\b", text)
        if len(bombs) > 100:  # 適当なしきい値
            return False, f"too_many_BOMBS: {len(bombs)}"
        return True, "NORMAL TERMINATION"
    else:
        return False, "no_NORMAL_TERMINATION_in_log"

def run_step(name, func, *args):
    try:
        ok, msg = func(*args)
        return ok, f"[{name}] {msg}"
    except Exception as e:
        return False, f"[{name}] Failed: {str(e)}"

```

```python
##必要！
def run_pipeline_for_row(row, rules, token_kinds_set, out_csv_path, file_path, run_remote, 
                         t_end_ps, dt_ps, skip_ps):
    
    seq_id = row.ID
    started = now_str()
    steps = []

    # --- 1. Clean_Sequence → list パース ---
    try:
        input_list = ast.literal_eval(row.Clean_Sequence)
    except Exception as e:
        append_results([{"ID": seq_id, "Status": "X", "FailedStep": "parse_sequence", 
                         "Message": f"パース失敗: {e}", "StartedAt": started, 
                         "EndedAt": now_str(), "WorkDir": "", "Logs": ""}], out_csv_path)
        return

    # --- 2. トークン検査 (ここがご質問の該当箇所です) ---
    # input_list の全要素が token_kinds_set (kind) に含まれているかチェック
    unknown_tokens = [tok for tok in input_list if tok not in token_kinds_set]
    if unknown_tokens:
        # 未知のトークンが見つかった場合、エラーを記録して「return」でこの行を終了
        append_results([{"ID": seq_id, "Status": "X", "FailedStep": "token_check", 
                         "Message": f"未知トークンあり: {unknown_tokens}", 
                         "StartedAt": started, "EndedAt": now_str(), "WorkDir": "", "Logs": ""}], out_csv_path)
        print(f"   [SKIP] ID {seq_id}: 未知のアミノ酸が含まれているためスキップします。")
        return # これにより、次のアミノ酸（次の行）の処理へ移行します

    # --- 3. パラメータ準備 ---
    C_number = int(row.newC_number) if pd.notna(row.newC_number) else None
    N_number = int(row.newN_number) if pd.notna(row.newN_number) else None
    Modifier = ("" if pd.isna(row.Modifier) else str(row.Modifier)).strip()
    Patch    = ("" if pd.isna(row.Cyc_Patch) else str(row.Cyc_Patch)).strip()
    
    today = datetime.today().strftime('%Y.%m.%d')
    value = f"{seq_id}_{today}"
    work_dir = f"/Users/teraimao/experiment/confirm/{value}_work"
    ensure_dir(os.path.join(work_dir, "logs"))

    # --- 4. ローカル処理ステップ ---
    # write_all_gbsw_inputs 等に UIから取得した t_end_ps 等を渡すように設定
    local_steps = [
        ("copy_seed_work", copy_seed_work, [value]),
        ("copy_seed_pdb", copy_seed_pdb, [value, input_list, C_number]),
        ("make_rename_tcl", make_rename_tcl, [input_list, value, C_number]),
        ("in_charminp", in_charminp, [value, input_list, C_number, N_number, Modifier, Patch, rules]),
        ("write_all_gbsw_inputs", write_all_gbsw_inputs, [value, input_list, C_number, N_number, 
                                                          Modifier, Patch, rules, t_end_ps, dt_ps, refz_map]),
        ("write_all_hbond_inputs", write_all_hbond_inputs, [value, t_end_ps, dt_ps, skip_ps]),
        ("write_all_sasa_inputs", write_all_sasa_inputs, [value, t_end_ps, dt_ps, skip_ps])
    ]

    for name, func, args in local_steps:
        ok, msg = run_step(name, func, *args)
        steps.append(msg)
        if not ok:
            append_results([{"ID": seq_id, "Status": "X", "FailedStep": name, "Message": msg, 
                             "StartedAt": started, "EndedAt": now_str(), "WorkDir": work_dir, 
                             "Logs": "\n".join(steps)}], out_csv_path)
            return

    # --- 5. リモート処理ステップ ---
    if run_remote:
        client = connect_server()
        if client is None:
            append_results([{"ID": seq_id, "Status": "X", "FailedStep": "SSH_Connect", "Message": "接続失敗", 
                             "StartedAt": started, "EndedAt": now_str(), "WorkDir": work_dir, 
                             "Logs": "\n".join(steps)}], out_csv_path)
            return
        
        try:
            ok, msg = run_step("send_dir", send_dir, value, client)
            steps.append(msg)
            if not ok:
                append_results([{"ID": seq_id, "Status": "X", "FailedStep": "send_dir", 
                                 "Message": msg, "StartedAt": started, "EndedAt": now_str(), 
                                 "WorkDir": work_dir, "Logs": "\n".join(steps)}], out_csv_path)
                return

            ok, msg = run_step("return_copy_work", return_copy_work, value, client)
            steps.append(msg)
            if not ok:
                append_results([{"ID": seq_id, "Status": "X", "FailedStep": "return_copy_work", 
                                 "Message": msg, "StartedAt": started, "EndedAt": now_str(), 
                                 "WorkDir": work_dir, "Logs": "\n".join(steps)}], out_csv_path)
                return
        finally:
            client.close()
    else:
        steps.append("remote: SKIP")

    # --- 6. 最終検証と結果保存 ---
    ok_verify, verify_msg = verify_charmm_success(value)
    steps.append(f"verify_charmm_success: {verify_msg}")
    
    status = "O" if ok_verify else "X"
    append_results([{
        "ID": seq_id, "Status": status, "FailedStep": "" if ok_verify else "verify_charmm_success",
        "Message": "All steps completed" if ok_verify else verify_msg,
        "StartedAt": started, "EndedAt": now_str(),
        "WorkDir": work_dir, "Logs": "\n".join(steps)
    }], out_csv_path)
```

### copy_seed_work(value) seed_workからコピーする 

```python
def copy_seed_work(value):
    src = "/Users/teraimao/experiment/confirm/seed_work"
    dst = f"/Users/teraimao/experiment/confirm/{value}_work"

    shutil.copytree(src, dst, dirs_exist_ok=True)

    # ここで「コピー先だけ」ロック解除＆書き込みOKにする
    try:
        # ロック(uchgフラグ)を外す
        subprocess.run(["chflags", "-R", "nouchg", dst], check=False)
        # 念のため自分に書き込み権限を付ける
        subprocess.run(["chmod", "-R", "u+rwX", dst], check=False)
    except Exception as e:
        return False, f"copy_seed_work: copied, but failed to clear flags: {e}"

    return True, f"Seed merged into '{dst}'."
```

### make_rename_tcl build.tcl実行

```python
vmd_executable = "/Applications/VMD.app/Contents/vmd/vmd_MACOSXARM64"  # VMDのフルパス

def make_rename_tcl(input_list, value,C_number):
    # VMDスクリプト内容を生成
    rename_build_tcl = rename_pdb_atom(input_list,C_number)
    
    # 作業ディレクトリとスクリプトの出力先を定義
    working_dir = f"/Users/teraimao/experiment/confirm/{value}_work/rename"
    destination_path = os.path.join(working_dir, "build.tcl")
    
    # 作業ディレクトリが存在しなければ作成
    if not os.path.exists(working_dir):
        os.makedirs(working_dir)
    
    # build.tclファイルを作成
    with open(destination_path, 'w') as file:
        file.write(rename_build_tcl)
    
    # build.tclファイルが正しく作成されたか確認
    if not os.path.exists(destination_path):
        print(f"Error: {destination_path} が存在しません。ファイルパスを確認してください。")
        return
    
    # VMDの実行
    try:
        log_path = os.path.join(working_dir, "log")
        with open(log_path, "w") as log_file:
            subprocess.run(
                [vmd_executable, "-dispdev", "text", "-e", destination_path],
                stdout=log_file, stderr=subprocess.PIPE, check=True, cwd=working_dir
            )
        print(f"VMD executed successfully, and output saved to '{log_path}'.")
    except subprocess.CalledProcessError as e:
        print(f"Error occurred during VMD execution: {e.stderr}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
```

### 分子内水素結合のcharmm.inp

```python
def hbond_charmm_inp(
    pos: str,
    t_end_ps: float,
    dt_ps: float,
    skip_ps: float,
    segid: str = "PROA",
    dist: float = 3.0,
    angle: float = 120.0,
    out_name: str = "output.dat",
):
    """
    CHARMM: 分子内水素結合数（?NHBOND）を time(ps) と一緒に output.dat に書く inp を生成
    前提:
      - PSF: ../../../gbsw_{pos}/peptide_{pos}.psf
      - DCD: ../../../gbsw_{pos}/gbsw_{pos}.dcd
      - dt_ps は「DCDのフレーム間隔(ps/frame)」
      - skip_ps は「最初に飛ばす時間(ps)」
    """
    pos = str(pos)

    if dt_ps <= 0:
        raise ValueError("dt_ps must be > 0")
    if t_end_ps <= 0:
        raise ValueError("t_end_ps must be > 0")
    if skip_ps < 0:
        skip_ps = 0.0

    # skip(ps) -> skip(frames)
    skip_frames = int(round(skip_ps / dt_ps))
    if skip_frames < 0:
        skip_frames = 0

    return f"""\
open read card name ../../../toppar/4.1top_all36_prot.rtf unit 10
read rtf card unit 10
close unit 10

open read card name ../../../toppar/2_1gbsw_par_all36m_prot.prm unit 10
read parameter card unit 10
close unit 10

open read card name ../../../gbsw_{pos}/peptide_{pos}.psf unit 10
read psf card unit 10
close unit 10

open read unform name ../../../gbsw_{pos}/gbsw_{pos}.dcd unit 20
traj firstu 20 nunit 1 skip {skip_frames}

open write card name {out_name} unit 30
write title unit 30
* Time(ps) NHBOND
*

set time 0.0
set incr {dt_ps}

label loop
traj read
if ?nstep eq 0 goto endloop
if ?nfile eq 0 goto endloop

coor hbond dist {dist:.3f} angle {angle:.1f} -
     select segid {segid} .and. hydrogen end -
     select segid {segid} .and. (type O* .or. type N*) end

set nhbond ?NHBOND

write title unit 30
* @time @nhbond
*

incr time by @incr
if time lt {t_end_ps} goto loop

label endloop
close unit 30
stop
"""

```

### 溶媒露出表面積のcharmm.inp

```python
def sasa_charmm_inp(pos: str, t_end_ps: float, dt_ps: float, skip_ps: float,
                    segid: str = "PROA",
                    rprobe: float = 1.4,
                    out_name: str = "output.dat"):
    """
    CHARMM: total ASA と 3D-PSA を trajectory から計算する inp を生成
    - pos: "1"〜"5"（gbsw_{pos} を参照）
    - t_end_ps: 解析終端 (ps)
    - dt_ps: DCD の保存間隔 (ps/frame) → time の増分にも使う
    - skip_ps: 解析開始まで飛ばす時間 (ps)（DCD の frame skip に変換）
    """
    pos = str(pos)

    if t_end_ps <= 0:
        raise ValueError("t_end_ps must be > 0")
    if dt_ps <= 0:
        raise ValueError("dt_ps must be > 0")
    if skip_ps < 0:
        raise ValueError("skip_ps must be >= 0")

    # skip(ps) → skip(frames)
    skip_frames = int(round(skip_ps / dt_ps))
    if skip_frames < 0:
        skip_frames = 0

    return f"""\
open read card name ../../../toppar/4.1top_all36_prot.rtf unit 10
read rtf card unit 10
close unit 10

open read card name ../../../toppar/2_1gbsw_par_all36m_prot.prm unit 10
read parameter card unit 10
close unit 10

open read card name ../../../gbsw_{pos}/peptide_{pos}.psf unit 10
read psf card unit 10
close unit 10

open read unform name ../../../gbsw_{pos}/gbsw_{pos}.dcd unit 20
traj firstu 20 nunit 1 skip {skip_frames}

open write card name {out_name} unit 30
write title unit 30
* Time(ps) total_asa 3D-PSA
*

set time 0.0
set incr {dt_ps}

label loop
traj read
if ?nstep eq 0 goto endloop
if ?nfile eq 0 goto endloop

! total ASA
coor surf rprobe {rprobe} sele all end
scalar wmain statistics sele all end
set totalasa ?STOT

BOMLEV -1

! 3D-PSA (O/N only)
coor surf rprobe {rprobe} acce sele (type O* .or. type N*) end
scalar wmain statistics sele (type O* .or. type N*) end
if ?NSEL gt 0 set 3DPSA ?STOT
if ?NSEL eq 0 set 3DPSA 0.0
...
* @time @totalasa @3DPSA
*

incr time by @incr
if time lt {t_end_ps} goto loop

label endloop
close unit 30
stop
"""
```

```python
#コピーディレクトリをサーバーに送って、一気に全ジョブを実行する関数
def send_dir(value, client):
    # clientはメイン関数から渡されるので、ここでのconnect_serverは不要です
    sftp = client.open_sftp()
    
    local_path = f"/Users/teraimao/experiment/confirm/{value}_work"
    # ★ /swork に修正
    remote_path = f"/swork/teraimao/experiment/confirm/{value}_work"

    try:
        # 1. リモートディレクトリの準備（既存のを消して作り直すのが最も安全）
        client.exec_command(f"rm -rf {remote_path}")
        client.exec_command(f"mkdir -p {remote_path}")
        
        # 2. ファイルのアップロード
        print(f"  [UPLOAD] Mac -> Server: {remote_path}")
        upload_directory(sftp, local_path, remote_path)
        
        # 3. PSF作成 (Run.csh を直接実行)
        # 前のやり取りの通り、./Run.csh で実行して完了を待ちます
        print(f"  [EXEC] Step1: PSF generation (Run.csh)...")
        ok_csh, msg_csh = run_csh(value, client)
        if not ok_csh:
            return False, f"PSF generation failed: {msg_csh}"

        # 4. GBSW MDジョブ投入 (sbatch)
        print(f"  [SUBMIT] Step2: GBSW MD jobs...")
        ok_md, msg_md = run_sh_gbsw(value, client)
        if not ok_md:
            return False, f"MD submission failed: {msg_md}"

        # 5. 解析実行 (hbond / sasa)
        # ※ MDが終わる前に解析を投げてもエラーになる場合は、MDの終了を待つ必要があります
        print(f"  [EXEC] Step3: Analysis (hbond / sasa)...")
        run_sh_hbond(value, client)
        run_sh_surf(value, client)

        return True, "Upload and all jobs/scripts triggered successfully"

    except Exception as e:
        return False, f"Error in send_dir sequence: {str(e)}"
    finally:
        sftp.close() # client.close() はメイン関数で行うのでここでは閉じない
```

def upload_directory(sftp, local_path, remote_path):
    for item in os.listdir(local_path):

        remote_item = os.path.join(remote_path, item)

        if os.path.isdir(local_item):
            try:
                sftp.stat(remote_item)
            except FileNotFoundError:
                sftp.mkdir(remote_item)
            upload_directory(sftp, local_item, remote_item)
        else:
            sftp.put(local_item, remote_item)

```python
def upload_directory(sftp, local_path, remote_path):
    """ディレクトリを再帰的にアップロードする"""
    try:
        sftp.mkdir(remote_path)
    except IOError:
        pass  # 既に存在する場合は無視

    for item in os.listdir(local_path):
        lp = os.path.join(local_path, item)
        rp = os.path.join(remote_path, item)
        if os.path.isdir(lp):
            upload_directory(sftp, lp, rp)
        else:
            sftp.put(lp, rp)
```

```python
def run_csh(value, client):
    """
    Run.csh を直接実行する。
    swork のディレクトリへ移動し、パスを通した上で実行。
    """
    remote_dir = f"/swork/teraimao/experiment/confirm/{value}_work/charmm"
    
    # 実行コマンド: 
    # 1. cd で移動
    # 2. chmod で実行権限付与
    # 3. ./Run.csh で実行
    # ※ ログを log ファイルに書き出すようにしておくと、後で確認しやすいです
    command = f"cd {remote_dir} && chmod +x Run.csh && ./Run.csh > log 2>&1"
    
    print(f"  [EXEC] Running Run.csh in {remote_dir}...")
    
    stdin, stdout, stderr = client.exec_command(command)
    
    # 直接実行の場合、終わるまで待機して結果を受け取ります
    exit_status = stdout.channel.recv_exit_status() 
    
    if exit_status == 0:
        return True, "Run.csh completed successfully."
    else:
        err_msg = stderr.read().decode().strip()
        return False, f"Run.csh failed with status {exit_status}. Error: {err_msg}"
```

```python
#後半10psの溶媒露出表面積の平均を求めるコード
def calculate_surf_average(value):
    column_indices = [1, 2]  # 2列目、3列目、4列目、5列目のインデックス
    time_index = 0  # 計算時間のインデックス
    peptide_positions = ["in", "sur", "out"]  # 各位置で処理
    directory = "surf_area"  # hbond専用ディレクトリ名

    try:
        for position in peptide_positions:
            dir_path = f"/Users/teraimao/experiment/work/{value}_work/analysis/ana_gbsw_{position}/{directory}"

            # ディレクトリの存在確認
            if not os.path.isdir(dir_path):
                print(f"Directory not found: {dir_path}")
                continue

            # ディレクトリ内のファイルをリストアップ
            data_files = [f for f in os.listdir(dir_path) if os.path.isfile(os.path.join(dir_path, f))]
            if not data_files:
                print(f"No files found in directory: {dir_path}")
                continue

            # 最初のファイルを選択（複数ファイルがある場合は要確認）
            file_path = os.path.join(dir_path, data_files[0])
            column_values = {index: [] for index in column_indices}

            with open(file_path, 'r') as file:
                next(file)  # ヘッダーをスキップ
                for line in file:
                    if not line.strip():  # 空行をスキップ
                        continue
                    columns = line.split()
                    time = float(columns[time_index])  # 時間列を取得
                    if time >= 400.0:  # 時間が40以上の場合のみ値を取得
                        for index in column_indices:
                            column_values[index].append(float(columns[index]))

            # 各列の平均を計算
            averages = {
                index: (sum(column_values[index]) / len(column_values[index])) if column_values[index] else 0.0
                for index in column_indices
            }

            # データがなかった場合の処理
            if not any(column_values.values()):
                print(f"No valid hbond values found in {file_path} for averaging.")
                continue

            # "average" ディレクトリを作成
            average_dir_path = os.path.join(dir_path, "average")
            os.makedirs(average_dir_path, exist_ok=True)

            # 結果を保存
            average_file_path = os.path.join(average_dir_path, "result.txt")
            try:
                with open(average_file_path, 'w') as avg_file:
                    for index, avg in averages.items():
                        avg_file.write(f"Column {index+1} average: {avg:.6f}\n")
                    print(f"Averages saved to {average_file_path}")
            except Exception as e:
                print(f"Failed to write averages to {average_file_path}: {e}")

    except Exception as e:
        print(f"An unexpected error occurred: {e}")
```

```python
#平均値を計算する必要のない表面積を解析ディレクトリに入れる
import os
from collections import OrderedDict

def extract_total_from_log(log_file, target_commands):
    """ログファイルから指定されたターゲットコマンドの次の `total =` の値を抽出する"""
    target_areas = OrderedDict((cmd, None) for cmd in target_commands)  # 順番を保証する辞書

    try:
        with open(log_file, 'r') as file:
            lines = file.readlines()
            for i, line in enumerate(lines):
                for key in target_areas.keys():
                    if key in line:  # ターゲットコマンドを検出
                        for j in range(i + 1, len(lines)):
                            if "total  =" in lines[j]:
                                target_areas[key] = lines[j].split()[-1]  # 値を取得
                                break
        return target_areas
    except FileNotFoundError:
        print(f"Log file not found: {log_file}")
        return None
    except Exception as e:
        print(f"Error reading log file: {e}")
        return None

def save_results(value):
    peptide_positions = ["in", "sur", "out"]
    target_commands = [
        "COOR SURF RPROBE 1.4 SELE all END",  # surf_area 用
        "COOR SURF RPROBE 1.4 SELE type O* .or. type N* END",  # polar_surf_area 用
    ]

    for position in peptide_positions:
        # パスの定義
        log_file = f"/Users/teraimao/experiment/work/{value}_work/gbsw_{position}/log"
        dir_polar = f"/Users/teraimao/experiment/work/{value}_work/analysis/ana_gbsw_{position}/polar_surf_area/average"
        dir_surf = f"/Users/teraimao/experiment/work/{value}_work/analysis/ana_gbsw_{position}/surf_area/average"

        # ログファイルからデータを抽出
        target_areas = extract_total_from_log(log_file, target_commands)
        if not target_areas:
            print(f"Skipping {log_file} due to missing or invalid data.")
            continue

        # 値を保存
        if None not in target_areas.values():  # 全てのターゲットが見つかった場合
            os.makedirs(dir_polar, exist_ok=True)
            os.makedirs(dir_surf, exist_ok=True)

            # 正しい対応関係で保存する
            with open(os.path.join(dir_surf, "result.txt"), 'w') as file_surf:
                file_surf.write(target_areas["COOR SURF RPROBE 1.4 SELE all END"] + "\n")

            with open(os.path.join(dir_polar, "result.txt"), 'w') as file_polar:
                file_polar.write(target_areas["COOR SURF RPROBE 1.4 SELE type O* .or. type N* END"] + "\n")

            print(f"Values saved successfully for {position}:")
            print(f" - {target_areas['COOR SURF RPROBE 1.4 SELE all END']} saved to {os.path.join(dir_surf, 'result.txt')}")
            print(f" - {target_areas['COOR SURF RPROBE 1.4 SELE type O* .or. type N* END']} saved to {os.path.join(dir_polar, 'result.txt')}")
        else:
            print(f"Error: Missing data for some target commands in {log_file}.")
```
