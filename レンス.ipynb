```python
import re, unicodedata

subscript_map = str.maketrans('₀₁₂₃₄₅₆₇₈₉', '0123456789')

def normalize_literal(s: str) -> str:
    s = unicodedata.normalize('NFKC', s)
    return (s.replace('\u2212', '-')  # minus → hyphen-minus
             .replace('\u2010', '-').replace('\u2011', '-')
             .replace('\u2012', '-').replace('\u2013', '-')
             .replace('\u2014', '-').replace('\u2015', '-')
             .replace('\u200b', '')
             .replace('（','(').replace('）',')')
             .replace('，',',')
            )

def normalize_token(t: str) -> str:
    t = normalize_literal(t).translate(subscript_map)
    t = re.sub(r'\s+', '', t)      # 余分な空白除去
    t = t.replace('--','-').replace('−','-')
    return t

norm_target = normalize_token("Hph(4-CF3,3-Cl)")
norm_row = [normalize_token(x) for x in row_tokens]

print("== 正規化後比較:", norm_target in norm_row)
if norm_target in norm_row:
    print("一致OK:", norm_target)
else:
    print("非一致例:", [x for x in norm_row if x.startswith('Hph(')])
```

```python
import ast, re, unicodedata
import pandas as pd

# --- 正規化ユーティリティ（先にこれで整形） ---
subscript_map = str.maketrans('₀₁₂₃₄₅₆₇₈₉', '0123456789')
def normalize_literal(s: str) -> str:
    s = unicodedata.normalize('NFKC', str(s))
    return (s.replace('\u2212','-').replace('\u2010','-').replace('\u2011','-')
             .replace('\u2012','-').replace('\u2013','-').replace('\u2014','-').replace('\u2015','-')
             .replace('\u200b','').replace('（','(').replace('）',')').replace('，',','))

def normalize_token(t: str) -> str:
    t = normalize_literal(t).translate(subscript_map)
    t = re.sub(r'\s+', '', t)
    t = t.replace('--','-').replace('−','-')
    return t

def to_list(s):
    s = normalize_literal(s)
    try:
        return ast.literal_eval(s)
    except Exception:
        return None  # パースできない行は None

# --- パースして爆発 ---
seq_lists = df['Sequence'].map(to_list)
tokens_df = (
    df.assign(__tokens=seq_lists)
      .dropna(subset=['__tokens'])
      .explode('__tokens', ignore_index=False)  # 元の行indexを保持
      .rename(columns={'__tokens':'token'})
)

# 正規化
tokens_df['token_norm'] = tokens_df['token'].astype(str).map(normalize_token)

# ターゲットも正規化して比較
target = "Hph(4-CF3,3-Cl)"
target_norm = normalize_token(target)

hit_tokens = tokens_df[tokens_df['token_norm'] == target_norm]
print("ヒット（トークン行）:", len(hit_tokens))

# 親行（元の df の該当行）を取得
hit_rows = df.loc[hit_tokens.index.unique()]
print("ヒット（元行）:", len(hit_rows))
```

```python
import pandas as pd
import re

# N-numberとC-numberをHELM列から抽出する関数
def extract_nc_numbers(helm_series):
    n_list = []
    c_list = []
    pattern = re.compile(r',(\d+):R\d+[-\^](\d+):R\d+')

    for helm in helm_series:
        match = pattern.search(helm)
        if match:
            n_number = match.group(1)
            c_number = match.group(2)
        else:
            n_number = None
            c_number = None
        n_list.append(n_number)
        c_list.append(c_number)

    return pd.DataFrame({'N-number': n_list, 'C-number': c_list})

# CSVの読み込み
kouhann = '/Users/teraimao/experiment/CycPeptMPDB_All.csv'
rinrin = pd.read_csv(kouhann, dtype=str)

# N-numberとC-numberを抽出
nc_df = extract_nc_numbers(rinrin['HELM'])

# HELM列の右隣に列を挿入する
helm_index = rinrin.columns.get_loc('HELM')
for col in ['N-number', 'C-number']:
    rinrin.insert(helm_index + 1, col, nc_df[col])

# 保存
rinrin.to_csv("/Users/teraimao/experiment/data/all_csv_data/CycPeptMPDB_All_+NC.csv", index=False, encoding="utf-8")
```

```python
# 端capの種類と側
CAP_TOKENS = {'ac-': 'N', '-pip': 'C', '-pyrro': 'C'}

# 端の残基で cap の有無に応じて resname を切り替えるスイッチ（必要に応じて増やす）
EDGE_SWITCH = {
    # token: { 'default': 非cap, 'with_cap': cap隣接時 }
    'T':   {'default': 'THR', 'with_cap': 'T'},
    'D':   {'default': 'ASP', 'with_cap': 'D'},
    'meD': {'default': 'DSP', 'with_cap': 'D'},  # NME は後段のルールで付与
}

def adjust_indices_after_exclude(seq, N_idx, C_idx):
    """除外前 1始まり index を、除外トークン除去後の 1始まりに補正。併せて mask を返す。"""
    mask = [tok not in EXCLUDE_TOKENS for tok in seq]
    def to_filtered(i1):
        i0 = i1 - 1
        if not mask[i0]:
            raise ValueError(f"Ring index {i1} is an excluded token ({seq[i0]}).")
        return sum(mask[:i1])
    return to_filtered(N_idx), to_filtered(C_idx), mask  # ← mask も返す

def map_resnames(seq_filtered, N_adj, C_adj, rules, kind_2, *,
                 original_seq=None, original_mask=None):
    """
    除外後配列と N′/C′（1始まり）から最終 resname を決める。
    T/D/meD のような『端capの有無で切り替える』特例も反映。
    返り値: mapped_names, (N_rule, C_rule)
    """
    # 除外後 index -> 除外前 index（0始まり）の対応
    idx_map = []
    if original_seq is not None and original_mask is not None:
        idx_map = [i for i, keep in enumerate(original_mask) if keep]

    def has_N_cap_before(i_filtered_1based):
        """除外後の先頭(=1)の残基で、除外前直前が N-cap(ac-) か？"""
        if not idx_map or i_filtered_1based != 1:
            return False
        j = idx_map[0]  # 除外前 index
        return (j-1) >= 0 and CAP_TOKENS.get(original_seq[j-1]) == 'N'

    def has_C_cap_after(i_filtered_1based):
        """除外後の末尾の残基で、除外前直後が C-cap(-pip/-pyrro) か？"""
        if not idx_map or i_filtered_1based != len(idx_map):
            return False
        j = idx_map[-1]
        return (j+1) < len(original_seq) and CAP_TOKENS.get(original_seq[j+1]) == 'C'

    mapped = []
    N_rule = C_rule = None
    L = len(seq_filtered)

    for i, token in enumerate(seq_filtered, start=1):
        rule = find_rule(token, rules)
        is_ring_pos = (i == N_adj or i == C_adj)

        if is_ring_pos:
            # 端かどうか（端以外はcapの影響なし）
            cap_adjacent = (i == 1 and has_N_cap_before(i)) or (i == L and has_C_cap_after(i))

            # 端特例（T/D/meD 等）の切り替え
            if token in EDGE_SWITCH and cap_adjacent:
                mapped.append(EDGE_SWITCH[token]['with_cap'])
            elif token in EDGE_SWITCH and not cap_adjacent:
                mapped.append(EDGE_SWITCH[token]['default'])
            # 通常のリング時ルール
            elif rule and rule.get('ring_res'):
                mapped.append(rule['ring_res'])
            else:
                mapped.append(kind_2.get(token, token))

            if i == N_adj: N_rule = rule
            if i == C_adj: C_rule = rule

        else:
            # リング非参加位置
            if rule and rule.get('nonring_res'):
                mapped.append(rule['nonring_res'])
            else:
                mapped.append(kind_2.get(token, token))

    return mapped, (N_rule, C_rule)
```

```python
import pandas as pd
import ast
import math

# 入力ファイル
in_csv = "/Users/teraimao/experiment/data/all_csv_data/CycPeptMPDB_All_+NC_modified.csv"

def parse_sequence_cell(x):
    """Sequence列をlistに変換（リスト風文字列対応）"""
    if isinstance(x, list):
        return x
    if x is None or (isinstance(x, float) and math.isnan(x)):
        return []
    if isinstance(x, str):
        try:
            v = ast.literal_eval(x.strip())
            if isinstance(v, list):
                return v
        except Exception:
            pass
    return []

# CSV読み込み
df = pd.read_csv(in_csv, dtype={'Sequence': str}, low_memory=False)

# Sequenceをリスト化して -pip と ac- が両方含まれるか判定
mask = df['Sequence'].apply(lambda s: set(parse_sequence_cell(s)) >= {"-pip", "-pyrrp"})

# 抽出
df_both = df.loc[mask, ['ID', 'Sequence']]

# 件数と内容を表示
print(f"-pip と ac- が同時に含まれる行数: {len(df_both)}")
print(df_both.to_string(index=False))

# 必要ならCSVに出力
# df_both.to_csv("/Users/teraimao/experiment/data/all_csv_data/with_pip_and_ac.csv", index=False)
```

```python
import pandas as pd
import ast
import math

# 入出力
in_csv  = "/Users/teraimao/experiment/data/all_csv_data/CycPeptMPDB_All_+NC_+MODI+CYC.csv"
out_csv = "/Users/teraimao/experiment/data/all_csv_data/CycPeptMPDB_All_+NC_cleaned.csv"

def parse_sequence_cell(x):
    if isinstance(x, list):
        return x
    if x is None or (isinstance(x, float) and math.isnan(x)):
        return []
    if isinstance(x, str):
        try:
            v = ast.literal_eval(x.strip())
            if isinstance(v, list):
                return v
        except Exception:
            pass
    return []

# 相互排他的な修飾子（同時に2つは存在しない想定）
MODIFIERS = ["-pip", "ac-", "-pyrro"]

# 修飾子→パッチ名（必要に応じて変えてください）
PATCH_MAP = {
    "-pip":  "C_TERM_PIP",      # 例
    "ac-":   "N_TERM_ACETYL",   # 例
    "-pyrro":"C_TERM_PYRRO",    # 例
}

df = pd.read_csv(in_csv, dtype={'Sequence': str}, low_memory=False)

def process_row(seq_str):
    seq = parse_sequence_cell(seq_str)
    present = [m for m in MODIFIERS if m in seq]

    # 相互排他のはずだが、もし2つ以上あれば検知（記録だけにして続行）
    warning = ""
    if len(present) > 1:
        warning = f"Multiple modifiers found: {present}"

    modifier = present[0] if present else ""
    patch = PATCH_MAP.get(modifier, "") if modifier else ""

    # 修飾子を抜いたクリーン配列
    clean_seq = [tok for tok in seq if tok not in MODIFIERS]

    return pd.Series({
        "Modifier": modifier,              # '', '-pip', 'ac-', '-pyrro'
        "Patch_From_Modifier": patch,      # '', or パッチ名
        "Clean_Sequence": str(clean_seq),  # 残基のみ（PDB生成に使用）
        "Modifier_Warning": warning        # 複数検出時の警告（通常は空）
    })

added = df["Sequence"].apply(process_row)
df_out = pd.concat([df, added], axis=1)

# Sequence の直後に新列を移動
cols = list(df_out.columns)
if "Sequence" in cols:
    base = cols.index("Sequence")
    for name in ["Modifier", "Patch_From_Modifier", "Clean_Sequence", "Modifier_Warning"]:
        if name in cols:
            cols.insert(base + 1, cols.pop(cols.index(name)))
            base += 1
    df_out = df_out[cols]

df_out.to_csv(out_csv, index=False)
print(f"Saved: {out_csv}")
```

```python
import pandas as pd

# 入出力
in_csv  = "/Users/teraimao/experiment/data/all_csv_data/CycPeptMPDB_All_+NC_cleaned.csv"
out_csv = "/Users/teraimao/experiment/data/all_csv_data/CycPeptMPDB_All_+NC_with_newNC.csv"

df = pd.read_csv(in_csv, low_memory=False)

# 整数に変換（欠損なし前提）
df["N-number"] = df["N-number"].apply(lambda x: int(float(x)))
df["C-number"] = df["C-number"].apply(lambda x: int(float(x)))

# ac- のときだけ 1 減算、その他はそのまま
is_ac = (df["Modifier"] == "ac-")
df["newN_number"] = df["N-number"] - is_ac.astype(int)  # True→1, False→0 を利用
df["newC_number"] = df["C-number"] - is_ac.astype(int)

# 列を Clean_Sequence の右隣に移動
cols = list(df.columns)
if "Clean_Sequence" in cols:
    base = cols.index("Clean_Sequence")
    for name in ["newN_number", "newC_number"]:
        cols.insert(base + 1, cols.pop(cols.index(name)))
        base += 1
    df = df[cols]

# 保存（整数のまま出力）
df.to_csv(out_csv, index=False)
print(f"Saved: {out_csv}")
```

```python
import pandas as pd

# 入出力
in_csv  = "/Users/teraimao/experiment/data/all_csv_data/CycPeptMPDB_All_+NC_with_newNC.csv"
out_csv = "/Users/teraimao/experiment/data/all_csv_data/CycPeptMPDB_All_+NC_with_length.csv"

df = pd.read_csv(in_csv, low_memory=False)

# 差に1を足して新しい列を追加
df["Cycle_Length"] = (df["newC_number"] - df["newN_number"]) + 1

# Clean_Sequence の右隣に列を移動
cols = list(df.columns)
if "Clean_Sequence" in cols:
    base = cols.index("Clean_Sequence")
    cols.insert(base + 1, cols.pop(cols.index("Cycle_Length")))
    df = df[cols]

# 保存
df.to_csv(out_csv, index=False)
print(f"Saved: {out_csv}")
```
