```python
def now_str():
    return datetime.now().strftime("%Y-%m-%d %H:%M:%S")

def run_step(step_name, func, *args, **kwargs):
    try:
        func(*args, **kwargs)
        return True, f"{step_name}: OK"
    except NameError as e:
        return False, f"{step_name}: NameError (未定義関数/変数?) -> {e}"
    except subprocess.CalledProcessError as e:
        return False, f"{step_name}: CalledProcessError -> {e}"
    except Exception as e:
        return False, f"{step_name}: Exception -> {e}"

def ensure_dir(p):
    os.makedirs(p, exist_ok=True)
    return p

def append_results(rows, out_csv_path):
    ensure_dir(os.path.dirname(out_csv_path))
    df = pd.DataFrame(rows)
    header = not os.path.exists(out_csv_path)
    df.to_csv(out_csv_path, index=False, mode="a", header=header)

# --- 重要: 既存関数のシグネチャ微調整（※本体はあなたの実装を使う） ---
# make_rename_tcl に C_number を追加してください:
# def make_rename_tcl(input_list, value, C_number): ...

# --- 追加: パイプライン実行本体 ---
def run_pipeline_for_row(row, rules, token_kinds_set, out_csv_path, file_path):
    seq_id = row.ID
    started = now_str()

    try:
        input_list = ast.literal_eval(row.Clean_Sequence)
    except Exception as e:
        append_results([{
            "ID": seq_id, "Status": "X", "FailedStep": "parse_sequence",
            "Message": f"Clean_Sequenceのパース失敗: {e}",
            "StartedAt": started, "EndedAt": now_str(),
            "WorkDir": "", "Logs": ""
        }], out_csv_path)
        return

    if not all(tok in token_kinds_set for tok in seq_list):
        append_results([{
            "ID": seq_id, "Status": "X", "FailedStep": "token_check",
            "Message": f"未知トークンを含む: {set(seq_list) - token_kinds_set}",
            "StartedAt": started, "EndedAt": now_str(),
            "WorkDir": "", "Logs": ""
        }], out_csv_path)
        return

    N_number = row.newN_number
    C_number = row.newC_number
    Modifier = row.Modifier
    Patch    = row.Cyc_Patch

    today = datetime.today().strftime('%Y.%m.%d')
    value = f"{seq_id}_{today}"
    work_dir = f"/Users/teraimao/experiment/confirm/{value}_work"
    logs_dir = ensure_dir(os.path.join(work_dir, "logs"))

    steps = []

    ok, msg = run_step("copy_seed_work", copy_seed_work, value)
    steps.append(msg)
    if not ok: return append_results([{
        "ID": seq_id, "Status": "X", "FailedStep": "copy_seed_work",
        "Message": msg, "StartedAt": started, "EndedAt": now_str(),
        "WorkDir": work_dir, "Logs": "\n".join(steps)
    }], out_csv_path)

    ok, msg = run_step("copy_seed_pdb", copy_seed_pdb, value, input_list, C_number)
    steps.append(msg)
    if not ok: return append_results([{
        "ID": seq_id, "Status": "X", "FailedStep": "copy_seed_pdb",
        "Message": msg, "StartedAt": started, "EndedAt": now_str(),
        "WorkDir": work_dir, "Logs": "\n".join(steps)
    }], out_csv_path)

    ok, msg = run_step("make_rename_tcl", make_rename_tcl, input_list, value, C_number)
    steps.append(msg)
    if not ok: return append_results([{
        "ID": seq_id, "Status": "X", "FailedStep": "make_rename_tcl",
        "Message": msg, "StartedAt": started, "EndedAt": now_str(),
        "WorkDir": work_dir, "Logs": "\n".join(steps)
    }], out_csv_path)

    ok, msg = run_step("in_charminp", in_charminp, value)
    steps.append(msg)
    if not ok: return append_results([{
        "ID": seq_id, "Status": "X", "FailedStep": "in_charminp",
        "Message": msg, "StartedAt": started, "EndedAt": now_str(),
        "WorkDir": work_dir, "Logs": "\n".join(steps)
    }], out_csv_path)

    ok, msg = run_step("send_dir", send_dir, value)
    steps.append(msg)
    if not ok: return append_results([{
        "ID": seq_id, "Status": "X", "FailedStep": "send_dir",
        "Message": msg, "StartedAt": started, "EndedAt": now_str(),
        "WorkDir": work_dir, "Logs": "\n".join(steps)
    }], out_csv_path)

    ok, msg = run_step("return_copy_work", return_copy_work, value)
    steps.append(msg)
    if not ok:
        return append_results([{
            "ID": seq_id, "Status": "X", "FailedStep": "return_copy_work",
            "Message": msg, "StartedAt": started, "EndedAt": now_str(),
            "WorkDir": work_dir, "Logs": "\n".join(steps)
        }], out_csv_path)

    ok, msg = run_step("save_results", save_results, value)
    steps.append(msg)
    if not ok:
        return append_results([{
            "ID": seq_id, "Status": "X", "FailedStep": "save_results",
            "Message": msg, "StartedAt": started, "EndedAt": now_str(),
            "WorkDir": work_dir, "Logs": "\n".join(steps)
        }], out_csv_path)

    ok, msg = run_step("in_csv", in_csv, file_path)  # ★ file_path を引数で受け取るようにした
    steps.append(msg)
    if not ok:
        return append_results([{
            "ID": seq_id, "Status": "X", "FailedStep": "in_csv",
            "Message": msg, "StartedAt": started, "EndedAt": now_str(),
            "WorkDir": work_dir, "Logs": "\n".join(steps)
        }], out_csv_path)

    append_results([{
        "ID": seq_id, "Status": "O", "FailedStep": "",
        "Message": "All steps completed",
        "StartedAt": started, "EndedAt": now_str(),
        "WorkDir": work_dir, "Logs": "\n".join(steps)
    }], out_csv_path)
```

```python
def run_pipeline_for_row_hdgb(row, rules, token_kinds_set, out_csv_path, file_path, run_remote, 
                              t_end_ps, dt_ps, skip_ps, protocol):
    
    seq_id = row.ID
    started = now_str()
    steps = []

    print("DEBUG A: entered run_pipeline_for_row_hdgb")
    print("DEBUG B: run_remote =", run_remote)

    # --- 1. Clean_Sequence → list パース ---
    try:
        input_list = ast.literal_eval(row.Clean_Sequence)
    except Exception as e:
        append_results([{
            "ID": seq_id,
            "Status": "X",
            "FailedStep": "parse_sequence",
            "Message": f"パース失敗: {e}",
            "StartedAt": started,
            "EndedAt": now_str(),
            "WorkDir": "",
            "Logs": ""
        }], out_csv_path)
        return

    # --- 2. トークン検査 ---
    unknown_tokens = [tok for tok in input_list if tok not in token_kinds_set]
    if unknown_tokens:
        append_results([{
            "ID": seq_id,
            "Status": "X",
            "FailedStep": "token_check",
            "Message": f"未知トークンあり: {unknown_tokens}",
            "StartedAt": started,
            "EndedAt": now_str(),
            "WorkDir": "",
            "Logs": ""
        }], out_csv_path)
        print(f"   [SKIP] ID {seq_id}: 未知のアミノ酸が含まれているためスキップします。")
        return

    # --- 3. パラメータ準備 ---
    C_number = int(row.newC_number) if pd.notna(row.newC_number) else None
    N_number = int(row.newN_number) if pd.notna(row.newN_number) else None
    Modifier = ("" if pd.isna(row.Modifier) else str(row.Modifier)).strip()
    Patch    = ("" if pd.isna(row.Cyc_Patch) else str(row.Cyc_Patch)).strip()

    today = datetime.today().strftime('%Y.%m.%d')
    value = f"{seq_id}_{today}_{int(t_end_ps)}ps"
    work_dir = f"/Users/teraimao/experiment/confirm/{value}_work"
    ensure_dir(os.path.join(work_dir, "logs"))

    # --- 4. ローカル処理ステップ ---
    local_steps = [
        ("copy_seed_hdgb_work", copy_seed_hdgb_work, [value]),
        ("copy_seed_pdb", copy_seed_pdb, [value, input_list, C_number]),
        ("make_rename_tcl", make_rename_tcl, [input_list, value, C_number]),
        ("in_charminp", in_charminp, [value, input_list, C_number, N_number, Modifier, Patch, rules]),

        ("write_all_hdgb_inputs", write_all_hdgb_inputs,
         [value, input_list, C_number, N_number, Modifier, Patch, rules,
          t_end_ps, dt_ps, refz_map, seq_id, protocol]),

        ("write_all_hdgb_hbond_inputs", write_all_hdgb_hbond_inputs,
         [value, t_end_ps, dt_ps, skip_ps, protocol]),

        ("write_all_hdgb_sasa_inputs", write_all_hdgb_sasa_inputs,
         [value, t_end_ps, dt_ps, skip_ps, protocol]),

        ("write_all_hdgb_run_csh", write_all_hdgb_run_csh, [value]),
    ]

    for name, func, args in local_steps:
        ok, msg = run_step(name, func, *args)
        steps.append(msg)
        if not ok:
            append_results([{
                "ID": seq_id,
                "Status": "X",
                "FailedStep": name,
                "Message": msg,
                "StartedAt": started,
                "EndedAt": now_str(),
                "WorkDir": work_dir,
                "Logs": "\n".join(steps)
            }], out_csv_path)
            return

    print("DEBUG C: finished local_steps")
    print("DEBUG D: about to check run_remote")

    # --- 5. リモート処理ステップ ---
    if run_remote:
        print("DEBUG E: entered remote block")

        client = connect_server()
        if client is None:
            append_results([{
                "ID": seq_id,
                "Status": "X",
                "FailedStep": "SSH_Connect",
                "Message": "接続失敗",
                "StartedAt": started,
                "EndedAt": now_str(),
                "WorkDir": work_dir,
                "Logs": "\n".join(steps)
            }], out_csv_path)
            return

        try:
            print("DEBUG F: before send_dir_hdgb")
            ok, msg = run_step("send_dir_hdgb", send_dir_hdgb, value, client)
            print("DEBUG G: after send_dir_hdgb", ok, msg)
            steps.append(msg)

            if not ok:
                append_results([{
                    "ID": seq_id,
                    "Status": "X",
                    "FailedStep": "send_dir_hdgb",
                    "Message": msg,
                    "StartedAt": started,
                    "EndedAt": now_str(),
                    "WorkDir": work_dir,
                    "Logs": "\n".join(steps)
                }], out_csv_path)
                return

            print("DEBUG J: before run_sh_hdgb")
            ok, msg = run_step("run_sh_hdgb", run_sh_hdgb, value, client)
            print("DEBUG K: after run_sh_hdgb", ok, msg)
            steps.append(msg)

            if not ok:
                append_results([{
                    "ID": seq_id,
                    "Status": "X",
                    "FailedStep": "run_sh_hdgb",
                    "Message": msg,
                    "StartedAt": started,
                    "EndedAt": now_str(),
                    "WorkDir": work_dir,
                    "Logs": "\n".join(steps)
                }], out_csv_path)
                return

        finally:
            client.close()
    else:
        print("DEBUG H: remote skipped")
        steps.append("remote: SKIP")

    # --- 6. 最終検証と結果保存 ---
    if run_remote:
        status = "O"
        msg = "HDGB jobs submitted to server. Waiting for execution."
    else:
        ok_verify, verify_msg = verify_charmm_success(value)
        status = "O" if ok_verify else "X"
        msg = verify_msg if not ok_verify else "All HDGB steps completed"

    append_results([{
        "ID": seq_id,
        "Status": status,
        "FailedStep": "" if status == "O" else "verify_charmm_success",
        "Message": msg,
        "StartedAt": started,
        "EndedAt": now_str(),
        "WorkDir": work_dir,
        "Logs": "\n".join(steps)
    }], out_csv_path)
```

```python
def copy_seed_hdgb_work(value):
    src = "/Users/teraimao/experiment/confirm/seed_hdgbvdw_work"
    dst = f"/Users/teraimao/experiment/confirm/{value}_work"

    if not os.path.exists(src):
        return False, f"seed source not found: {src}"

    try:
        # 既存の作業ディレクトリを丸ごと消してからコピー
        if os.path.exists(dst):
            shutil.rmtree(dst)

        shutil.copytree(src, dst)

        # コピー先だけロック解除＆書き込みOK
        subprocess.run(["chflags", "-R", "nouchg", dst], check=False)
        subprocess.run(["chmod", "-R", "u+rwX", dst], check=False)

    except Exception as e:
        return False, f"copy_seed_hdgbvdw_work failed: {e}"

    return True, f"Seed copied cleanly into '{dst}'."
```

### make_rename_tcl build.tcl実行

```python
vmd_executable = "/Applications/VMD.app/Contents/vmd/vmd_MACOSXARM64"  # VMDのフルパス

def make_rename_tcl(input_list, value, C_number):
    # 1. 前の関数から「2つの値」を受け取る
    ok_rename, rename_build_tcl = rename_pdb_atom(input_list, C_number)
    
    # rename_pdb_atom が失敗していたらここで止める
    if not ok_rename:
        return False, f"rename_pdb_atom failed: {rename_build_tcl}"
    
    # 作業ディレクトリとスクリプトの出力先を定義
    working_dir = f"/Users/teraimao/experiment/confirm/{value}_work/rename"
    destination_path = os.path.join(working_dir, "build.tcl")
    
    try:
        # 作業ディレクトリが存在しなければ作成
        if not os.path.exists(working_dir):
            os.makedirs(working_dir)
        
        # build.tclファイルを作成
        with open(destination_path, 'w') as file:
            file.write(rename_build_tcl)
        
        # VMDの実行
        log_path = os.path.join(working_dir, "log")
        with open(log_path, "w") as log_file:
            subprocess.run(
                [vmd_executable, "-dispdev", "text", "-e", destination_path],
                stdout=log_file, 
                stderr=subprocess.PIPE, 
                check=True, 
                cwd=working_dir
            )
        
        msg = f"VMD executed successfully, output saved to '{log_path}'."
        print(msg)
        # ★成功を返す
        return True, msg

    except subprocess.CalledProcessError as e:
        err_msg = f"Error occurred during VMD execution: {e.stderr.decode() if e.stderr else e}"
        print(err_msg)
        return False, err_msg
    except Exception as e:
        err_msg = f"An unexpected error occurred in make_rename_tcl: {e}"
        print(err_msg)
        return False, err_msg
```

### 分子内水素結合のcharmm.inp

```python
def hbond_hdgb_charmm_inp(pos, t_end_ps, dt_ps, skip_ps,
                          protocol="constant",
                          segid="PROA", dist=3.0, angle=120.0,
                          out_name="output.dat"):
    pos = str(pos)
    steps_per_frame = int(round(dt_ps / 0.002))
    if steps_per_frame < 1:
        steps_per_frame = 1

    if protocol == "anneal":
        p3_start_ps = 0.75 * t_end_ps
        dcd_name = f"../../../hdgbvdw_{pos}/hdgbvdw_{pos}_p3.dcd"

        if skip_ps < p3_start_ps:
            actual_start_ps = p3_start_ps
        else:
            actual_start_ps = skip_ps

        begin_step = int(round(actual_start_ps / 0.002)) + steps_per_frame

    elif protocol == "constant":
        dcd_name = f"../../../hdgbvdw_{pos}/hdgbvdw_{pos}.dcd"
        actual_start_ps = max(skip_ps, 0.0)
        begin_step = int(round(actual_start_ps / 0.002)) + steps_per_frame

    else:
        raise ValueError(f"Unknown protocol: {protocol}")

    return f"""* HBOND Analysis for hdgbvdw_{pos}
*
open read card name ../../../toppar/4.1top_all36_prot.rtf unit 10
read rtf card unit 10
close unit 10
open read card name ../../../toppar/2_1gbsw_par_all36m_prot.prm unit 10
read parameter card unit 10
close unit 10
open read card name ../../../hdgbvdw_{pos}/peptide_{pos}.psf unit 10
read psf card unit 10
close unit 10

open read unform name {dcd_name} unit 20
traj query unit 20
set nframes = ?NFILE

open write card name {out_name} unit 30
set time {actual_start_ps}
set incr {dt_ps}

traj firstu 20 nunit 1 begin {begin_step} skip {steps_per_frame}

label loop
traj read
if ?EOF .eq. 1 goto end

coor hbond dist {dist:.3f} angle {angle:.1f} -
select segid {segid} .and. hydrogen end -
select segid {segid} .and. (type O* .or. type N*) end

set nhbond ?NHBOND
write title unit 30
* @time @nhbond
*
incr time by @incr
goto loop

label end
close unit 30
stop
"""
```

### コピーディレクトリをサーバーに

```python
def run_remote_command(client, command):
    stdin, stdout, stderr = client.exec_command(command)
    exit_status = stdout.channel.recv_exit_status()
    out = stdout.read().decode()
    err = stderr.read().decode()
    return exit_status, out, err


def send_dir_hdgb(value, client):
    print("DEBUG I: send_dir_hdgb called")

    sftp = client.open_sftp()
    local_path = f"/Users/teraimao/experiment/confirm/{value}_work"
    remote_path = f"/home/teraimao/experiment/confirm/{value}_work"

    try:
        # 1. サーバー側を作り直す
        status, out, err = run_remote_command(client, f"rm -rf {remote_path}")
        if status != 0:
            return False, f"Remote rm failed: {err}"

        status, out, err = run_remote_command(client, f"mkdir -p {remote_path}")
        if status != 0:
            return False, f"Remote mkdir failed: {err}"

        # 2. ローカル -> サーバーへアップロード
        upload_directory(sftp, local_path, remote_path)

        # 3. アップロード確認
        status, out, err = run_remote_command(client, f"ls -ld {remote_path} {remote_path}/charmm")
        if status != 0:
            return False, f"Upload check failed: {err}\nremote_path={remote_path}"

        print("[send_dir_hdgb] Upload check:")
        print(out)

        # 4. charmm/Run.csh を実行
        ok_csh, msg_csh = run_csh(value, client)
        if not ok_csh:
            return False, f"PSF failed: {msg_csh}"

        return True, "Uploaded files and finished run_csh."

    except Exception as e:
        return False, f"Error: {str(e)}"

    finally:
        sftp.close()
```

```python
def upload_directory(sftp, local_path, remote_path):
    """ディレクトリを再帰的にアップロードする"""
    try:
        sftp.mkdir(remote_path)
    except IOError:
        pass  # 既に存在する場合は無視

    for item in os.listdir(local_path):
        lp = os.path.join(local_path, item)
        rp = os.path.join(remote_path, item)
        if os.path.isdir(lp):
            upload_directory(sftp, lp, rp)
        else:
            sftp.put(lp, rp)
```

```python
def run_csh(value, client):
    """
    Run.csh を直接実行する。
    swork のディレクトリへ移動し、パスを通した上で実行。
    """
    remote_dir = f"/home/teraimao/experiment/confirm/{value}_work/charmm"
    
    # 実行コマンド: 
    # 1. cd で移動
    # 2. chmod で実行権限付与
    # 3. ./Run.csh で実行
    # ※ ログを log ファイルに書き出すようにしておくと、後で確認しやすいです
    command = f"cd {remote_dir} && chmod +x Run.csh && ./Run.csh > log 2>&1"
    
    print(f"  [EXEC] Running Run.csh in {remote_dir}...")
    
    stdin, stdout, stderr = client.exec_command(command)
    
    # 直接実行の場合、終わるまで待機して結果を受け取ります
    exit_status = stdout.channel.recv_exit_status() 
    
    if exit_status == 0:
        return True, "Run.csh completed successfully."
    else:
        err_msg = stderr.read().decode().strip()
        return False, f"Run.csh failed with status {exit_status}. Error: {err_msg}"
```

```python
def run_sh_hdgb(value, client):
    if client is None:
        return False, "Failed to connect to the server."

    peptide_positions = ["1", "2", "3", "4", "5"]
    results = []
    errors = []

    for pos in peptide_positions:
        remote_directory = f'/home/teraimao/experiment/confirm/{value}_work/hdgbvdw_{pos}'
        script_name = "Run.csh"

        submit_command = f'cd {remote_directory} && chmod +x {script_name} && sbatch -p f03.q {script_name}'

        stdin, stdout, stderr = client.exec_command(submit_command)
        out = stdout.read().decode().strip()
        err = stderr.read().decode().strip()

        if out:
            print(f"[hdgb_{pos}] {out}")
            results.append(f"{pos}: {out}")

        if err:
            print(f"[hdgb_{pos}] Error: {err}")
            errors.append(f"{pos}: {err}")

    if errors:
        return False, f"Some HDGB jobs failed: {errors}"

    if len(results) != 5:
        return False, f"Expected 5 submissions, but got {len(results)}"

    return True, f"Submitted HDGB jobs: {len(results)} positions"
```

```python

SAVE_DIR = "/Users/teraimao/experiment/data/解析結果/"
os.makedirs(SAVE_DIR, exist_ok=True)

REMOTE_SCRIPT_PATH = "/home/teraimao/experiment/scripts/aggregate_md_fast_universal.py"
REMOTE_OUT_DIR = "/home/teraimao/experiment/aggregation_results"


# =========================================================
# 固定モデル設定
# =========================================================
# ここだけ変える
# "gbsw", "hdgb", "hdgbvdw" のどれか
FIXED_MODEL = "gbsw"


def normalize_csv_name(name):
    """
    CSV名を整える。
    """
    name = str(name).strip()

    if name == "":
        name = f"master_database_{FIXED_MODEL}.csv"

    # パスが入っていた場合はファイル名だけ使う
    name = os.path.basename(name)

    if not name.endswith(".csv"):
        name += ".csv"

    return name


def get_custom_master_path(csv_name):
    csv_name = normalize_csv_name(csv_name)
    return os.path.join(SAVE_DIR, csv_name)


def submit_md_aggregation_background_custom_name(
    client,
    model,
    base_dir,
    start_id,
    end_id,
    target_date="ALL",
    target_t_end_ps="500",
    skip_ps=250.0,
    dt_ps=1.0,
    tolerance_frames=0,
    save_csv_name=None,
):
    """
    サーバー側で集計をバックグラウンド実行。
    ローカル保存用CSV名を指定できる版。
    """

    install_remote_universal_aggregator(client)

    model = model.lower()

    if model not in ["gbsw", "hdgb", "hdgbvdw"]:
        raise ValueError("model は 'gbsw', 'hdgb', 'hdgbvdw' のどれかにしてください")

    save_csv_name = normalize_csv_name(
        save_csv_name or f"master_database_{model}.csv"
    )

    base_name = save_csv_name.replace(".csv", "")

    tag_date = str(target_date).replace("*", "ALL").replace("/", "_")
    tag_tend = str(target_t_end_ps).replace("*", "ALL")
    timestamp = time.strftime("%Y%m%d_%H%M%S")

    # サーバー側の一時CSV名
    remote_csv = (
        f"{REMOTE_OUT_DIR}/{base_name}_"
        f"{model}_ID{start_id}_{end_id}_{tag_date}_{tag_tend}ps_{timestamp}.csv"
    )

    remote_log = remote_csv.replace(".csv", ".log")

    cmd = (
        f"nohup python3 {REMOTE_SCRIPT_PATH} "
        f"--model '{model}' "
        f"--base_dir '{base_dir}' "
        f"--start {start_id} "
        f"--end {end_id} "
        f"--date '{target_date}' "
        f"--tend '{target_t_end_ps}' "
        f"--skip {skip_ps} "
        f"--dt {dt_ps} "
        f"--tolerance_frames {tolerance_frames} "
        f"--out '{remote_csv}' "
        f"> '{remote_log}' 2>&1 & echo $!"
    )

    stdin, stdout, stderr = client.exec_command(cmd)
    pid = stdout.read().decode().strip()
    err = stderr.read().decode().strip()

    if err:
        print("stderr:", err)

    print("🚀 集計をバックグラウンド投入しました")
    print("model:", model)
    print("base_dir:", base_dir)
    print("PID:", pid)
    print("remote_csv:", remote_csv)
    print("remote_log:", remote_log)
    print("local_save_csv:", save_csv_name)

    return {
        "model": model,
        "pid": pid,
        "remote_csv": remote_csv,
        "remote_log": remote_log,
        "base_dir": base_dir,
        "save_csv_name": save_csv_name,
    }


def download_and_merge_md_aggregation_custom_name(client, job):
    """
    完了CSVをダウンロードし、指定したCSV名に統合する。
    """

    model = job["model"].lower()
    remote_csv = job["remote_csv"]
    save_csv_name = normalize_csv_name(job.get("save_csv_name", f"master_database_{model}.csv"))

    local_master_csv = get_custom_master_path(save_csv_name)

    local_tmp = os.path.join(
        SAVE_DIR,
        os.path.basename(remote_csv)
    )

    sftp = client.open_sftp()
    try:
        sftp.get(remote_csv, local_tmp)
    finally:
        sftp.close()

    print("✅ downloaded temporary csv:")
    print(local_tmp)

    new_df = pd.read_csv(local_tmp)

    if os.path.exists(local_master_csv):
        old_df = pd.read_csv(local_master_csv)

        final_df = pd.concat([old_df, new_df], ignore_index=True)

        final_df = final_df.drop_duplicates(
            subset=["Model", "ID", "Date", "Pos_Num", "t_end_ps"],
            keep="last"
        )
    else:
        final_df = new_df

    final_df = final_df.sort_values(
        ["ID", "Date", "t_end_ps", "Pos_Num"]
    ).reset_index(drop=True)

    final_df.to_csv(local_master_csv, index=False)

    print("✅ merged master saved:")
    print(local_master_csv)
    print("master shape:", final_df.shape)

    display(final_df.tail(10))

    return final_df
```

```python
import os
import time
import textwrap
import pandas as pd

SAVE_DIR = "/Users/teraimao/experiment/data/解析結果/"
MASTER_CSV_PATH = os.path.join(SAVE_DIR, "master_database_hdgbvdw.csv")
os.makedirs(SAVE_DIR, exist_ok=True)

REMOTE_SCRIPT_PATH = "/home/teraimao/experiment/scripts/aggregate_hdgbvdw_fast.py"
REMOTE_OUT_DIR = "/home/teraimao/experiment/aggregation_results"


REMOTE_SCRIPT_TEXT = r'''#!/usr/bin/env python3
import os
import re
import csv
import glob
import argparse
import math

BASE_DIR = "/home/teraimao/experiment/confirm"

valid_pattern = re.compile(r"^(\d+)_(\d{4}\.\d{2}\.\d{2})_(\d+)ps_work$")


def is_float(x):
    try:
        float(x)
        return True
    except Exception:
        return False


def to_float_or_nan(x):
    try:
        return float(x)
    except Exception:
        return math.nan


def parse_charmm_energies(log_path):
    """
    CHARMM logから LAVE EXTERN / LAVE PBEQ を読む。
    vdw  = LAVE EXTERN> の VDWaals
    coul = LAVE EXTERN> の ELEC
    solv = LAVE PBEQ> の GBEnr
    """
    vdw, coul, gbenr = math.nan, math.nan, math.nan

    if not os.path.exists(log_path):
        return vdw, coul, gbenr

    try:
        with open(log_path, "r", errors="ignore") as f:
            for line in f:
                if "LAVE EXTERN>" in line:
                    parts = line.split()
                    if len(parts) >= 4:
                        vdw = to_float_or_nan(parts[2])
                        coul = to_float_or_nan(parts[3])

                elif "LAVE PBEQ>" in line:
                    parts = line.split()
                    if len(parts) >= 5:
                        gbenr = to_float_or_nan(parts[4])
    except Exception:
        pass

    return vdw, coul, gbenr


def parse_refz(charmm_inp_path):
    """
    charmm.inpから refz の値を読む。
    """
    if not os.path.exists(charmm_inp_path):
        return ""

    try:
        with open(charmm_inp_path, "r", errors="ignore") as f:
            for line in f:
                parts = line.replace(",", " ").split()
                for i, p in enumerate(parts):
                    if p.lower() == "refz" and i + 1 < len(parts):
                        return parts[i + 1]
    except Exception:
        pass

    return ""


def mean_last_numeric_column(path):
    """
    hbond output.dat / sasa dat をなるべく頑健に読む。
    各行の「最後の数値」を平均する。

    hbondが
      * 250.0 3
    でも
      250.0 3
    でも、最後の 3 を拾う。
    """
    if not os.path.exists(path):
        return math.nan, 0

    vals = []

    try:
        with open(path, "r", errors="ignore") as f:
            for line in f:
                line = line.strip()
                if not line:
                    continue

                tokens = line.replace("*", " ").split()
                nums = []
                for t in tokens:
                    if is_float(t):
                        nums.append(float(t))

                if len(nums) == 0:
                    continue

                vals.append(nums[-1])

    except Exception:
        return math.nan, 0

    if len(vals) == 0:
        return math.nan, 0

    return sum(vals) / len(vals), len(vals)


def main():
    parser = argparse.ArgumentParser()

    parser.add_argument("--start", type=int, required=True)
    parser.add_argument("--end", type=int, required=True)
    parser.add_argument("--date", type=str, default="ALL")
    parser.add_argument("--tend", type=str, default="500")
    parser.add_argument("--skip", type=float, default=250.0)
    parser.add_argument("--dt", type=float, default=1.0)
    parser.add_argument("--out", type=str, required=True)
    parser.add_argument("--tolerance_frames", type=int, default=0)

    args = parser.parse_args()

    os.makedirs(os.path.dirname(args.out), exist_ok=True)

    if args.date.upper() in ["", "ALL", "すべて"]:
        search_date = "*"
    else:
        search_date = args.date

    if args.tend.upper() in ["", "ALL", "すべて"]:
        t_end_filter = None
    else:
        t_end_filter = int(float(args.tend))

    rows = []

    for seq_id in range(args.start, args.end + 1):
        pattern = os.path.join(BASE_DIR, f"{seq_id}_{search_date}_*ps_work")
        folder_paths = sorted(glob.glob(pattern))

        if not folder_paths:
            print(f"[{seq_id}] no folder")
            continue

        for path in folder_paths:
            folder_name = os.path.basename(path)
            m = valid_pattern.match(folder_name)

            if not m:
                continue

            extracted_id = int(m.group(1))
            extracted_date = m.group(2)
            t_end_ps = float(m.group(3))

            if t_end_filter is not None and int(t_end_ps) != t_end_filter:
                continue

            expected = max(0, int((t_end_ps - args.skip) / args.dt))
            min_required_hbond_n = max(0, expected - args.tolerance_frames)

            for pos in range(1, 6):
                pos_dir = os.path.join(path, f"hdgbvdw_{pos}")
                analysis_dir = os.path.join(path, "analysis", f"ana_hdgbvdw_{pos}")

                charmm_inp = os.path.join(pos_dir, "charmm.inp")
                log_path = os.path.join(pos_dir, "log")
                hb_path = os.path.join(analysis_dir, "hbond", "output.dat")
                sasa_path = os.path.join(analysis_dir, "surf_area", f"sasa_{pos}.dat")

                z_coord = parse_refz(charmm_inp)
                vdw, coul, solv = parse_charmm_energies(log_path)
                hb_avg, hb_n = mean_last_numeric_column(hb_path)
                sasa_avg, sasa_n = mean_last_numeric_column(sasa_path)

                energy_ok = (
                    not math.isnan(vdw) and
                    not math.isnan(coul) and
                    not math.isnan(solv)
                )

                hbond_ok = hb_n >= min_required_hbond_n

                status = "Success" if (energy_ok and hbond_ok) else "Failed"

                rows.append({
                    "ID": extracted_id,
                    "Date": extracted_date,
                    "Pos_Num": pos,
                    "Z_Coord": z_coord,
                    "t_end_ps": t_end_ps,
                    "Status": status,
                    "HB_Avg": hb_avg,
                    "SASA_Avg": sasa_avg,
                    "Solv_Free_Avg": solv,
                    "vdW_Avg": vdw,
                    "Coul_Avg": coul,
                    "HB_N": hb_n,
                    "SASA_N": sasa_n,
                    "Expected_N": expected,
                    "Source_Folder": path,
                })

        print(f"[{seq_id}] done")

    fieldnames = [
        "ID", "Date", "Pos_Num", "Z_Coord", "t_end_ps",
        "Status", "HB_Avg", "SASA_Avg", "Solv_Free_Avg",
        "vdW_Avg", "Coul_Avg",
        "HB_N", "SASA_N", "Expected_N", "Source_Folder",
    ]

    with open(args.out, "w", newline="") as f:
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(rows)

    print("saved:", args.out)
    print("rows:", len(rows))


if __name__ == "__main__":
    main()
'''


def run_remote_command(client, command):
    stdin, stdout, stderr = client.exec_command(command)
    status = stdout.channel.recv_exit_status()
    out = stdout.read().decode()
    err = stderr.read().decode()
    return status, out, err


def install_remote_hdgbvdw_aggregator(client):
    """
    サーバーに高速集計スクリプトを置く。
    """
    run_remote_command(client, "mkdir -p /home/teraimao/experiment/scripts")
    run_remote_command(client, f"mkdir -p {REMOTE_OUT_DIR}")

    sftp = client.open_sftp()
    try:
        with sftp.open(REMOTE_SCRIPT_PATH, "w") as f:
            f.write(REMOTE_SCRIPT_TEXT)
    finally:
        sftp.close()

    status, out, err = run_remote_command(client, f"chmod +x {REMOTE_SCRIPT_PATH}")

    if status != 0:
        raise RuntimeError(err)

    print("✅ remote script installed:")
    print(REMOTE_SCRIPT_PATH)


def submit_hdgbvdw_aggregation_background(
    client,
    start_id,
    end_id,
    target_date="ALL",
    target_t_end_ps="500",
    skip_ps=250.0,
    dt_ps=1.0,
    tolerance_frames=0,
):
    """
    集計をサーバー上でバックグラウンド実行する。
    Notebookはすぐ返ってくる。
    """
    install_remote_hdgbvdw_aggregator(client)

    tag_date = str(target_date).replace("*", "ALL").replace("/", "_")
    tag_tend = str(target_t_end_ps).replace("*", "ALL")

    timestamp = time.strftime("%Y%m%d_%H%M%S")

    remote_csv = (
        f"{REMOTE_OUT_DIR}/hdgbvdw_aggregation_"
        f"ID{start_id}_{end_id}_{tag_date}_{tag_tend}ps_{timestamp}.csv"
    )
    remote_log = remote_csv.replace(".csv", ".log")

    cmd = (
        f"nohup python3 {REMOTE_SCRIPT_PATH} "
        f"--start {start_id} "
        f"--end {end_id} "
        f"--date '{target_date}' "
        f"--tend '{target_t_end_ps}' "
        f"--skip {skip_ps} "
        f"--dt {dt_ps} "
        f"--tolerance_frames {tolerance_frames} "
        f"--out '{remote_csv}' "
        f"> '{remote_log}' 2>&1 & echo $!"
    )

    stdin, stdout, stderr = client.exec_command(cmd)
    pid = stdout.read().decode().strip()
    err = stderr.read().decode().strip()

    if err:
        print("stderr:", err)

    print("🚀 集計をバックグラウンド投入しました")
    print("PID:", pid)
    print("remote_csv:", remote_csv)
    print("remote_log:", remote_log)

    return {
        "pid": pid,
        "remote_csv": remote_csv,
        "remote_log": remote_log,
    }
```

```python
# --- 設定：保存先のディレクトリとファイル名 ---
SAVE_DIR = "/Users/teraimao/experiment/data/解析結果/"
MASTER_CSV_PATH = os.path.join(SAVE_DIR, "master_database_hdgbvdw.csv")

if not os.path.exists(SAVE_DIR):
    os.makedirs(SAVE_DIR)

def parse_charmm_energies(log_lines):
    vdw, coul, gbenr = np.nan, np.nan, np.nan
    for line in log_lines:
        if "LAVE EXTERN>" in line:
            parts = line.split()
            vdw, coul = float(parts[2]), float(parts[3])
        elif "LAVE PBEQ>" in line:
            parts = line.split()
            gbenr = float(parts[4])
    return vdw, coul, gbenr

def run_aggregation_progressive_hdgb(client, start_id, end_id, target_date, target_t_end_ps, skip_ps, dt_ps):
    results = []
    search_date = "*" if str(target_date).upper() in ["", "ALL", "すべて"] else target_date

    if str(target_t_end_ps).upper() in ["", "ALL", "すべて"]:
        t_end_filter = None
    else:
        t_end_filter = int(float(target_t_end_ps))

    valid_pattern = re.compile(r"^(\d+)_(\d{4}\.\d{2}\.\d{2})_(\d+)ps_work$")

    for seq_id in range(start_id, end_id + 1):
        print(f"🔎 [{seq_id}/{end_id}] フォルダを検索中...", end="\r")

        cmd = f"ls -d /home/teraimao/experiment/confirm/{seq_id}_{search_date}_*ps_work 2>/dev/null"
        _, stdout, _ = client.exec_command(cmd)
        folder_paths = stdout.read().decode().splitlines()

        if not folder_paths:
            continue

        for path in folder_paths:
            folder_name = os.path.basename(path)
            match = valid_pattern.match(folder_name)
            if not match:
                continue

            extracted_id = int(match.group(1))
            extracted_date = match.group(2)
            t_end_ps = float(match.group(3))

            if t_end_filter is not None and int(t_end_ps) != t_end_filter:
                continue

            # skip_ps 以降を dt_ps 間隔で解析している前提
            expected = max(0, int((t_end_ps - skip_ps) / dt_ps))

            for pos in range(1, 6):
                # A. Z座標取得
                z_cmd = f"grep 'refz' {path}/hdgbvdw_{pos}/charmm.inp | awk '{{for(i=1;i<=NF;i++) if($i==\"refz\") print $(i+1)}}'"
                _, out, _ = client.exec_command(z_cmd)
                z_coord = out.read().decode().strip()

                # B. エネルギー平均取得
                _, out, _ = client.exec_command(f"grep -A 10 'LAVE>' {path}/hdgbvdw_{pos}/log")
                vdw, coul, solv = parse_charmm_energies(out.read().decode().splitlines())

                # C. HBOND取得
                hb_path = f"{path}/analysis/ana_hdgbvdw_{pos}/hbond/output.dat"
                _, out, _ = client.exec_command(
                    f"awk '{{sum+=$2; n++}} END {{if(n>0) print sum/n, n; else print \"0 0\"}}' {hb_path}"
                )
                hb_res = out.read().decode().split()
                hb_avg, hb_n = (float(hb_res[0]), int(hb_res[1])) if len(hb_res) >= 2 else (np.nan, 0)

                # D. SASA取得（HDGBvdWでは必須にしない）
                sasa_path = f"{path}/analysis/ana_hdgbvdw_{pos}/surf_area/sasa_{pos}.dat"
                _, out, _ = client.exec_command(
                    f"awk '{{sum+=$2; n++}} END {{if(n>0) print sum/n, n; else print \"nan 0\"}}' {sasa_path} 2>/dev/null"
                )
                sasa_res = out.read().decode().split()
                try:
                    sasa_avg, sasa_n = (float(sasa_res[0]), int(sasa_res[1])) if len(sasa_res) >= 2 else (np.nan, 0)
                except Exception:
                    sasa_avg, sasa_n = np.nan, 0

                # E. Status判定
                # HDGBvdWではSASAは必須にしない
                energy_ok = (
                    not np.isnan(vdw) and
                    not np.isnan(coul) and
                    not np.isnan(solv)
                )

                hbond_ok = hb_n >= expected

                status = "Success" if (hbond_ok and energy_ok) else "Failed"

                results.append({
                    "ID": extracted_id,
                    "Date": extracted_date,
                    "Pos_Num": pos,
                    "Z_Coord": z_coord,
                    "t_end_ps": t_end_ps,
                    "Status": status,
                    "HB_Avg": hb_avg,
                    "SASA_Avg": sasa_avg,
                    "Solv_Free_Avg": solv,
                    "vdW_Avg": vdw,
                    "Coul_Avg": coul
                })

        print(f"✅ ID {seq_id} 完了 ({len(folder_paths)}フォルダ分)")

    if not results:
        return 0

    new_df = pd.DataFrame(results)
    if os.path.exists(MASTER_CSV_PATH):
        old_df = pd.read_csv(MASTER_CSV_PATH)
        final_df = pd.concat([old_df, new_df]).drop_duplicates(
            subset=['ID', 'Date', 'Pos_Num', 't_end_ps'],
            keep='last'
        )
    else:
        final_df = new_df

    final_df.sort_values(['ID', 'Date', 't_end_ps', 'Pos_Num']).to_csv(MASTER_CSV_PATH, index=False)
    return len(results)


# --- 操作パネル ---
style = {'description_width': 'initial'}
txt_start = widgets.IntText(value=1, description='開始ID:', style=style)
txt_end = widgets.IntText(value=10, description='終了ID:', style=style)
txt_date = widgets.Text(value='ALL', description='日付 (ALLで全日程):', style=style)
txt_tend = widgets.Text(value='500', description='計算時間ps (ALLで全件):', style=style)
txt_skip = widgets.FloatText(value=250.0, description='skip_ps:', style=style)
txt_dt = widgets.FloatText(value=1.0, description='dt_ps:', style=style)

btn_exec = widgets.Button(description='🚀 HDGB集計開始', button_style='danger', layout={'width': '300px'})
out_log = widgets.Output()

def on_click_exec_hdgb(b):
    with out_log:
        clear_output()
        print(f"🚀 HDGB集計を開始します... ID {txt_start.value} ~ {txt_end.value}")
        print(f"📁 保存先: {MASTER_CSV_PATH}")
        print("-" * 50)
        try:
            count = run_aggregation_progressive_hdgb(
                client,
                txt_start.value,
                txt_end.value,
                txt_date.value,
                txt_tend.value,
                txt_skip.value,
                txt_dt.value
            )
            if count > 0:
                print("-" * 50)
                print(f"🎉 すべて完了！ 合計 {count}件のデータを追加しました。")
                df_check = pd.read_csv(MASTER_CSV_PATH)
                display(df_check.tail(5))
            else:
                print("⚠️ 条件に合うデータはありませんでした。")
        except Exception as e:
            print(f"❌ エラーが発生しました: {e}")

btn_exec.on_click(on_click_exec_hdgb)

display(widgets.VBox([
    widgets.HTML("<h3>🐚 HDGB 解析データ集計</h3>"),
    widgets.HBox([txt_start, txt_end, txt_date, txt_tend, txt_skip, txt_dt]),
    btn_exec,
    out_log
]))
```

```python

def process_single_id(client, seq_id, today, t_end_ps):
    """
    指定された1つのIDフォルダに対して、サーバー上で集計・検品を行う。
    """
    folder_name = f"{seq_id}_{today}_{int(t_end_ps)}ps"
    base_path = f"/home/teraimao/experiment/confirm/{folder_name}_work"
    
    # 期待される行数の計算 (生産相25%分)
    # 500ps -> 125枚, 100ps -> 25枚 (dt=1.0ps想定)
    expected_lines = int((t_end_ps * 0.25) / 1.0)
    
    results_for_this_id = []

    for pos in range(1, 6):
        # --- 1. Z座標の取得 (charmm.inpからrefzを抜き出す) ---
        inp_path = f"{base_path}/gbsw_{pos}/charmm.inp"
        # refz の後の数値を grep と sed で抽出
        z_cmd = f"grep 'refz' {inp_path} | awk '{{print $8}}'"
        stdin, stdout, stderr = client.exec_command(z_cmd)
        z_coord = stdout.read().decode().strip()

        # --- 2. エネルギーデータの抽出 (awkコマンド) ---
        # ログから指定時間範囲の Solv(PBEQ $5), vdW(EXTERN $2), Coul(EXTERN $3) を抽出
        start_time = t_end_ps * 0.75
        log_path = f"{base_path}/gbsw_{pos}/log"
        
        # 例として Solv(GBEnr) を抽出して平均を出す
        extract_cmd = f"""
        awk '/^DYNA>/ {{time = $3; next}} 
             /^DYNA PBEQ>/ {{if (time >= {start_time}) print $5}}' {log_path}
        """
        stdin, stdout, stderr = client.exec_command(extract_cmd)
        solv_data = stdout.read().decode().splitlines()
        solv_avg = np.mean([float(x) for x in solv_data]) if solv_data else np.nan
        
        # --- 3. 解析データの行数チェック (HBONDを代表として使用) ---
        hb_dat = f"{base_path}/analysis/ana_gbsw_{pos}/hbond/output.dat"
        stdin, stdout, stderr = client.exec_command(f"wc -l < {hb_dat}")
        actual_lines = int(stdout.read().decode().strip() or 0)
        
        # 検品ステータス
        status = "Success" if actual_lines >= expected_lines else "Failed"
        
        # (同様に SASA, vdW, Coul の平均計算をここに追加...)
        
        results_for_this_id.append({
            "ID": seq_id,
            "Date": today,
            "Pos_Num": pos,
            "Z_Coord": z_coord,
            "t_end_ps": t_end_ps,
            "Status": status,
            "Solv_Free_Avg": solv_avg,
            # 他の項目も追加
        })
        
    return results_for_this_id
```

```python
def aggregate_results(client, start_id, end_id, today, master_csv_path="master_database.csv"):
    """
    指定した範囲のIDを集計し、master_database.csvに追記する
    """
    all_rows = []
    
    for seq_id in range(start_id, end_id + 1):
        # 1. ディレクトリの存在確認とt_end_psの自動判別
        # サーバー上でlsコマンドを使い、その日のIDフォルダを探す
        find_cmd = f"ls -d /home/teraimao/experiment/confirm/{seq_id}_{today}_*ps_work"
        stdin, stdout, stderr = client.exec_command(find_cmd)
        folder_paths = stdout.read().decode().splitlines()
        
        if not folder_paths:
            print(f"Skipping ID {seq_id}: Directory not found.")
            continue
            
        for base_path in folder_paths:
            folder_name = os.path.basename(base_path)
            # フォルダ名からt_end_psを抽出 (例: 500ps)
            t_end_ps = float(re.search(r"(\d+)ps", folder_name).group(1))
            expected_lines = int((t_end_ps * 0.25) / 1.0) # 25%分のデータ行数

            for pos in range(1, 6):
                # --- ステップ1: Z座標取得 ---
                z_cmd = f"grep 'refz' {base_path}/gbsw_{pos}/charmm.inp | awk '{{print $8}}'"
                _, stdout, _ = client.exec_command(z_cmd)
                z_coord = stdout.read().decode().strip()

                # --- ステップ2: エネルギー値取得 (案A: LAVEセクション) ---
                # LAVEセクションから VDWaals, ELEC, GBEnr を一気に取得
                ene_cmd = f"grep -A 8 'LAVE>' {base_path}/gbsw_{pos}/log | tail -n 8"
                _, stdout, _ = client.exec_command(ene_cmd)
                log_lines = stdout.read().decode().splitlines()
                
                # ここで各数値を解析 (行位置などは実際のlogの並び順に合わせる)
                vdw, elec, gbenr = parse_charmm_log_energies(log_lines)

                # --- ステップ3: 解析データの平均 (hbond, sasa) ---
                hb_dat = f"{base_path}/analysis/ana_gbsw_{pos}/hbond/output.dat"
                hb_avg, hb_count = get_dat_average(client, hb_dat)
                
                sasa_dat = f"{base_path}/analysis/ana_gbsw_{pos}/surf_area/sasa_1.dat"
                sasa_avg, sasa_count = get_dat_average(client, sasa_dat)

                # --- ステップ4: Status判定 ---
                # hbondとsasaの両方が予定の行数(125行など)あればSuccess
                status = "Success" if (hb_count >= expected_lines and sasa_count >= expected_lines) else "Failed"

                # リストに追加
                all_rows.append({
                    "ID": seq_id, "Date": today, "Pos_Num": pos, "Z_Coord": z_coord,
                    "t_end_ps": t_end_ps, "Status": status, "HB_Avg": hb_avg,
                    "SASA_Avg": sasa_avg, "Solv_Free_Avg": gbenr, 
                    "vdW_Avg": vdw, "Coul_Avg": elec
                })

    # --- ステップ5: CSVへの統合・追記 ---
    new_df = pd.DataFrame(all_rows)
    if os.path.exists(master_csv_path):
        old_df = pd.read_csv(master_csv_path)
        combined_df = pd.concat([old_df, new_df]).drop_duplicates(subset=['ID', 'Date', 'Pos_Num'], keep='last')
        combined_df.to_csv(master_csv_path, index=False)
    else:
        new_df.to_csv(master_csv_path, index=False)
    
    print(f"Aggregation complete. Master CSV updated.")
```

```python
def parse_charmm_energies(log_lines):
    """LAVEセクションから特定のエネルギー値を抽出する"""
    vdw, coul, gbenr = np.nan, np.nan, np.nan
    for line in log_lines:
        if line.startswith("LAVE EXTERN>"):
            parts = line.split()
            vdw = float(parts[2])  # VDWaals
            coul = float(parts[3]) # ELEC (Coulomb)
        elif line.startswith("LAVE PBEQ>"):
            parts = line.split()
            gbenr = float(parts[4]) # GBEnr (Solv Free)
    return vdw, coul, gbenr

def aggregate_to_master(client, start_id, end_id, today, master_csv="master_database.csv"):
    results = []
    
    for seq_id in range(start_id, end_id + 1):
        # 1. サーバー上のディレクトリを特定 (100ps/500ps自動判別)
        cmd = f"ls -d /home/teraimao/experiment/confirm/{seq_id}_{today}_*ps_work"
        _, stdout, _ = client.exec_command(cmd)
        folder_paths = stdout.read().decode().splitlines()
        
        if not folder_paths:
            print(f"ID {seq_id}: フォルダが見つかりません。スキップします。")
            continue
            
        for path in folder_paths:
            folder_name = os.path.basename(path)
            t_end_ps = float(re.search(r"(\d+)ps", folder_name).group(1))
            expected_lines = int((t_end_ps * 0.25) / 1.0) # 生産相25%分

            for pos in range(1, 6):
                # --- A. Z座標取得 ---
                z_cmd = f"grep 'refz' {path}/gbsw_{pos}/charmm.inp | awk '{{print $8}}'"
                _, stdout, _ = client.exec_command(z_cmd)
                z_coord = stdout.read().decode().strip()

                # --- B. エネルギー平均取得 (LAVE) ---
                # LAVE> から始まる10行程度を抜き出して解析
                ene_cmd = f"grep -A 10 'LAVE>' {path}/gbsw_{pos}/log"
                _, stdout, _ = client.exec_command(ene_cmd)
                vdw, coul, solv = parse_charmm_energies(stdout.read().decode().splitlines())

                # --- C. 解析データ平均 (HBOND, SASA) と行数チェック ---
                # HBOND
                hb_path = f"{path}/analysis/ana_gbsw_{pos}/hbond/output.dat"
                _, stdout, _ = client.exec_command(f"awk '{{sum+=$2; n++}} END {{if(n>0) print sum/n, n; else print \"0 0\"}}' {hb_path}")
                hb_res = stdout.read().decode().split()
                hb_avg, hb_count = float(hb_res[0]), int(hb_res[1])

                # SASA
                sasa_path = f"{path}/analysis/ana_gbsw_{pos}/surf_area/sasa_1.dat"
                _, stdout, _ = client.exec_command(f"awk '{{sum+=$2; n++}} END {{if(n>0) print sum/n, n; else print \"0 0\"}}' {sasa_path}")
                sasa_res = stdout.read().decode().split()
                sasa_avg, sasa_count = float(sasa_res[0]), int(sasa_res[1])

                # --- D. Status 判定 ---
                status = "Success" if (hb_count >= expected_lines and sasa_count >= expected_lines) else "Failed"

                results.append({
                    "ID": seq_id, "Date": today, "Pos_Num": pos, "Z_Coord": z_coord,
                    "t_end_ps": t_end_ps, "Status": status, "HB_Avg": hb_avg,
                    "SASA_Avg": sasa_avg, "Solv_Free_Avg": solv, "vdW_Avg": vdw, "Coul_Avg": coul
                })

    # CSVへの書き出し (追記モード)
    new_df = pd.DataFrame(results)
    if os.path.exists(master_csv):
        old_df = pd.read_csv(master_csv)
        final_df = pd.concat([old_df, new_df]).drop_duplicates(subset=['ID', 'Date', 'Pos_Num'], keep='last')
    else:
        final_df = new_df
    
    final_df.to_csv(master_csv, index=False)
    print(f"集計完了: {len(new_df)} 行のデータを {master_csv} に反映しました。")
```

```python
PATH_SMILES = "/Users/teraimao/experiment/data/all_csv_data/this_is_CycPeptMPDB.csv"
PATH_MASTER_HDGBVDW = "/Users/teraimao/experiment/data/解析結果/master_database_hdgbvdw.csv"
PATH_OUTPUT_HDGBVDW = "/Users/teraimao/experiment/data/解析結果/training_dataset_hdgbvdw_500ps_ID1_100.csv"

def build_integrated_dataset_hdgbvdw(start_id=1, end_id=100):
    if not os.path.exists(PATH_MASTER_HDGBVDW) or not os.path.exists(PATH_SMILES):
        raise FileNotFoundError("入力ファイルが見つかりません。")

    df_md = pd.read_csv(PATH_MASTER_HDGBVDW)
    df_sl = pd.read_csv(PATH_SMILES)

    df_md = df_md[(df_md["ID"] >= start_id) & (df_md["ID"] <= end_id)].copy()

    initial_count = df_md["ID"].nunique()

    df_md = df_md[df_md["Status"] == "Success"].copy()

    # HDGBvdWではSASAを必須にしない
    required_cols = ["HB_Avg", "Solv_Free_Avg", "vdW_Avg", "Coul_Avg"]
    df_md = df_md.dropna(subset=required_cols)

    # 5点そろっているIDだけ残す
    pos_counts = df_md.groupby("ID")["Pos_Num"].nunique()
    valid_ids = pos_counts[pos_counts == 5].index
    df_md = df_md[df_md["ID"].isin(valid_ids)].copy()

    final_count = df_md["ID"].nunique()
    removed_count = initial_count - final_count

    z_map = df_md.groupby("Pos_Num")["Z_Coord"].mean().to_dict()

    target_values = ["HB_Avg", "Solv_Free_Avg", "vdW_Avg", "Coul_Avg"]
    df_pivot = df_md.pivot(index="ID", columns="Pos_Num", values=target_values)

    new_columns = []
    for val, pos in df_pivot.columns:
        z_val = int(float(z_map.get(pos, pos)))
        new_columns.append(f"{val}_z{z_val}")

    df_pivot.columns = new_columns
    df_pivot = df_pivot.reset_index()

    final_df = pd.merge(df_sl, df_pivot, on="ID", how="inner")

    final_df.to_csv(PATH_OUTPUT_HDGBVDW, index=False)

    print(f"✅ 保存完了: {PATH_OUTPUT_HDGBVDW}")
    print(f"統合件数: {len(final_df)}")
    print(f"除外ID数: {removed_count}")

    return final_df

df_hdgbvdw = build_integrated_dataset_hdgbvdw(1, 100)
display(df_hdgbvdw.head(3))
```

```python
# =========================================================
# パス設定
# =========================================================

SAVE_DIR = "/Users/teraimao/experiment/data/解析結果/"

PATH_RAW = "/Users/teraimao/experiment/data/all_csv_data/this_is_CycPeptMPDB.csv"

PATH_MASTER = os.path.join(
    SAVE_DIR,
    "一定温度master_database_gbsw8436.csv"
)

PATH_OUTPUT = os.path.join(
    SAVE_DIR,
    "training_dataset_一定_5z_all.csv"
)

# =========================================================
# 設定
# =========================================================

TARGET_T_END_PS = 500
TARGET_Z = [0, 7, 15, 17, 20]

# 使うMD特徴量
target_values = [
    "HB_Avg",
    "SASA_Avg",
    "Solv_Free_Avg",
    "vdW_Avg",
    "Coul_Avg",
]

# =========================================================
# 読み込み
# =========================================================

df_raw = pd.read_csv(PATH_RAW, low_memory=False)
df_md = pd.read_csv(PATH_MASTER, low_memory=False)

print("raw shape   :", df_raw.shape)
print("master shape:", df_md.shape)

print("\nmaster columns:")
print(df_md.columns.tolist())

# =========================================================
# 型変換
# =========================================================

df_md["ID"] = pd.to_numeric(df_md["ID"], errors="coerce")
df_md["Pos_Num"] = pd.to_numeric(df_md["Pos_Num"], errors="coerce")
df_md["t_end_ps"] = pd.to_numeric(df_md["t_end_ps"], errors="coerce")
df_md["Z_Coord"] = pd.to_numeric(df_md["Z_Coord"], errors="coerce")

for c in target_values:
    if c in df_md.columns:
        df_md[c] = pd.to_numeric(df_md[c], errors="coerce")
    else:
        raise ValueError(f"{c} が master CSV にありません")

df_md = df_md.dropna(subset=["ID", "Pos_Num", "t_end_ps", "Z_Coord"]).copy()
df_md["ID"] = df_md["ID"].astype(int)
df_md["Pos_Num"] = df_md["Pos_Num"].astype(int)

# Zを整数ラベル化
df_md["Z_Int"] = df_md["Z_Coord"].round().astype(int)

# =========================================================
# 対象条件でフィルタ
# =========================================================

df_md = df_md[df_md["t_end_ps"] == TARGET_T_END_PS].copy()

if "Status" in df_md.columns:
    df_md = df_md[df_md["Status"] == "Success"].copy()

df_md = df_md[df_md["Z_Int"].isin(TARGET_Z)].copy()

print("\nAfter filter:")
print("shape:", df_md.shape)
print("ID数:", df_md["ID"].nunique())
print("Z counts:")
print(df_md["Z_Int"].value_counts().sort_index())

# =========================================================
# 同じID・Date内で5Z点そろっているworkを選ぶ
# =========================================================
# Dateが複数あるIDがあるため、
# 1 IDにつき「5Z点がそろっている最新Date」を採用する。
# =========================================================

# Dateがない場合に備える
if "Date" not in df_md.columns:
    df_md["Date"] = "unknown"

# 同じID, Date, Zで重複がある場合は最後を採用
df_md = df_md.sort_values(["ID", "Date", "Z_Int", "Pos_Num"]).copy()
df_md = df_md.drop_duplicates(
    subset=["ID", "Date", "Z_Int", "t_end_ps"],
    keep="last"
)

# ID-Dateごとに何種類のZがあるか確認
z_count_by_work = (
    df_md
    .groupby(["ID", "Date"])["Z_Int"]
    .nunique()
    .reset_index(name="Z_count")
)

complete_work = z_count_by_work[z_count_by_work["Z_count"] == len(TARGET_Z)].copy()

print("\n5Z complete work数:", len(complete_work))
print("5Z complete ID数:", complete_work["ID"].nunique())

# 5Zそろったworkだけ残す
df_md_complete = df_md.merge(
    complete_work[["ID", "Date"]],
    on=["ID", "Date"],
    how="inner"
)

# 1 IDに複数Dateがある場合、最新Dateを採用
latest_date_by_id = (
    df_md_complete
    .groupby("ID")["Date"]
    .max()
    .reset_index()
)

df_md_final = df_md_complete.merge(
    latest_date_by_id,
    on=["ID", "Date"],
    how="inner"
)

print("\nAfter selecting latest complete work:")
print("shape:", df_md_final.shape)
print("ID数:", df_md_final["ID"].nunique())

# 念のため最終確認
check_z = df_md_final.groupby("ID")["Z_Int"].nunique()
valid_ids = check_z[check_z == len(TARGET_Z)].index
df_md_final = df_md_final[df_md_final["ID"].isin(valid_ids)].copy()

print("\nFinal valid ID数:", df_md_final["ID"].nunique())
print("Expected rows:", df_md_final["ID"].nunique() * len(TARGET_Z))
print("Actual rows  :", len(df_md_final))

# =========================================================
# 横持ち化
# =========================================================

df_pivot = df_md_final.pivot(
    index="ID",
    columns="Z_Int",
    values=target_values
)

# 列名を HB_Avg_z0 形式にする
new_cols = []
for val, z in df_pivot.columns:
    new_cols.append(f"{val}_z{int(z)}")

df_pivot.columns = new_cols
df_pivot = df_pivot.reset_index()

print("\nPivot shape:", df_pivot.shape)
display(df_pivot.head())

# =========================================================
# rawデータと統合
# =========================================================

df_raw["ID"] = pd.to_numeric(df_raw["ID"], errors="coerce")
df_raw = df_raw.dropna(subset=["ID"]).copy()
df_raw["ID"] = df_raw["ID"].astype(int)

final_df = pd.merge(
    df_raw,
    df_pivot,
    on="ID",
    how="inner"
)

print("\nIntegrated dataset:")
print("shape:", final_df.shape)
print("ID数:", final_df["ID"].nunique())

# =========================================================
# 保存
# =========================================================

final_df.to_csv(PATH_OUTPUT, index=False)

print("\n✅ 保存完了:")
print(PATH_OUTPUT)

display(final_df.head())
```

```python
# =========================================================
# MD-only comparison:
# GBSW only / HDGB only / HDGBvdW only
# ID 1-100
# 
# 出力:
# 1. common_HB_Solv 比較
# 2. all_available_MD 比較
# =========================================================

# =========================================================
# 0. パス設定
# =========================================================

SAVE_DIR = "/Users/teraimao/experiment/data/解析結果/"
OUT_DIR = os.path.join(SAVE_DIR, "MD_only_GBSW_HDGB_HDGBvdW_ID1_100")
os.makedirs(OUT_DIR, exist_ok=True)

PATHS = {
    "GBSW": os.path.join(SAVE_DIR, "training_dataset_const500_5zMatched_ID1_100.csv"),
    "HDGB": os.path.join(SAVE_DIR, "training_dataset_hdgb_500ps_ID1_100.csv"),
    "HDGBvdW": os.path.join(SAVE_DIR, "training_dataset_hdgbvdw_500ps_ID1_100.csv"),
}

ID_MIN = 1
ID_MAX = 100
RANDOM_STATE = 42
N_SPLITS = 5

# =========================================================
# 1. ユーティリティ
# =========================================================

def resolve_path(path):
    """
    拡張子なしで指定した場合にも対応。
    """
    if os.path.exists(path):
        return path
    if not path.endswith(".csv") and os.path.exists(path + ".csv"):
        return path + ".csv"
    return path


def find_target_column(df):
    """
    目的変数列を自動検出。
    優先順位は PAMPA / Permeability 系。
    """
    candidates = [
        "PAMPA",
        "Permeability",
        "logPm",
        "LogPm",
        "logPM",
        "logPapp",
        "target",
        "y",
    ]
    for c in candidates:
        if c in df.columns:
            return c

    raise ValueError(
        "目的変数列が見つかりません。候補: "
        + str(candidates)
        + "\n現在の列: "
        + str(df.columns.tolist())
    )


def get_cols_by_prefix(df, prefixes):
    cols = []
    for c in df.columns:
        for p in prefixes:
            if c.startswith(p):
                cols.append(c)
                break
    return cols


def get_md_features(df, mode="all_available"):
    """
    mode:
      common_HB_Solv:
          全モデル共通の最小MD特徴量
          HB_Avg_z*, Solv_Free_Avg_z*
      
      all_available:
          CSVに存在するMD特徴量を全部使う
          HB_Avg_z*, SASA_Avg_z*, Solv_Free_Avg_z*, vdW_Avg_z*, Coul_Avg_z*
    """
    if mode == "common_HB_Solv":
        prefixes = [
            "HB_Avg_z",
            "Solv_Free_Avg_z",
        ]

    elif mode == "all_available":
        prefixes = [
            "HB_Avg_z",
            "SASA_Avg_z",
            "Solv_Free_Avg_z",
            "vdW_Avg_z",
            "Coul_Avg_z",
        ]

    else:
        raise ValueError(f"Unknown mode: {mode}")

    cols = get_cols_by_prefix(df, prefixes)

    # 念のためIDや目的変数っぽい列は除外
    leakage_like = {
        "ID", "SMILES", "PAMPA", "Permeability",
        "logPm", "LogPm", "logPM", "logPapp",
        "target", "y"
    }
    cols = [c for c in cols if c not in leakage_like]

    return cols


def prepare_model_df(df, model_name, mode):
    """
    1モデル分のデータを整形。
    ID, y, featureのみ残す。
    """
    df = df.copy()

    if "ID" not in df.columns:
        raise ValueError(f"{model_name}: ID列がありません")

    target_col = find_target_column(df)
    feature_cols = get_md_features(df, mode=mode)

    if len(feature_cols) == 0:
        raise ValueError(f"{model_name}: {mode} でMD特徴量が見つかりません")

    df = df[(df["ID"] >= ID_MIN) & (df["ID"] <= ID_MAX)].copy()

    keep_cols = ["ID", target_col] + feature_cols
    out = df[keep_cols].copy()

    # 数値化
    out["ID"] = pd.to_numeric(out["ID"], errors="coerce")
    out[target_col] = pd.to_numeric(out[target_col], errors="coerce")

    for c in feature_cols:
        out[c] = pd.to_numeric(out[c], errors="coerce")

    # 欠損除去
    out = out.dropna(subset=["ID", target_col] + feature_cols).copy()

    # IDは整数化
    out["ID"] = out["ID"].astype(int)

    # ID重複があれば最後を採用
    out = out.drop_duplicates(subset=["ID"], keep="last").copy()

    # 目的変数名を共通化
    out = out.rename(columns={target_col: "y"})

    return out, feature_cols


def evaluate_one_model(model_df, feature_cols, model_name, mode, common_ids):
    """
    common_idsにそろえてRandomForest 5-fold CV。
    """
    use_df = model_df[model_df["ID"].isin(common_ids)].copy()
    use_df = use_df.sort_values("ID").reset_index(drop=True)

    X = use_df[feature_cols].values
    y = use_df["y"].values

    n = len(use_df)

    if n < 5:
        raise ValueError(f"{model_name}: Nが少なすぎます N={n}")

    k = min(N_SPLITS, n)

    cv = KFold(
        n_splits=k,
        shuffle=True,
        random_state=RANDOM_STATE
    )

    model = RandomForestRegressor(
        n_estimators=500,
        random_state=RANDOM_STATE,
        min_samples_leaf=2,
        n_jobs=-1
    )

    y_pred = cross_val_predict(model, X, y, cv=cv)

    mae = mean_absolute_error(y, y_pred)
    rmse = np.sqrt(mean_squared_error(y, y_pred))
    r2 = r2_score(y, y_pred)

    if np.std(y) == 0 or np.std(y_pred) == 0:
        r = np.nan
        p = np.nan
    else:
        r, p = pearsonr(y, y_pred)

    summary = {
        "Mode": mode,
        "Model": model_name,
        "N": n,
        "Feature_N": len(feature_cols),
        "Pearson_R": r,
        "R2": r2,
        "MAE": mae,
        "RMSE": rmse,
        "p_value": p,
        "Features": ",".join(feature_cols),
    }

    pred_df = pd.DataFrame({
        "Mode": mode,
        "Model": model_name,
        "ID": use_df["ID"].values,
        "y_true": y,
        "y_pred": y_pred,
        "error": y_pred - y,
    })

    return summary, pred_df


def run_comparison(mode):
    print("\n" + "=" * 80)
    print(f"Running MD-only comparison mode: {mode}")
    print("=" * 80)

    prepared = {}
    feature_dict = {}

    # -----------------------------
    # 1. 読み込み & 整形
    # -----------------------------
    for model_name, path in PATHS.items():
        path = resolve_path(path)

        if not os.path.exists(path):
            raise FileNotFoundError(f"{model_name}: ファイルが見つかりません: {path}")

        df = pd.read_csv(path, low_memory=False)

        model_df, feature_cols = prepare_model_df(
            df,
            model_name=model_name,
            mode=mode
        )

        prepared[model_name] = model_df
        feature_dict[model_name] = feature_cols

        print(f"\n[{model_name}]")
        print("path:", path)
        print("usable N before common-ID filtering:", len(model_df))
        print("feature N:", len(feature_cols))
        print("features:")
        for c in feature_cols:
            print("  -", c)

    # -----------------------------
    # 2. 共通IDにそろえる
    # -----------------------------
    id_sets = [
        set(prepared[m]["ID"].tolist())
        for m in prepared.keys()
    ]

    common_ids = sorted(set.intersection(*id_sets))

    print("\nCommon IDs:")
    print("N =", len(common_ids))
    print(common_ids[:20], "..." if len(common_ids) > 20 else "")

    if len(common_ids) < 5:
        raise ValueError("3モデル共通で使えるIDが少なすぎます")

    # -----------------------------
    # 3. 評価
    # -----------------------------
    summaries = []
    predictions = []

    for model_name in ["GBSW", "HDGB", "HDGBvdW"]:
        summary, pred_df = evaluate_one_model(
            model_df=prepared[model_name],
            feature_cols=feature_dict[model_name],
            model_name=model_name,
            mode=mode,
            common_ids=common_ids
        )

        summaries.append(summary)
        predictions.append(pred_df)

        print("\n" + "-" * 60)
        print(model_name)
        print("-" * 60)
        print(f"N         = {summary['N']}")
        print(f"Feature_N = {summary['Feature_N']}")
        print(f"Pearson R = {summary['Pearson_R']:.4f}")
        print(f"R2        = {summary['R2']:.4f}")
        print(f"MAE       = {summary['MAE']:.4f}")
        print(f"RMSE      = {summary['RMSE']:.4f}")

    summary_df = pd.DataFrame(summaries)
    pred_all_df = pd.concat(predictions, ignore_index=True)

    summary_df = summary_df.sort_values(
        ["Pearson_R", "RMSE"],
        ascending=[False, True]
    ).reset_index(drop=True)

    return summary_df, pred_all_df


# =========================================================
# 2. 実行
# =========================================================

all_summary = []
all_predictions = []

for mode in ["common_HB_Solv", "all_available"]:
    summary_df, pred_df = run_comparison(mode)
    all_summary.append(summary_df)
    all_predictions.append(pred_df)

summary_all_df = pd.concat(all_summary, ignore_index=True)
prediction_all_df = pd.concat(all_predictions, ignore_index=True)

# =========================================================
# 3. 保存
# =========================================================

summary_path = os.path.join(
    OUT_DIR,
    "MD_only_GBSW_HDGB_HDGBvdW_summary_ID1_100.csv"
)

prediction_path = os.path.join(
    OUT_DIR,
    "MD_only_GBSW_HDGB_HDGBvdW_predictions_ID1_100.csv"
)

summary_all_df.to_csv(summary_path, index=False)
prediction_all_df.to_csv(prediction_path, index=False)

print("\n\n" + "=" * 80)
print("FINAL SUMMARY")
print("=" * 80)

display(summary_all_df[
    ["Mode", "Model", "N", "Feature_N", "Pearson_R", "R2", "MAE", "RMSE"]
])

print("\n✅ saved:")
print(summary_path)
print(prediction_path)
```

#### モデル比較 Repeated CV

```python
SAVE_DIR = "/Users/teraimao/experiment/data/解析結果/"

PATHS = {
    "GBSW": os.path.join(SAVE_DIR, "training_dataset_一定_5z_all.csv"),
    "HDGB": os.path.join(SAVE_DIR, "training_dataset_hdgb_500ps_ID1_100.csv"),
    "HDGBvdW": os.path.join(SAVE_DIR, "training_dataset_hdgbvdw_500ps_ID1_100.csv"),
}

OUT_DIR = os.path.join(SAVE_DIR, "MD_model_comparison_repeatedCV_ID1_100")
os.makedirs(OUT_DIR, exist_ok=True)

TARGET_Z = [0, 7, 15, 17, 20]

FEATURE_BASES = {
    "all_available_MD": {
        "GBSW": ["HB_Avg", "SASA_Avg", "Solv_Free_Avg", "vdW_Avg", "Coul_Avg"],
        "HDGB": ["HB_Avg", "SASA_Avg", "Solv_Free_Avg", "vdW_Avg", "Coul_Avg"],
        "HDGBvdW": ["HB_Avg", "Solv_Free_Avg", "vdW_Avg", "Coul_Avg"],
    },
    "common_noSASA": {
        "GBSW": ["HB_Avg", "Solv_Free_Avg", "vdW_Avg", "Coul_Avg"],
        "HDGB": ["HB_Avg", "Solv_Free_Avg", "vdW_Avg", "Coul_Avg"],
        "HDGBvdW": ["HB_Avg", "Solv_Free_Avg", "vdW_Avg", "Coul_Avg"],
    }
}

N_SPLITS = 5
SEEDS = range(50)


def find_target_column(df):
    candidates = ["PAMPA", "Permeability", "logPm", "LogPm", "logPM", "target", "y"]
    for c in candidates:
        if c in df.columns:
            return c
    raise ValueError("目的変数列が見つかりません")


def get_feature_cols(df, model_name, mode):
    bases = FEATURE_BASES[mode][model_name]
    cols = []
    missing = []

    for base in bases:
        for z in TARGET_Z:
            col = f"{base}_z{z}"
            if col in df.columns:
                cols.append(col)
            else:
                missing.append(col)

    if missing:
        print(f"\n⚠️ {model_name} missing columns in {mode}:")
        for c in missing:
            print("  -", c)

    return cols


def prepare_dataset(model_name, path, mode):
    df = pd.read_csv(path, low_memory=False)

    print("\n" + "=" * 80)
    print(model_name, mode)
    print("=" * 80)
    print("path :", path)
    print("shape:", df.shape)

    target_col = find_target_column(df)
    feature_cols = get_feature_cols(df, model_name, mode)

    use_cols = ["ID", target_col] + feature_cols
    use_df = df[use_cols].copy()

    use_df["ID"] = pd.to_numeric(use_df["ID"], errors="coerce")
    use_df[target_col] = pd.to_numeric(use_df[target_col], errors="coerce")

    for c in feature_cols:
        use_df[c] = pd.to_numeric(use_df[c], errors="coerce")

    before = len(use_df)

    use_df = use_df.dropna(subset=["ID", target_col] + feature_cols).copy()
    use_df["ID"] = use_df["ID"].astype(int)

    # ID1〜100に制限
    use_df = use_df[(use_df["ID"] >= 1) & (use_df["ID"] <= 100)].copy()

    # ID重複があれば最後を採用
    use_df = use_df.drop_duplicates(subset=["ID"], keep="last").copy()

    use_df = use_df.rename(columns={target_col: "y"})

    print("target_col:", target_col)
    print("feature N :", len(feature_cols))
    print("N before dropna:", before)
    print("N usable       :", len(use_df))
    print("ID range       :", use_df["ID"].min(), "~", use_df["ID"].max())

    return use_df, feature_cols


def evaluate_repeated(df, feature_cols, model_name, mode, common_ids):
    use_df = df[df["ID"].isin(common_ids)].copy()
    use_df = use_df.sort_values("ID").reset_index(drop=True)

    X = use_df[feature_cols].values
    y = use_df["y"].values

    rows = []
    pred_rows = []

    for seed in SEEDS:
        cv = KFold(
            n_splits=N_SPLITS,
            shuffle=True,
            random_state=seed
        )

        rf = RandomForestRegressor(
            n_estimators=500,
            random_state=seed,
            min_samples_leaf=2,
            n_jobs=-1
        )

        y_pred = cross_val_predict(rf, X, y, cv=cv)

        r, p = pearsonr(y, y_pred)
        r2 = r2_score(y, y_pred)
        mae = mean_absolute_error(y, y_pred)
        rmse = np.sqrt(mean_squared_error(y, y_pred))

        rows.append({
            "Mode": mode,
            "Model": model_name,
            "Seed": seed,
            "N": len(use_df),
            "Feature_N": len(feature_cols),
            "Pearson_R": r,
            "R2": r2,
            "MAE": mae,
            "RMSE": rmse,
            "p_value": p,
        })

        tmp_pred = pd.DataFrame({
            "Mode": mode,
            "Model": model_name,
            "Seed": seed,
            "ID": use_df["ID"].values,
            "y_true": y,
            "y_pred": y_pred,
            "error": y_pred - y,
        })
        pred_rows.append(tmp_pred)

    return pd.DataFrame(rows), pd.concat(pred_rows, ignore_index=True)


all_results = []
all_preds = []

for mode in ["all_available_MD", "common_noSASA"]:
    print("\n\n" + "#" * 90)
    print("MODE:", mode)
    print("#" * 90)

    datasets = {}
    feature_sets = {}

    for model_name, path in PATHS.items():
        use_df, feature_cols = prepare_dataset(model_name, path, mode)
        datasets[model_name] = use_df
        feature_sets[model_name] = feature_cols

    common_ids = set(datasets["GBSW"]["ID"])
    common_ids &= set(datasets["HDGB"]["ID"])
    common_ids &= set(datasets["HDGBvdW"]["ID"])
    common_ids = sorted(common_ids)

    print("\nCommon ID N:", len(common_ids))
    print("Common ID range:", min(common_ids), "~", max(common_ids))

    for model_name in ["GBSW", "HDGB", "HDGBvdW"]:
        result_df, pred_df = evaluate_repeated(
            df=datasets[model_name],
            feature_cols=feature_sets[model_name],
            model_name=model_name,
            mode=mode,
            common_ids=common_ids
        )

        all_results.append(result_df)
        all_preds.append(pred_df)

results_all = pd.concat(all_results, ignore_index=True)
preds_all = pd.concat(all_preds, ignore_index=True)

summary = (
    results_all
    .groupby(["Mode", "Model"])
    .agg(
        N=("N", "first"),
        Feature_N=("Feature_N", "first"),
        R_mean=("Pearson_R", "mean"),
        R_std=("Pearson_R", "std"),
        R2_mean=("R2", "mean"),
        R2_std=("R2", "std"),
        MAE_mean=("MAE", "mean"),
        MAE_std=("MAE", "std"),
        RMSE_mean=("RMSE", "mean"),
        RMSE_std=("RMSE", "std"),
    )
    .reset_index()
    .sort_values(["Mode", "R_mean"], ascending=[True, False])
)

display(summary)

out_summary = os.path.join(OUT_DIR, "MD_model_comparison_repeatedCV_summary.csv")
out_all = os.path.join(OUT_DIR, "MD_model_comparison_repeatedCV_all_seeds.csv")
out_pred = os.path.join(OUT_DIR, "MD_model_comparison_repeatedCV_predictions.csv")

summary.to_csv(out_summary, index=False)
results_all.to_csv(out_all, index=False)
preds_all.to_csv(out_pred, index=False)

print("\n✅ saved:")
print(out_summary)
print(out_all)
print(out_pred)
```

```python
# =========================================================
# GBSW only prediction
# training_dataset_一定_5z_all
# constant temperature / 5 z positions / ~1000 samples
# =========================================================
# =========================================================
# 0. パス設定
# =========================================================

SAVE_DIR = "/Users/teraimao/experiment/data/解析結果/"

PATH_GBSW = os.path.join(SAVE_DIR, "training_dataset_一定_5z_all.csv")

OUT_DIR = os.path.join(SAVE_DIR, "GBSW_only_const_5z_all")
os.makedirs(OUT_DIR, exist_ok=True)

OUT_SUMMARY = os.path.join(OUT_DIR, "GBSW_only_const_5z_all_summary.csv")
OUT_PRED = os.path.join(OUT_DIR, "GBSW_only_const_5z_all_predictions.csv")

# =========================================================
# 1. ファイル読み込み
# =========================================================

if not os.path.exists(PATH_GBSW):
    # 拡張子なしで存在する場合にも対応
    alt_path = PATH_GBSW.replace(".csv", "")
    if os.path.exists(alt_path):
        PATH_GBSW = alt_path
    else:
        raise FileNotFoundError(f"ファイルが見つかりません: {PATH_GBSW}")

df = pd.read_csv(PATH_GBSW, low_memory=False)

print("Loaded:", PATH_GBSW)
print("shape:", df.shape)
print("columns:")
print(df.columns.tolist())

# =========================================================
# 2. 目的変数列の検出
# =========================================================

target_candidates = [
    "PAMPA",
    "Permeability",
    "logPm",
    "LogPm",
    "logPM",
    "logPapp",
    "target",
    "y",
]

target_col = None
for c in target_candidates:
    if c in df.columns:
        target_col = c
        break

if target_col is None:
    raise ValueError(
        "目的変数列が見つかりません。候補: "
        + str(target_candidates)
    )

print("\nTarget column:", target_col)

# =========================================================
# 3. GBSW MD-only特徴量を選ぶ
# =========================================================
# baseline leakageを避けるため、MD由来列だけ使う
#
# core:
#   HB, SASA, Solv_Free
#
# all_available:
#   HB, SASA, Solv_Free, vdW, Coul
#   GBSW CSVにvdW/Coulがあるなら使う
# =========================================================

feature_modes = {
    "core_HB_SASA_Solv": [
        "HB_Avg_z",
        "SASA_Avg_z",
        "Solv_Free_Avg_z",
    ],
    "all_available_MD": [
        "HB_Avg_z",
        "SASA_Avg_z",
        "Solv_Free_Avg_z",
        "vdW_Avg_z",
        "Coul_Avg_z",
    ],
}

leakage_like_cols = {
    "ID", "SMILES", "PAMPA", "Permeability",
    "logPm", "LogPm", "logPM", "logPapp",
    "target", "y",
}

def get_feature_cols(df, prefixes):
    cols = [
        c for c in df.columns
        if any(c.startswith(p) for p in prefixes)
    ]
    cols = [c for c in cols if c not in leakage_like_cols]
    return cols

# =========================================================
# 4. 評価関数
# =========================================================

def evaluate_gbsw_only(df, feature_cols, mode_name, n_splits=5, random_state=42):
    use_cols = []

    if "ID" in df.columns:
        use_cols.append("ID")

    use_cols += [target_col] + feature_cols

    use_df = df[use_cols].copy()

    # 数値化
    use_df[target_col] = pd.to_numeric(use_df[target_col], errors="coerce")
    for c in feature_cols:
        use_df[c] = pd.to_numeric(use_df[c], errors="coerce")

    # 欠損除去
    before_n = len(use_df)
    use_df = use_df.dropna(subset=[target_col] + feature_cols).copy()
    after_n = len(use_df)

    # ID重複があれば最後を採用
    if "ID" in use_df.columns:
        use_df = use_df.drop_duplicates(subset=["ID"], keep="last").copy()

    print("\n" + "=" * 70)
    print(f"Mode: {mode_name}")
    print("=" * 70)
    print("Features:")
    for c in feature_cols:
        print("  -", c)

    print(f"\nN before dropna: {before_n}")
    print(f"N after dropna : {after_n}")
    print(f"N after duplicate removal: {len(use_df)}")

    if len(use_df) < 5:
        raise ValueError(f"{mode_name}: usable sampleが少なすぎます N={len(use_df)}")

    X = use_df[feature_cols].values
    y = use_df[target_col].values

    k = min(n_splits, len(use_df))

    cv = KFold(
        n_splits=k,
        shuffle=True,
        random_state=random_state
    )

    model = RandomForestRegressor(
        n_estimators=500,
        random_state=random_state,
        min_samples_leaf=2,
        n_jobs=-1
    )

    y_pred = cross_val_predict(model, X, y, cv=cv)

    mae = mean_absolute_error(y, y_pred)
    rmse = np.sqrt(mean_squared_error(y, y_pred))
    r2 = r2_score(y, y_pred)

    if np.std(y) == 0 or np.std(y_pred) == 0:
        r, p = np.nan, np.nan
    else:
        r, p = pearsonr(y, y_pred)

    summary = {
        "Mode": mode_name,
        "N": len(use_df),
        "Feature_N": len(feature_cols),
        "Pearson_R": r,
        "R2": r2,
        "MAE": mae,
        "RMSE": rmse,
        "p_value": p,
        "Features": ",".join(feature_cols),
    }

    pred_df = use_df.copy()
    pred_df["Mode"] = mode_name
    pred_df["Pred_GBSW_only"] = y_pred
    pred_df["Error"] = pred_df["Pred_GBSW_only"] - pred_df[target_col]

    print("\n===== Result =====")
    print(f"N         = {summary['N']}")
    print(f"Feature_N = {summary['Feature_N']}")
    print(f"Pearson R = {summary['Pearson_R']:.4f}")
    print(f"R2        = {summary['R2']:.4f}")
    print(f"MAE       = {summary['MAE']:.4f}")
    print(f"RMSE      = {summary['RMSE']:.4f}")
    print(f"p-value   = {summary['p_value']:.3e}")

    return summary, pred_df

# =========================================================
# 5. 実行
# =========================================================

summary_rows = []
pred_rows = []

for mode_name, prefixes in feature_modes.items():
    feature_cols = get_feature_cols(df, prefixes)

    if len(feature_cols) == 0:
        print(f"[SKIP] {mode_name}: 特徴量が見つかりません")
        continue

    summary, pred_df = evaluate_gbsw_only(
        df=df,
        feature_cols=feature_cols,
        mode_name=mode_name,
        n_splits=5,
        random_state=42,
    )

    summary_rows.append(summary)
    pred_rows.append(pred_df)

summary_df = pd.DataFrame(summary_rows)

if len(summary_df) > 0:
    summary_df = summary_df.sort_values(
        ["Pearson_R", "RMSE"],
        ascending=[False, True]
    ).reset_index(drop=True)

    summary_df.to_csv(OUT_SUMMARY, index=False)

    print("\n\n" + "=" * 70)
    print("FINAL SUMMARY")
    print("=" * 70)
    display(summary_df[["Mode", "N", "Feature_N", "Pearson_R", "R2", "MAE", "RMSE"]])
    print("\n✅ summary saved:")
    print(OUT_SUMMARY)

if len(pred_rows) > 0:
    all_pred_df = pd.concat(pred_rows, ignore_index=True)
    all_pred_df.to_csv(OUT_PRED, index=False)

    print("\n✅ predictions saved:")
    print(OUT_PRED)
    display(all_pred_df.head(10))
```

### 8000件　一定と変温の比較

```python
# =========================================================
# GBSW only comparison
# 変温 vs 一定
# training_dataset_変温_5z_all
# training_dataset_一定_5z_all
# =========================================================
# =========================================================
# 0. パス設定
# =========================================================

SAVE_DIR = "/Users/teraimao/experiment/data/解析結果/"

PATHS = {
    "GBSW_anneal": os.path.join(SAVE_DIR, "training_dataset_変温_5z_all.csv"),
    "GBSW_constant": os.path.join(SAVE_DIR, "training_dataset_一定_5z_all.csv"),
}

OUT_DIR = os.path.join(SAVE_DIR, "GBSW_only_constant_vs_anneal_5z_all")
os.makedirs(OUT_DIR, exist_ok=True)

OUT_SUMMARY = os.path.join(OUT_DIR, "GBSW_only_constant_vs_anneal_summary.csv")
OUT_PRED = os.path.join(OUT_DIR, "GBSW_only_constant_vs_anneal_predictions.csv")

TARGET_Z = [0, 7, 15, 17, 20]

RANDOM_STATE = 42
N_SPLITS = 5

# =========================================================
# 1. 関数
# =========================================================

def resolve_path(path):
    if os.path.exists(path):
        return path
    if not path.endswith(".csv") and os.path.exists(path + ".csv"):
        return path + ".csv"
    raise FileNotFoundError(path)


def find_target_column(df):
    candidates = [
        "PAMPA",
        "Permeability",
        "logPm",
        "LogPm",
        "logPM",
        "logPapp",
        "target",
        "y",
    ]
    for c in candidates:
        if c in df.columns:
            return c
    raise ValueError(f"目的変数列が見つかりません: {df.columns.tolist()}")


def get_md_feature_cols(df, mode):
    """
    mode:
      core_HB_SASA_Solv:
          HB_Avg_z*, SASA_Avg_z*, Solv_Free_Avg_z*
      all_available_MD:
          HB_Avg_z*, SASA_Avg_z*, Solv_Free_Avg_z*, vdW_Avg_z*, Coul_Avg_z*
    """
    if mode == "core_HB_SASA_Solv":
        bases = [
            "HB_Avg",
            "SASA_Avg",
            "Solv_Free_Avg",
        ]
    elif mode == "all_available_MD":
        bases = [
            "HB_Avg",
            "SASA_Avg",
            "Solv_Free_Avg",
            "vdW_Avg",
            "Coul_Avg",
        ]
    else:
        raise ValueError(mode)

    cols = []
    missing = []

    for base in bases:
        for z in TARGET_Z:
            col = f"{base}_z{z}"
            if col in df.columns:
                cols.append(col)
            else:
                missing.append(col)

    return cols, missing


def prepare_df(path, dataset_name, mode):
    path = resolve_path(path)
    df = pd.read_csv(path, low_memory=False)

    print("\n" + "=" * 80)
    print(dataset_name)
    print("=" * 80)
    print("path:", path)
    print("shape:", df.shape)

    if "ID" not in df.columns:
        raise ValueError(f"{dataset_name}: ID列がありません")

    target_col = find_target_column(df)
    feature_cols, missing_cols = get_md_feature_cols(df, mode)

    print("target_col:", target_col)
    print("feature N :", len(feature_cols))

    if missing_cols:
        print("missing columns:")
        for c in missing_cols:
            print("  -", c)

    if len(feature_cols) == 0:
        raise ValueError(f"{dataset_name}: {mode} のMD特徴量がありません")

    use_cols = ["ID", target_col] + feature_cols
    use_df = df[use_cols].copy()

    use_df["ID"] = pd.to_numeric(use_df["ID"], errors="coerce")
    use_df[target_col] = pd.to_numeric(use_df[target_col], errors="coerce")

    for c in feature_cols:
        use_df[c] = pd.to_numeric(use_df[c], errors="coerce")

    before_n = len(use_df)

    use_df = use_df.dropna(subset=["ID", target_col] + feature_cols).copy()

    after_dropna_n = len(use_df)

    use_df["ID"] = use_df["ID"].astype(int)

    # ID重複があれば最後を採用
    use_df = use_df.drop_duplicates(subset=["ID"], keep="last").copy()

    final_n = len(use_df)

    use_df = use_df.rename(columns={target_col: "y"})

    print("N before dropna      :", before_n)
    print("N after dropna       :", after_dropna_n)
    print("N after duplicate rm :", final_n)

    return use_df, feature_cols


def evaluate_one_dataset(df, feature_cols, dataset_name, mode, common_ids=None):
    if common_ids is not None:
        use_df = df[df["ID"].isin(common_ids)].copy()
    else:
        use_df = df.copy()

    use_df = use_df.sort_values("ID").reset_index(drop=True)

    X = use_df[feature_cols].values
    y = use_df["y"].values

    n = len(use_df)

    if n < 5:
        raise ValueError(f"{dataset_name}: Nが少なすぎます N={n}")

    k = min(N_SPLITS, n)

    cv = KFold(
        n_splits=k,
        shuffle=True,
        random_state=RANDOM_STATE
    )

    model = RandomForestRegressor(
        n_estimators=500,
        random_state=RANDOM_STATE,
        min_samples_leaf=2,
        n_jobs=-1
    )

    y_pred = cross_val_predict(model, X, y, cv=cv)

    mae = mean_absolute_error(y, y_pred)
    rmse = np.sqrt(mean_squared_error(y, y_pred))
    r2 = r2_score(y, y_pred)

    if np.std(y) == 0 or np.std(y_pred) == 0:
        r, p = np.nan, np.nan
    else:
        r, p = pearsonr(y, y_pred)

    summary = {
        "Dataset": dataset_name,
        "Mode": mode,
        "N": n,
        "Feature_N": len(feature_cols),
        "Pearson_R": r,
        "R2": r2,
        "MAE": mae,
        "RMSE": rmse,
        "p_value": p,
        "Features": ",".join(feature_cols),
    }

    pred_df = pd.DataFrame({
        "Dataset": dataset_name,
        "Mode": mode,
        "ID": use_df["ID"].values,
        "y_true": y,
        "y_pred": y_pred,
        "error": y_pred - y,
    })

    return summary, pred_df


def run_mode(mode):
    print("\n\n" + "#" * 90)
    print(f"MODE: {mode}")
    print("#" * 90)

    prepared = {}
    features = {}

    for name, path in PATHS.items():
        use_df, feature_cols = prepare_df(path, name, mode)
        prepared[name] = use_df
        features[name] = feature_cols

    # =====================================================
    # 共通ID
    # =====================================================
    common_ids = sorted(
        set(prepared["GBSW_anneal"]["ID"]).intersection(
            set(prepared["GBSW_constant"]["ID"])
        )
    )

    print("\n" + "=" * 80)
    print("Common ID filtering")
    print("=" * 80)
    print("Anneal usable N  :", len(prepared["GBSW_anneal"]))
    print("Constant usable N:", len(prepared["GBSW_constant"]))
    print("Common ID N      :", len(common_ids))

    if len(common_ids) < 5:
        raise ValueError("共通IDが少なすぎます")

    summaries = []
    preds = []

    for name in ["GBSW_anneal", "GBSW_constant"]:
        # 重要：比較の公平性のため、共通IDだけ使う
        summary, pred_df = evaluate_one_dataset(
            df=prepared[name],
            feature_cols=features[name],
            dataset_name=name,
            mode=mode,
            common_ids=common_ids
        )

        summaries.append(summary)
        preds.append(pred_df)

        print("\n" + "-" * 70)
        print(name)
        print("-" * 70)
        print(f"N         = {summary['N']}")
        print(f"Feature_N = {summary['Feature_N']}")
        print(f"Pearson R = {summary['Pearson_R']:.4f}")
        print(f"R2        = {summary['R2']:.4f}")
        print(f"MAE       = {summary['MAE']:.4f}")
        print(f"RMSE      = {summary['RMSE']:.4f}")

    summary_df = pd.DataFrame(summaries)
    pred_df = pd.concat(preds, ignore_index=True)

    return summary_df, pred_df


# =========================================================
# 2. 実行
# =========================================================

all_summary = []
all_pred = []

for mode in ["core_HB_SASA_Solv", "all_available_MD"]:
    summary_df, pred_df = run_mode(mode)
    all_summary.append(summary_df)
    all_pred.append(pred_df)

summary_all_df = pd.concat(all_summary, ignore_index=True)
pred_all_df = pd.concat(all_pred, ignore_index=True)

# 見やすくソート
summary_all_df = summary_all_df.sort_values(
    ["Mode", "Pearson_R"],
    ascending=[True, False]
).reset_index(drop=True)

# =========================================================
# 3. 保存
# =========================================================

summary_all_df.to_csv(OUT_SUMMARY, index=False)
pred_all_df.to_csv(OUT_PRED, index=False)

print("\n\n" + "=" * 90)
print("FINAL SUMMARY")
print("=" * 90)

display(summary_all_df[
    ["Dataset", "Mode", "N", "Feature_N", "Pearson_R", "R2", "MAE", "RMSE"]
])

print("\n✅ saved:")
print(OUT_SUMMARY)
print(OUT_PRED)
```

```python
# =========================================================
# パス設定
# =========================================================

SAVE_DIR = "/Users/teraimao/experiment/data/解析結果/"

PATHS = {
    "GBSW_cons100": os.path.join(SAVE_DIR, "training_dataset_一定_5z_all.csv"),
    "GBSW_cons50": os.path.join(SAVE_DIR, "training_dataset_cons50_500ps_ID1_100.csv"),
    "GBSW_cons1": os.path.join(SAVE_DIR, "training_dataset_cons1_500ps_ID1_100.csv"),
}

OUT_DIR = os.path.join(SAVE_DIR, "GBSW_only_constraint_comparison_ID1_100")
os.makedirs(OUT_DIR, exist_ok=True)

OUT_SUMMARY = os.path.join(OUT_DIR, "GBSW_only_constraint_summary.csv")
OUT_PRED = os.path.join(OUT_DIR, "GBSW_only_constraint_predictions.csv")

TARGET_Z = [0, 7, 15, 17, 20]

FEATURE_BASES = [
    "HB_Avg",
    "SASA_Avg",
    "Solv_Free_Avg",
    "vdW_Avg",
    "Coul_Avg",
]

RANDOM_STATE = 42
N_SPLITS = 5


# =========================================================
# 関数
# =========================================================

def resolve_path(path):
    if os.path.exists(path):
        return path
    if not path.endswith(".csv") and os.path.exists(path + ".csv"):
        return path + ".csv"
    raise FileNotFoundError(path)


def find_target_column(df):
    candidates = [
        "PAMPA",
        "Permeability",
        "logPm",
        "LogPm",
        "logPM",
        "target",
        "y",
    ]
    for c in candidates:
        if c in df.columns:
            return c
    raise ValueError("目的変数列が見つかりません")


def get_md_feature_cols(df):
    cols = []
    missing = []

    for base in FEATURE_BASES:
        for z in TARGET_Z:
            col = f"{base}_z{z}"
            if col in df.columns:
                cols.append(col)
            else:
                missing.append(col)

    return cols, missing


def prepare_dataset(name, path):
    path = resolve_path(path)
    df = pd.read_csv(path, low_memory=False)

    print("\n" + "=" * 80)
    print(name)
    print("=" * 80)
    print("path :", path)
    print("shape:", df.shape)

    if "ID" not in df.columns:
        raise ValueError(f"{name}: ID列がありません")

    target_col = find_target_column(df)
    feature_cols, missing_cols = get_md_feature_cols(df)

    print("target_col:", target_col)
    print("feature N :", len(feature_cols))

    if missing_cols:
        print("missing columns:")
        for c in missing_cols:
            print("  -", c)

    if len(feature_cols) != 25:
        print("⚠️ all_available_MDが25列そろっていません。確認してください。")

    use_cols = ["ID", target_col] + feature_cols
    use_df = df[use_cols].copy()

    use_df["ID"] = pd.to_numeric(use_df["ID"], errors="coerce")
    use_df[target_col] = pd.to_numeric(use_df[target_col], errors="coerce")

    for c in feature_cols:
        use_df[c] = pd.to_numeric(use_df[c], errors="coerce")

    before = len(use_df)

    use_df = use_df.dropna(subset=["ID", target_col] + feature_cols).copy()
    use_df["ID"] = use_df["ID"].astype(int)

    # cons100の大規模datasetはID1〜100に制限
    if name == "GBSW_cons100":
        use_df = use_df[(use_df["ID"] >= 1) & (use_df["ID"] <= 100)].copy()

    # ID重複があれば最後を採用
    use_df = use_df.drop_duplicates(subset=["ID"], keep="last").copy()

    use_df = use_df.rename(columns={target_col: "y"})

    print("N before dropna:", before)
    print("N usable       :", len(use_df))
    print("ID range       :", use_df["ID"].min(), "~", use_df["ID"].max())

    return use_df, feature_cols


def evaluate_dataset(df, feature_cols, name, common_ids):
    use_df = df[df["ID"].isin(common_ids)].copy()
    use_df = use_df.sort_values("ID").reset_index(drop=True)

    X = use_df[feature_cols].values
    y = use_df["y"].values

    cv = KFold(
        n_splits=min(N_SPLITS, len(use_df)),
        shuffle=True,
        random_state=RANDOM_STATE
    )

    model = RandomForestRegressor(
        n_estimators=500,
        random_state=RANDOM_STATE,
        min_samples_leaf=2,
        n_jobs=-1,
    )

    y_pred = cross_val_predict(model, X, y, cv=cv)

    r, p = pearsonr(y, y_pred)
    r2 = r2_score(y, y_pred)
    mae = mean_absolute_error(y, y_pred)
    rmse = np.sqrt(mean_squared_error(y, y_pred))

    summary = {
        "Dataset": name,
        "N": len(use_df),
        "Feature_N": len(feature_cols),
        "Pearson_R": r,
        "R2": r2,
        "MAE": mae,
        "RMSE": rmse,
        "p_value": p,
        "Features": ",".join(feature_cols),
    }

    pred_df = pd.DataFrame({
        "Dataset": name,
        "ID": use_df["ID"].values,
        "y_true": y,
        "y_pred": y_pred,
        "error": y_pred - y,
    })

    return summary, pred_df


# =========================================================
# 読み込み
# =========================================================

datasets = {}
feature_sets = {}

for name, path in PATHS.items():
    df_use, feature_cols = prepare_dataset(name, path)
    datasets[name] = df_use
    feature_sets[name] = feature_cols


# =========================================================
# 共通IDにそろえる
# =========================================================

common_ids = set(datasets["GBSW_cons100"]["ID"])

for name in ["GBSW_cons50", "GBSW_cons1"]:
    common_ids = common_ids.intersection(set(datasets[name]["ID"]))

common_ids = sorted(common_ids)

print("\n" + "=" * 80)
print("Common ID")
print("=" * 80)

for name, df in datasets.items():
    print(name, "usable N:", len(df))

print("Common ID N:", len(common_ids))
print("Common ID range:", min(common_ids), "~", max(common_ids))

if len(common_ids) < 10:
    raise ValueError("共通IDが少なすぎます")


# =========================================================
# 評価
# =========================================================

summaries = []
preds = []

for name in ["GBSW_cons100", "GBSW_cons50", "GBSW_cons1"]:
    summary, pred_df = evaluate_dataset(
        df=datasets[name],
        feature_cols=feature_sets[name],
        name=name,
        common_ids=common_ids,
    )

    summaries.append(summary)
    preds.append(pred_df)

    print("\n" + "-" * 70)
    print(name)
    print("-" * 70)
    print(f"N         = {summary['N']}")
    print(f"Feature_N = {summary['Feature_N']}")
    print(f"Pearson R = {summary['Pearson_R']:.4f}")
    print(f"R2        = {summary['R2']:.4f}")
    print(f"MAE       = {summary['MAE']:.4f}")
    print(f"RMSE      = {summary['RMSE']:.4f}")


summary_df = pd.DataFrame(summaries)
pred_df = pd.concat(preds, ignore_index=True)

summary_df = summary_df.sort_values("Pearson_R", ascending=False).reset_index(drop=True)

summary_df.to_csv(OUT_SUMMARY, index=False)
pred_df.to_csv(OUT_PRED, index=False)

print("\n" + "=" * 80)
print("FINAL SUMMARY")
print("=" * 80)

display(summary_df[[
    "Dataset", "N", "Feature_N", "Pearson_R", "R2", "MAE", "RMSE"
]])

print("\n✅ saved:")
print(OUT_SUMMARY)
print(OUT_PRED)
```

```python
def evaluate_repeated(df, feature_cols, name, common_ids, seeds=range(50)):
    rows = []

    use_df = df[df["ID"].isin(common_ids)].copy()
    use_df = use_df.sort_values("ID").reset_index(drop=True)

    X = use_df[feature_cols].values
    y = use_df["y"].values

    for seed in seeds:
        cv = KFold(
            n_splits=5,
            shuffle=True,
            random_state=seed
        )

        model = RandomForestRegressor(
            n_estimators=500,
            random_state=seed,
            min_samples_leaf=2,
            n_jobs=-1
        )

        y_pred = cross_val_predict(model, X, y, cv=cv)

        r, p = pearsonr(y, y_pred)
        r2 = r2_score(y, y_pred)
        mae = mean_absolute_error(y, y_pred)
        rmse = np.sqrt(mean_squared_error(y, y_pred))

        rows.append({
            "Dataset": name,
            "Seed": seed,
            "N": len(use_df),
            "Pearson_R": r,
            "R2": r2,
            "MAE": mae,
            "RMSE": rmse,
        })

    return pd.DataFrame(rows)


rep_results = []

for name in ["GBSW_cons100", "GBSW_cons50", "GBSW_cons1"]:
    rep_df = evaluate_repeated(
        df=datasets[name],
        feature_cols=feature_sets[name],
        name=name,
        common_ids=common_ids,
        seeds=range(50)
    )
    rep_results.append(rep_df)

rep_df_all = pd.concat(rep_results, ignore_index=True)

summary_rep = (
    rep_df_all
    .groupby("Dataset")
    .agg(
        R_mean=("Pearson_R", "mean"),
        R_std=("Pearson_R", "std"),
        R2_mean=("R2", "mean"),
        R2_std=("R2", "std"),
        MAE_mean=("MAE", "mean"),
        MAE_std=("MAE", "std"),
        RMSE_mean=("RMSE", "mean"),
        RMSE_std=("RMSE", "std"),
    )
    .reset_index()
    .sort_values("R_mean", ascending=False)
)

display(summary_rep)

out_path = "/Users/teraimao/experiment/data/解析結果/GBSW_only_constraint_comparison_ID1_100/GBSW_only_constraint_repeatedCV_summary.csv"
summary_rep.to_csv(out_path, index=False)

print("saved:", out_path)
```

```python
# =========================================================
# GBSW cons25 / cons50 / cons75 / cons100 Repeated CV 比較
# cons100 = training_dataset_一定_5z_all からID1-300を抽出
# 共通IDのみ / all_available_MD 25特徴量
# =========================================================
# =========================================================
# パス設定
# =========================================================

SAVE_DIR = "/Users/teraimao/experiment/data/解析結果/"
os.makedirs(SAVE_DIR, exist_ok=True)

# cons100の元データ候補
CONS100_SOURCE_CANDIDATES = [
    os.path.join(SAVE_DIR, "training_dataset_一定_5z_all.csv"),
    os.path.join(SAVE_DIR, "training_dataset_一定_5z_all"),
]

# 見つからなければglobで探す
CONS100_SOURCE = None

for p in CONS100_SOURCE_CANDIDATES:
    if os.path.exists(p):
        CONS100_SOURCE = p
        break

if CONS100_SOURCE is None:
    print("候補:")
    for p in CONS100_SOURCE_CANDIDATES:
        print(p)
    raise FileNotFoundError("training_dataset_一定_5z_all.csv が見つかりません。")

print("cons100 source:", CONS100_SOURCE)

# cons100 ID1-300切り出し後の保存先
CONS100_OUT = os.path.join(
    SAVE_DIR,
    "training_dataset_gbsw_cons100_500ps_ID1_300_from_constant5z_all.csv"
)

# 既存のcons条件
PATH_CONS25 = os.path.join(SAVE_DIR, "training_dataset_gbsw_cons25_500ps_ID1_300.csv")
PATH_CONS50 = os.path.join(SAVE_DIR, "training_dataset_gbsw_cons50_500ps_ID1_300_FINAL.csv")
PATH_CONS75 = os.path.join(SAVE_DIR, "training_dataset_gbsw_cons75_500ps_ID1_300.csv")

# 出力
OUT_SUMMARY = os.path.join(SAVE_DIR, "GBSW_constraint_ID1_300_with_cons100_repeatedCV_summary.csv")
OUT_ALL_SEEDS = os.path.join(SAVE_DIR, "GBSW_constraint_ID1_300_with_cons100_repeatedCV_all_seeds.csv")
OUT_PRED = os.path.join(SAVE_DIR, "GBSW_constraint_ID1_300_with_cons100_repeatedCV_predictions.csv")


# =========================================================
# 設定
# =========================================================

N_SPLITS = 5
SEEDS = list(range(50))

TARGET_CANDIDATES = [
    "Permeability",
    "logPm",
    "LogPm",
    "PAMPA",
    "PAMPA_logP",
    "Target",
    "target",
    "y",
]

MD_BASES = [
    "HB_Avg",
    "SASA_Avg",
    "Solv_Free_Avg",
    "vdW_Avg",
    "Coul_Avg",
]

Z_LIST = [0, 7, 15, 17, 20]

MD_FEATURES = [
    f"{base}_z{z}"
    for base in MD_BASES
    for z in Z_LIST
]

print("\nMD features:", len(MD_FEATURES))
print(MD_FEATURES)


# =========================================================
# 補助関数
# =========================================================

def find_target_column(df):
    for c in TARGET_CANDIDATES:
        if c in df.columns:
            return c
    raise ValueError(
        "目的変数列が見つかりません。候補: "
        + ", ".join(TARGET_CANDIDATES)
        + "\n実際の列:\n"
        + str(df.columns.tolist()[:120])
    )


def require_file(path):
    if not os.path.exists(path):
        raise FileNotFoundError(f"ファイルが見つかりません: {path}")


def check_features(df, name):
    missing = [c for c in MD_FEATURES if c not in df.columns]
    if missing:
        raise ValueError(f"{name}: MD特徴量が足りません:\n{missing}")


def make_cons100_id1_300():
    df = pd.read_csv(CONS100_SOURCE, low_memory=False)

    if "ID" not in df.columns:
        raise ValueError("cons100 sourceにID列がありません。")

    df["ID"] = pd.to_numeric(df["ID"], errors="coerce")
    df = df.dropna(subset=["ID"]).copy()
    df["ID"] = df["ID"].astype(int)

    target_col = find_target_column(df)
    check_features(df, "cons100 source")

    # 数値化
    df[target_col] = pd.to_numeric(df[target_col], errors="coerce")
    for c in MD_FEATURES:
        df[c] = pd.to_numeric(df[c], errors="coerce")

    # ID1-300に限定
    sub = df[(df["ID"] >= 1) & (df["ID"] <= 300)].copy()

    # MD特徴量とtargetがそろっているものだけ
    sub = sub.dropna(subset=[target_col] + MD_FEATURES).copy()
    sub = sub.sort_values("ID").drop_duplicates("ID", keep="last").reset_index(drop=True)

    sub.to_csv(CONS100_OUT, index=False)

    print("\n✅ cons100 ID1-300 dataset saved:")
    print(CONS100_OUT)
    print("N:", len(sub))
    print("ID range:", sub["ID"].min(), "~", sub["ID"].max())
    print("missing IDs in 1-300:", sorted(set(range(1, 301)) - set(sub["ID"].tolist())))
    print("target:", target_col)
    print("MD feature columns:", len([c for c in MD_FEATURES if c in sub.columns]))

    return CONS100_OUT


def check_dataset(name, path):
    require_file(path)

    df = pd.read_csv(path, low_memory=False)

    if "ID" not in df.columns:
        raise ValueError(f"{name}: ID列がありません")

    df["ID"] = pd.to_numeric(df["ID"], errors="coerce")
    df = df.dropna(subset=["ID"]).copy()
    df["ID"] = df["ID"].astype(int)

    target_col = find_target_column(df)
    check_features(df, name)

    df[target_col] = pd.to_numeric(df[target_col], errors="coerce")
    for c in MD_FEATURES:
        df[c] = pd.to_numeric(df[c], errors="coerce")

    usable = df.dropna(subset=[target_col] + MD_FEATURES).copy()

    print(f"\n===== {name} =====")
    print("path:", path)
    print("shape:", df.shape)
    print("target:", target_col)
    print("usable N:", len(usable))
    print("ID range:", usable["ID"].min(), "~", usable["ID"].max())
    print("missing MD/target rows:", len(df) - len(usable))
    print("missing IDs in 1-300:", sorted(set(range(1, 301)) - set(usable["ID"].tolist())))

    return df, target_col, set(usable["ID"].tolist())


def make_common_datasets(dataset_paths):
    loaded = {}
    usable_id_sets = []
    target_cols = {}

    for name, path in dataset_paths.items():
        df, target_col, usable_ids = check_dataset(name, path)
        loaded[name] = df
        target_cols[name] = target_col
        usable_id_sets.append(usable_ids)

    common_ids = sorted(set.intersection(*usable_id_sets))

    print("\n==============================")
    print("COMMON ID INFO")
    print("==============================")
    print("Common N:", len(common_ids))
    print("Missing from 1-300:", sorted(set(range(1, 301)) - set(common_ids)))

    data = {}

    for name, df in loaded.items():
        target_col = target_cols[name]

        sub = df[df["ID"].isin(common_ids)].copy()
        sub = sub.dropna(subset=[target_col] + MD_FEATURES).copy()
        sub = sub.sort_values("ID").reset_index(drop=True)

        X = sub[MD_FEATURES].values
        y = sub[target_col].values
        ids = sub["ID"].values

        data[name] = {
            "df": sub,
            "X": X,
            "y": y,
            "ids": ids,
            "target_col": target_col,
        }

        print(f"{name}: final N={len(sub)}")

    return data, common_ids


def calc_metrics(y_true, y_pred):
    if len(y_true) < 2:
        r = np.nan
    else:
        r = pearsonr(y_true, y_pred)[0]

    return {
        "Pearson_R": r,
        "R2": r2_score(y_true, y_pred),
        "MAE": mean_absolute_error(y_true, y_pred),
        "RMSE": np.sqrt(mean_squared_error(y_true, y_pred)),
    }


def repeated_cv_one_dataset(name, X, y, ids):
    all_seed_rows = []
    all_pred_rows = []

    for seed in SEEDS:
        kf = KFold(
            n_splits=N_SPLITS,
            shuffle=True,
            random_state=seed,
        )

        oof_pred = np.full(len(y), np.nan)

        for fold, (train_idx, test_idx) in enumerate(kf.split(X), start=1):
            model = RandomForestRegressor(
                n_estimators=500,
                random_state=seed,
                n_jobs=-1,
                max_features="sqrt",
                min_samples_leaf=1,
            )

            model.fit(X[train_idx], y[train_idx])
            pred = model.predict(X[test_idx])
            oof_pred[test_idx] = pred

            for idx, p in zip(test_idx, pred):
                all_pred_rows.append({
                    "Condition": name,
                    "Seed": seed,
                    "Fold": fold,
                    "ID": int(ids[idx]),
                    "y_true": float(y[idx]),
                    "y_pred": float(p),
                })

        metrics = calc_metrics(y, oof_pred)

        all_seed_rows.append({
            "Condition": name,
            "Seed": seed,
            "N": len(y),
            "Feature_N": X.shape[1],
            **metrics,
        })

    return all_seed_rows, all_pred_rows


# =========================================================
# 実行
# =========================================================

# 1. cons100を作る
PATH_CONS100 = make_cons100_id1_300()

# 2. 4条件比較
DATASETS = {
    "GBSW_cons25": PATH_CONS25,
    "GBSW_cons50": PATH_CONS50,
    "GBSW_cons75": PATH_CONS75,
    "GBSW_cons100": PATH_CONS100,
}

data, common_ids = make_common_datasets(DATASETS)

all_seed_rows = []
all_pred_rows = []

for name, d in data.items():
    print(f"\n🚀 Repeated CV: {name}")
    seed_rows, pred_rows = repeated_cv_one_dataset(
        name=name,
        X=d["X"],
        y=d["y"],
        ids=d["ids"],
    )
    all_seed_rows.extend(seed_rows)
    all_pred_rows.extend(pred_rows)

df_all_seeds = pd.DataFrame(all_seed_rows)
df_pred = pd.DataFrame(all_pred_rows)

summary_rows = []

for condition, g in df_all_seeds.groupby("Condition"):
    row = {
        "Condition": condition,
        "N": int(g["N"].iloc[0]),
        "Feature_N": int(g["Feature_N"].iloc[0]),
    }

    for metric in ["Pearson_R", "R2", "MAE", "RMSE"]:
        row[f"{metric}_mean"] = g[metric].mean()
        row[f"{metric}_std"] = g[metric].std()

    summary_rows.append(row)

df_summary = pd.DataFrame(summary_rows)
df_summary = df_summary.sort_values("Pearson_R_mean", ascending=False).reset_index(drop=True)

df_summary.to_csv(OUT_SUMMARY, index=False)
df_all_seeds.to_csv(OUT_ALL_SEEDS, index=False)
df_pred.to_csv(OUT_PRED, index=False)

print("\n\n==============================")
print("SUMMARY")
print("==============================")
display(df_summary)

print("\n✅ saved:")
print(OUT_SUMMARY)
print(OUT_ALL_SEEDS)
print(OUT_PRED)
```

### GBSWの派生特徴量

```python
# =========================================================
# パス設定
# =========================================================

SAVE_DIR = "/Users/teraimao/experiment/data/解析結果/"

PATHS = {
    "GBSW_cons100": os.path.join(SAVE_DIR, "training_dataset_一定_5z_all.csv"),
    "GBSW_cons50": os.path.join(SAVE_DIR, "training_dataset_cons50_500ps_ID1_100.csv"),
    "GBSW_cons1": os.path.join(SAVE_DIR, "training_dataset_cons1_500ps_ID1_100.csv"),
}

OUT_DIR = os.path.join(SAVE_DIR, "GBSW_constraint_derived_feature_comparison_ID1_100")
os.makedirs(OUT_DIR, exist_ok=True)

OUT_SUMMARY = os.path.join(OUT_DIR, "GBSW_constraint_derived_feature_repeatedCV_summary.csv")
OUT_ALLSEEDS = os.path.join(OUT_DIR, "GBSW_constraint_derived_feature_repeatedCV_all_seeds.csv")
OUT_PRED = os.path.join(OUT_DIR, "GBSW_constraint_derived_feature_repeatedCV_predictions.csv")

TARGET_Z = [0, 7, 15, 17, 20]

BASES = [
    "HB_Avg",
    "SASA_Avg",
    "Solv_Free_Avg",
    "vdW_Avg",
    "Coul_Avg",
]

N_SPLITS = 5
SEEDS = range(50)


# =========================================================
# 基本関数
# =========================================================

def find_target_column(df):
    candidates = ["PAMPA", "Permeability", "logPm", "LogPm", "logPM", "target", "y"]
    for c in candidates:
        if c in df.columns:
            return c
    raise ValueError("目的変数列が見つかりません")


def get_raw_md_cols(df):
    cols = []
    missing = []

    for base in BASES:
        for z in TARGET_Z:
            col = f"{base}_z{z}"
            if col in df.columns:
                cols.append(col)
            else:
                missing.append(col)

    return cols, missing


def add_derived_features(df, bases=BASES, z_list=TARGET_Z):
    """
    zごとの生MD特徴量から、物理的な変化量・統計量を作る。
    """
    out = df.copy()

    for base in bases:
        cols = [f"{base}_z{z}" for z in z_list]

        if not all(c in out.columns for c in cols):
            print(f"skip {base}: missing columns")
            continue

        arr = out[cols].astype(float).values
        z = np.array(z_list, dtype=float)

        z0 = out[f"{base}_z0"].astype(float)
        z7 = out[f"{base}_z7"].astype(float)
        z15 = out[f"{base}_z15"].astype(float)
        z17 = out[f"{base}_z17"].astype(float)
        z20 = out[f"{base}_z20"].astype(float)

        interface = (z15 + z17) / 2.0

        # 基本統計
        out[f"{base}_mean_z"] = np.nanmean(arr, axis=1)
        out[f"{base}_std_z"] = np.nanstd(arr, axis=1)
        out[f"{base}_min_z"] = np.nanmin(arr, axis=1)
        out[f"{base}_max_z"] = np.nanmax(arr, axis=1)
        out[f"{base}_range_z"] = out[f"{base}_max_z"] - out[f"{base}_min_z"]

        # どのZで最大/最小か
        out[f"{base}_argmin_z"] = z[np.nanargmin(arr, axis=1)]
        out[f"{base}_argmax_z"] = z[np.nanargmax(arr, axis=1)]

        # 膜中心・界面・水側の差
        out[f"{base}_z0_minus_z20"] = z0 - z20
        out[f"{base}_z20_minus_z0"] = z20 - z0

        out[f"{base}_z0_minus_interface"] = z0 - interface
        out[f"{base}_interface_minus_z0"] = interface - z0

        out[f"{base}_interface_minus_z20"] = interface - z20
        out[f"{base}_z20_minus_interface"] = z20 - interface

        out[f"{base}_z7_minus_z20"] = z7 - z20
        out[f"{base}_z0_minus_z7"] = z0 - z7

        # 最大値・最小値と水側との差
        out[f"{base}_max_minus_z20"] = out[f"{base}_max_z"] - z20
        out[f"{base}_z20_minus_min"] = z20 - out[f"{base}_min_z"]

        # 変化量の絶対値
        out[f"{base}_abs_z0_minus_z20"] = np.abs(z0 - z20)
        out[f"{base}_abs_interface_minus_z20"] = np.abs(interface - z20)

        # z方向の傾き
        # polyfitは全行まとめて書くと面倒なので内積で一次回帰 slope を計算
        z_centered = z - z.mean()
        denom = np.sum(z_centered ** 2)
        arr_centered = arr - np.nanmean(arr, axis=1, keepdims=True)
        slope = np.nansum(arr_centered * z_centered, axis=1) / denom
        out[f"{base}_linear_slope_z"] = slope

        # 台形積分っぽい全体量
        out[f"{base}_auc_z"] = np.trapz(arr, x=z, axis=1)

    # =====================================================
    # 物理的に名前を付けた派生特徴量
    # =====================================================
    # Solv: 膜中心や最大障壁っぽい量
    if all(f"Solv_Free_Avg_z{z}" in out.columns for z in z_list):
        solv_cols = [f"Solv_Free_Avg_z{z}" for z in z_list]
        solv_arr = out[solv_cols].astype(float).values

        out["Solv_barrier_max_minus_water"] = np.nanmax(solv_arr, axis=1) - out["Solv_Free_Avg_z20"].astype(float)
        out["Solv_center_minus_water"] = out["Solv_Free_Avg_z0"].astype(float) - out["Solv_Free_Avg_z20"].astype(float)
        out["Solv_interface_mean"] = (
            out["Solv_Free_Avg_z15"].astype(float) + out["Solv_Free_Avg_z17"].astype(float)
        ) / 2.0
        out["Solv_center_minus_interface"] = out["Solv_Free_Avg_z0"].astype(float) - out["Solv_interface_mean"]

    # HB: 水側と膜中心でのHbond変化
    if all(f"HB_Avg_z{z}" in out.columns for z in z_list):
        out["HB_water_minus_center"] = out["HB_Avg_z20"].astype(float) - out["HB_Avg_z0"].astype(float)
        out["HB_center_minus_water"] = out["HB_Avg_z0"].astype(float) - out["HB_Avg_z20"].astype(float)
        out["HB_interface_mean"] = (
            out["HB_Avg_z15"].astype(float) + out["HB_Avg_z17"].astype(float)
        ) / 2.0
        out["HB_water_minus_interface"] = out["HB_Avg_z20"].astype(float) - out["HB_interface_mean"]

    # SASA: 膜中心と水側での露出変化
    if all(f"SASA_Avg_z{z}" in out.columns for z in z_list):
        out["SASA_water_minus_center"] = out["SASA_Avg_z20"].astype(float) - out["SASA_Avg_z0"].astype(float)
        out["SASA_center_minus_water"] = out["SASA_Avg_z0"].astype(float) - out["SASA_Avg_z20"].astype(float)
        out["SASA_interface_mean"] = (
            out["SASA_Avg_z15"].astype(float) + out["SASA_Avg_z17"].astype(float)
        ) / 2.0
        out["SASA_interface_minus_center"] = out["SASA_interface_mean"] - out["SASA_Avg_z0"].astype(float)

    # vdW: 最も安定なvdW相互作用
    if all(f"vdW_Avg_z{z}" in out.columns for z in z_list):
        vdw_cols = [f"vdW_Avg_z{z}" for z in z_list]
        vdw_arr = out[vdw_cols].astype(float).values
        out["vdW_best_min"] = np.nanmin(vdw_arr, axis=1)
        out["vdW_best_minus_water"] = np.nanmin(vdw_arr, axis=1) - out["vdW_Avg_z20"].astype(float)
        out["vdW_range_physical"] = np.nanmax(vdw_arr, axis=1) - np.nanmin(vdw_arr, axis=1)

    # Coul: z方向でどれだけ変わるか
    if all(f"Coul_Avg_z{z}" in out.columns for z in z_list):
        coul_cols = [f"Coul_Avg_z{z}" for z in z_list]
        coul_arr = out[coul_cols].astype(float).values
        out["Coul_range_physical"] = np.nanmax(coul_arr, axis=1) - np.nanmin(coul_arr, axis=1)
        out["Coul_center_minus_water"] = out["Coul_Avg_z0"].astype(float) - out["Coul_Avg_z20"].astype(float)

    return out


def prepare_dataset(name, path):
    df = pd.read_csv(path, low_memory=False)

    print("\n" + "=" * 80)
    print(name)
    print("=" * 80)
    print("path :", path)
    print("shape:", df.shape)

    target_col = find_target_column(df)
    raw_cols, missing = get_raw_md_cols(df)

    if missing:
        print("missing raw MD columns:")
        for c in missing:
            print("  -", c)

    if len(raw_cols) != 25:
        print(f"⚠️ raw MD columns = {len(raw_cols)}。25列そろっていません。")

    keep_cols = ["ID", target_col] + raw_cols
    use_df = df[keep_cols].copy()

    use_df["ID"] = pd.to_numeric(use_df["ID"], errors="coerce")
    use_df[target_col] = pd.to_numeric(use_df[target_col], errors="coerce")

    for c in raw_cols:
        use_df[c] = pd.to_numeric(use_df[c], errors="coerce")

    before = len(use_df)

    use_df = use_df.dropna(subset=["ID", target_col] + raw_cols).copy()
    use_df["ID"] = use_df["ID"].astype(int)

    # cons100の大規模データはID1-100だけに制限
    if name == "GBSW_cons100":
        use_df = use_df[(use_df["ID"] >= 1) & (use_df["ID"] <= 100)].copy()

    # ID重複があれば最後を採用
    use_df = use_df.drop_duplicates(subset=["ID"], keep="last").copy()
    use_df = use_df.rename(columns={target_col: "y"})

    # 派生特徴量追加
    use_df = add_derived_features(use_df)

    # 派生特徴量列
    derived_cols = [
        c for c in use_df.columns
        if c not in ["ID", "y"] + raw_cols
    ]

    # 数値列だけに限定
    derived_cols = [
        c for c in derived_cols
        if pd.api.types.is_numeric_dtype(use_df[c])
    ]

    # infをNaNへ
    use_df[derived_cols] = use_df[derived_cols].replace([np.inf, -np.inf], np.nan)

    print("target_col:", target_col)
    print("raw feature N     :", len(raw_cols))
    print("derived feature N :", len(derived_cols))
    print("N before dropna   :", before)
    print("N usable          :", len(use_df))
    print("ID range          :", use_df["ID"].min(), "~", use_df["ID"].max())

    return use_df, raw_cols, derived_cols


def evaluate_repeated(df, feature_cols, dataset_name, feature_mode, common_ids):
    use_df = df[df["ID"].isin(common_ids)].copy()

    # featureに欠損がある行は落とす
    use_df = use_df.dropna(subset=["y"] + feature_cols).copy()
    use_df = use_df.sort_values("ID").reset_index(drop=True)

    X = use_df[feature_cols].values
    y = use_df["y"].values

    rows = []
    pred_rows = []

    for seed in SEEDS:
        cv = KFold(
            n_splits=N_SPLITS,
            shuffle=True,
            random_state=seed
        )

        rf = RandomForestRegressor(
            n_estimators=500,
            random_state=seed,
            min_samples_leaf=2,
            n_jobs=-1
        )

        y_pred = cross_val_predict(rf, X, y, cv=cv)

        r, p = pearsonr(y, y_pred)
        r2 = r2_score(y, y_pred)
        mae = mean_absolute_error(y, y_pred)
        rmse = np.sqrt(mean_squared_error(y, y_pred))

        rows.append({
            "Dataset": dataset_name,
            "Feature_Mode": feature_mode,
            "Seed": seed,
            "N": len(use_df),
            "Feature_N": len(feature_cols),
            "Pearson_R": r,
            "R2": r2,
            "MAE": mae,
            "RMSE": rmse,
            "p_value": p,
        })

        tmp_pred = pd.DataFrame({
            "Dataset": dataset_name,
            "Feature_Mode": feature_mode,
            "Seed": seed,
            "ID": use_df["ID"].values,
            "y_true": y,
            "y_pred": y_pred,
            "error": y_pred - y,
        })
        pred_rows.append(tmp_pred)

    return pd.DataFrame(rows), pd.concat(pred_rows, ignore_index=True)


# =========================================================
# 読み込み
# =========================================================

datasets = {}
raw_feature_sets = {}
derived_feature_sets = {}

for name, path in PATHS.items():
    df_use, raw_cols, derived_cols = prepare_dataset(name, path)
    datasets[name] = df_use
    raw_feature_sets[name] = raw_cols
    derived_feature_sets[name] = derived_cols


# =========================================================
# 共通ID
# =========================================================

common_ids = set(datasets["GBSW_cons100"]["ID"])
common_ids &= set(datasets["GBSW_cons50"]["ID"])
common_ids &= set(datasets["GBSW_cons1"]["ID"])
common_ids = sorted(common_ids)

print("\n" + "=" * 80)
print("Common ID")
print("=" * 80)
for name, df in datasets.items():
    print(name, "usable N:", len(df))

print("Common ID N:", len(common_ids))
print("Common ID range:", min(common_ids), "~", max(common_ids))


# =========================================================
# 評価
# =========================================================

all_results = []
all_preds = []

feature_modes = ["raw_MD", "derived_MD", "raw_plus_derived_MD"]

for dataset_name in ["GBSW_cons100", "GBSW_cons50", "GBSW_cons1"]:
    for feature_mode in feature_modes:

        raw_cols = raw_feature_sets[dataset_name]
        derived_cols = derived_feature_sets[dataset_name]

        if feature_mode == "raw_MD":
            feature_cols = raw_cols
        elif feature_mode == "derived_MD":
            feature_cols = derived_cols
        elif feature_mode == "raw_plus_derived_MD":
            feature_cols = raw_cols + derived_cols
        else:
            raise ValueError(feature_mode)

        result_df, pred_df = evaluate_repeated(
            df=datasets[dataset_name],
            feature_cols=feature_cols,
            dataset_name=dataset_name,
            feature_mode=feature_mode,
            common_ids=common_ids
        )

        all_results.append(result_df)
        all_preds.append(pred_df)

results_all = pd.concat(all_results, ignore_index=True)
preds_all = pd.concat(all_preds, ignore_index=True)

summary = (
    results_all
    .groupby(["Dataset", "Feature_Mode"])
    .agg(
        N=("N", "first"),
        Feature_N=("Feature_N", "first"),
        R_mean=("Pearson_R", "mean"),
        R_std=("Pearson_R", "std"),
        R2_mean=("R2", "mean"),
        R2_std=("R2", "std"),
        MAE_mean=("MAE", "mean"),
        MAE_std=("MAE", "std"),
        RMSE_mean=("RMSE", "mean"),
        RMSE_std=("RMSE", "std"),
    )
    .reset_index()
)

summary = summary.sort_values(
    ["Feature_Mode", "R_mean"],
    ascending=[True, False]
).reset_index(drop=True)

display(summary)

# 保存
summary.to_csv(OUT_SUMMARY, index=False)
results_all.to_csv(OUT_ALLSEEDS, index=False)
preds_all.to_csv(OUT_PRED, index=False)

print("\n✅ saved:")
print(OUT_SUMMARY)
print(OUT_ALLSEEDS)
print(OUT_PRED)
```
