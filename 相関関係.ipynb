```python
#オレンジと青の相関関係（Lys-Asn-Ser-Tyr-Phe）
```

```python
import pandas as pd
```

```python
# データフレームを作成（例: "energy" と "hbonds" の列があるとする）
df = pd.read_csv("/Users/teraimao/experiment/Lys-Asn-Ser-Tyr-Pheの相関関係のコピー.csv")
```

```python
# ピアソン相関

pearson_r = df["溶媒和自由エネルギー"].corr(df["水素結合数"], method="pearson")

# スピアマン相関
spearman_r = df["溶媒和自由エネルギー"].corr(df["水素結合数"], method="spearman")

print("Pearson相関:", pearson_r)
print("Spearman相関:", spearman_r)
```
